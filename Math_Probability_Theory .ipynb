## Ehtimal Nəzəriyyəsi:

### Ehtimal Nəzəriyyəsi:  
**Ehtimal nəzəriyyəsi** riyazi bir sahədir və hadisələrin başvermə ehtimallarını öyrənməklə məşğuldur. Bu nəzəriyyə qeyri-müəyyənliklə bağlı problemləri həll etməyə və təsadüfi dəyişənlərin davranışını analiz etməyə imkan verir. 

---

### Qısa İzah:  
Ehtimal nəzəriyyəsi hadisələrin müəyyən şərtlər altında başvermə ehtimallarını riyazi model vasitəsilə qiymətləndirir. Əsas məqsədi real həyatda müşahidə edilən qeyri-müəyyən prosesləri sistemləşdirmək və onlar haqqında dəqiq proqnozlar verməkdir.

---

### Geniş İzah:  
Ehtimal nəzəriyyəsi təsadüfi hadisələri riyazi olaraq təsvir etməyə çalışır. Təsadüfi hadisələr - başverməsi əvvəlcədən tam dəqiqliklə təxmin edilə bilməyən hadisələrdir. Məsələn, bir sikkənin atılması zamanı "gerb" və ya "pulu" çıxma ehtimalını nəzərdən keçirək: burada hər bir nəticənin baş vermə ehtimalı 0.5-dir. 

Ehtimal nəzəriyyəsi real dünyadakı müxtəlif təsadüfi prosesləri modelləşdirmək və təhlil etmək üçün istifadə olunur. Statistik analiz, süni intellekt, maliyyə analitikası və müxtəlif sənaye sahələrində geniş tətbiq sahəsi tapır. 

---

### Əsas Növləri:  

1. **Klassik Ehtimal:** 
   - Bütün mümkün nəticələrin bərabər ehtimallı olduğu hallarda tətbiq edilir.
   - Məsələn, ədalətli bir zar atıldıqda hər rəqəmin (1-dən 6-ya) çıxma ehtimalı eynidir.

2. **Empirik Ehtimal:** 
   - Tarixi və ya müşahidə olunan məlumatlardan istifadə edərək hesablanır.
   - Məsələn, bir şəhərdə yağış yağan günlərin statistikası əsasında yağış ehtimalı.

3. **Subyektiv Ehtimal:**
   - Fərdi şəxsin bir hadisənin baş verəcəyinə olan inamı ilə əlaqəlidir.
   - Məsələn, hava proqnozçusunun "sabah yağış yağma ehtimalı 70%-dir" deməsi.

4. **Şərti Ehtimal:**
   - Bir hadisənin baş verməsi başqa bir hadisənin baş verməsinə bağlıdır.
   - Məsələn, bir tələbənin universitetə qəbul olması, onun imtahandan keçməsinə bağlıdır.

---

### Data Science və Data Analitika üçün Əhəmiyyəti:  

1. **Model Qurulması:**
   - Təsadüfi dəyişənlərin davranışını təhlil etmək üçün ehtimal nəzəriyyəsi modelləri (məsələn, Gaussian paylanması) istifadə olunur.

2. **Statistik Testlər:**
   - Hipotez yoxlanılması zamanı ehtimal nəzəriyyəsi vasitəsilə nəticələrin təsadüfən baş vermə ehtimalı qiymətləndirilir.

3. **Maşın Öyrənməsi:**
   - Ehtimal paylanmaları, Naive Bayes kimi ehtimala əsaslanan modellərdə tətbiq olunur.

4. **Risk Təhlili:**
   - Maliyyə modelləşdirməsində və risklərin proqnozlaşdırılmasında ehtimal hesablamaları əsas rol oynayır.

5. **Təsadüfi Nümunə Götürmə:**
   - Analitikdə böyük datasetlər üzərində təsadüfi nümunə seçərək analizin sürətini artırmaq üçün istifadə edilir.

---

Ehtimal nəzəriyyəsi, sadəcə qeyri-müəyyənliklərlə işləməyi öyrətməklə qalmır, həm də real həyatda proqnozların dəqiqliyini artırmaq üçün etibarlı bir təməl yaradır.

### **Ehtimal Nəzəriyyəsinin Geniş İzahı**

Ehtimal nəzəriyyəsi qeyri-müəyyənliklərlə işləməyi öyrənən riyazi bir sahədir. Bu nəzəriyyənin kökü, təsadüfi hadisələrin təbiətini başa düşmək və onların baş vermə ehtimalını hesablayaraq gələcəkdə baş verə biləcək hadisələr haqqında proqnozlar verməkdən ibarətdir. Riyazi ehtimal nəzəriyyəsi, statistikanın təməli olaraq qəbul edilir və geniş tətbiq sahəsinə malikdir: təbiət elmlərindən maliyyə təhlilinə, süni intellektdən tibbi diaqnostikaya qədər bir çox sahədə istifadə edilir.

#### **Əsas Terminlər**

1. **Təsadüfi Hadisə:**  
   Müşahidələr zamanı nəticəsi əvvəlcədən müəyyən edilə bilməyən hadisələr. Məsələn, zar atıldıqda "6" çıxması bir təsadüfi hadisədir.

2. **Nəticə Məkanı (Sample Space, $ S $):**  
   Bütün mümkün nəticələrin toplusudur.  
   - Məsələn, bir zar atıldıqda nəticə məkanı $ S = \{1, 2, 3, 4, 5, 6\} $.

3. **Ehtimal ($ P $):**  
   Bir hadisənin baş vermə şansı.  
   Riyazi olaraq:  
   $$
   P(A) = \frac{\text{Hadisə üçün əlverişli nəticələrin sayı}}{\text{Ümumi nəticə sayı}}
   $$  
   Burada $ P(A) $ 0 ilə 1 arasında bir dəyərdir. $ P(A) = 0 $ hadisənin mümkünsüz, $ P(A) = 1 $ isə hadisənin mütləq baş verəcəyini göstərir.

#### **Əsas Prinsiplər**

1. **Ehtimalların Cəmi:**  
   Nəticə məkanındakı bütün mümkün nəticələrin ehtimallarının cəmi həmişə 1-ə bərabərdir.  
   $$
   \sum P(A_i) = 1
   $$

2. **Bir-birindən Asılı və Asılı Olmayan Hadisələr:**  
   - **Asılı Hadisələr:** Bir hadisənin baş verməsi digərinə təsir edir.  
     Məsələn, bir kart dəstəsindən çıxarılan bir kartın növbəti seçimi təsir etməsi.  
   - **Asılı Olmayan Hadisələr:** Bir hadisənin baş verməsi digərinə təsir etmir.  
     Məsələn, iki fərqli sikkənin eyni anda atılması.

3. **Şərti Ehtimal ($ P(A|B) $):**  
   Bir hadisənin baş vermə ehtimalı, digər bir hadisənin artıq baş verdiyi halda hesablanır. Riyazi ifadəsi:  
   $$
   P(A|B) = \frac{P(A \cap B)}{P(B)}
   $$  
   Burada $ P(A \cap B) $ həm \$ A $, həm də $ B $-nin baş vermə ehtimalıdır.

4. **Bayes Teoremi:**  
   Ehtimal nəzəriyyəsində şərti ehtimalların hesablanmasında mühüm alətdir. Bayes teoremi belə ifadə olunur:  
   $$
   P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)}
   $$  
   Burada $ P(A|B) $, $ A $-nın $ B $-yə bağlı olaraq baş vermə ehtimalıdır.

### **Ehtimal Nəzəriyyəsinin Tətbiqi Sahələri**

1. **Statistika və Hipotez Yoxlanışı:**  
   - Statistik testlərdə ehtimal nəzəriyyəsi istifadə olunur. Məsələn, $ p $-dəyərinin (statistik əhəmiyyətin) hesablanması.

2. **Maşın Öyrənməsi və Süni İntellekt:**  
   - Maşın öyrənməsi modellərində ehtimallar əsas rol oynayır. Xüsusilə Naive Bayes, Logistic Regression kimi modellər ehtimal anlayışını əsas götürür.  

3. **Maliyyə Riskləri:**  
   - Bank və sığorta sahəsində ehtimallar risklərin təhlilində və gələcək hadisələrin modelləşdirilməsində istifadə olunur.

4. **Marketinq və Müşahidəçi Davranışı:**  
   - Müştərilərin davranışlarının təxmin edilməsi üçün ehtimal modelləri tətbiq edilir. Məsələn, bir müştərinin məhsul alması ehtimalını təxmin etmək.

5. **Tibbi Tədqiqatlar:**  
   - Testlərin dəqiqliyini ölçmək üçün ehtimal nəzəriyyəsindən istifadə olunur (məsələn, xəstəlik diaqnostikasında yanlış pozitiv və yanlış neqativ ehtimallar).

---

### **Ehtimal Nəzəriyyəsinin Növləri**

1. **Diskret Ehtimal:**  
   - Nəticə məkanı sonlu və ya sayılabilən sonsuzdur. Məsələn, zarın atılması.

2. **Davamlı Ehtimal:**  
   - Nəticə məkanı sonsuz çoxluğa malikdir və ehtimal paylanma sıxlığı funksiyası ($ PDF $) ilə təsvir olunur. Məsələn, temperatur ölçüləri.

3. **Birləşmə Paylanmaları:**  
   - Birdən çox təsadüfi dəyişənin ehtimallarını nəzərə alan modellər. Məsələn, iki fərqli məhsulun satış ehtimalı.

---

### **Data Science və Data Analitika Üçün Əhəmiyyəti**

1. **Təsadüfi Proseslərin Modelləşdirilməsi:**  
   - Data science sahəsində ehtimal nəzəriyyəsi verilənlərin təsadüfi təbiətini başa düşmək və modelləşdirmək üçün istifadə olunur.

2. **Korelyasiya və Kovariasiya:**  
   - Verilənlər arasında əlaqələrin təsviri üçün ehtimal anlayışları tətbiq edilir.

3. **Paylanmaların Təhlili:**  
   - Normal, Binomial, Poisson və digər paylanmaların təhlili verilənlərin strukturu haqqında məlumat verir.

4. **Süni Öyrənmə Modelləri:**  
   - Logistic Regression və Naive Bayes kimi modellərdə ehtimal nəzəriyyəsi istifadə olunur. XGBoost və Random Forest kimi modellərdə də təsadüfilik elementləri daxildir.

5. **Monte Carlo Simulyasiyası:**  
   - Verilənlərdə təsadüfi dəyişikliklər üzərində proqnozlar yaratmaq üçün istifadə olunan simulyasiya metodudur.

---

Ehtimal nəzəriyyəsi, data scientist və analitiklər üçün həyati bir bacarıqdır. Həm model qurma, həm də nəticələri düzgün şərh etmək üçün bu sahənin əsas prinsiplərini anlamaq vacibdir.
