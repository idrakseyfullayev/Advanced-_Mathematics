## Qalıq tapma qaydası Python-da (%)

Python-da `%` operatoru **qalıq tapmaq** üçün istifadə olunur. Amma bu operatorun necə işlədiyi ilə bağlı bəzi incə məqamlar var. Gəlin məsələyə aydınlıq gətirək.

---

### **Qalıq tapma qaydası Python-da**
Python-da `%` operatoru belə işləyir:
$$
\text{qalıq} = \text{bölünən} - (\text{tam bölmə} \times \text{bölən})
$$

Bu, **tam bölmə** əməliyyatına əsaslanır və nəticə həmişə **bölənin işarəsi ilə eyni işarəyə** malik olur. Bu qayda Python-da dizayn qərarıdır və digər proqramlaşdırma dillərində fərqli ola bilər.

---

### **Məsələn, -17 % 3**
1. Tam bölməni tapırıq:
   $$
   \text{Tam bölmə} = \text{floor}(-17 / 3) = -6 \quad (\text{yuxarı yuvarlanmır, aşağı yuvarlanır})
   $$

2. Qalıq formuluna əsasən:
   $$
   \text{qalıq} = -17 - (-6 \times 3)
   $$
   $$
   \text{qalıq} = -17 - (-18) = 1
   $$

Nəticə:
$$
-17 \% 3 = 1
$$

---

### **Niyə -2 deyil?**
Bəzi dillərdə qalıq **bölünənin işarəsini** götürür. Python-da isə **qalıq bölənin işarəsini** izləyir. Bu səbəbdən nəticə $1$ olur, çünki $3$ müsbətdir.

---

### **Qarşılaşdırma:**
| Əməliyyat      | Python-da Nəticə | Bəzi Digər Dillərdə (C, C++) |
|-----------------|------------------|-----------------------------|
| \(-17 \% 3\)   | \(1\)            | \(-2\)                     |
| \(-17 \% -3\)  | \(-2\)           | \(-2\)                     |
| \(17 \% -3\)   | \(-2\)           | \(2\)                      |

---

### **Əgər nəticənin -2 olmasını istəyirsinizsə?**
Əgər qalıq həmişə bölünənin işarəsini götürməli olarsa, bu şəkildə hesablaya bilərsiniz:

```python
remainder = -17 % 3
if remainder and -17 < 0:
    remainder -= 3
print(remainder)  # -2
```

---

### **Son Qeyd**
Python dizayn baxımından "qalıq bölənin işarəsini izləməlidir" yanaşmasını seçib. Bu, bölmə və mod operatorunun uyğunluğu üçün edilir, çünki belə bir əlaqə var:
$$
a = b \times (a \div b) + (a \% b)
$$

Əlavə izaha ehtiyacınız varsa, mənə bildirin! 😊

##   $2^0 = 1$

$$
2^0 = 1
$$
bunun səbəbi **qüvvət funksiyasının əsas riyazi qaydalarına** əsaslanır. Qüvvət funksiyasını idarə edən qaydalar ardıcıl və məntiqi olduğundan, bu nəticə belə alınır.

### **Qüvvət Qaydası:**
Qüvvətlərin bölmə qaydasını nəzərdən keçirək:
$$
a^m \div a^n = a^{m-n}
$$

Məsələn, $2^3$ və $2^2$ üçün:
$$
\frac{2^3}{2^2} = 2^{3-2} = 2^1 = 2
$$

Bu qaydanı davam etdirərək \(m = n\) olduğu halı nəzərdən keçirək:
$$
\frac{2^n}{2^n} = 2^{n-n} = 2^0
$$

Amma $ \frac{2^n}{2^n} = 1$, çünki hər hansı bir ədədin özü-özünə bölünməsi həmişə $1$-ə bərabərdir. Buna görə:
$$
2^0 = 1
$$

### **Ümumi Məntiq**
1. **Hər hansı bir ədəd öz-özünə bölünəndə 1 olur:**
   $$
   a^0 = \frac{a^1}{a^1} = 1
   $$

2. **Qüvvət funksiyasının davamlılığı:** Riyaziyyatda $a^x$ funksiyası ardıcıl olaraq artır və ya azalır. Bu qaydanı qorumaq üçün $a^0 = 1$ təyin edilir.

3. **Boş hasil:**
   Riyaziyyatda heç bir ədədi hasilləmədikdə nəticə 1 hesab olunur. Məsələn, kombinatorikada $0!$ faktorialı 1-ə bərabərdir. Oxşar məntiqlə $2^0 = 1$ hesab olunur.

---

### **Nəticə**
$$
2^0 = 1
$$
bu, qüvvət funksiyasının təməl qaydalarından irəli gəlir və riyazi ardıcıllığı qorumaq üçün məntiqlidir. Əgər əlavə sualınız varsa, izah etməyə davam edə bilərəm! 😊

## **0! = 1**

**0! = 1** olması riyaziyyatın əsas qaydalarından biridir və bunun bir neçə məntiqli izahı var. Gəlin həm riyazi, həm də praktik baxımdan bunu araşdıraq.

---

### **1. Riyazi Təriflərdən Yanaşma**
Faktorial $n!$ ədədi $n$-dən 1-ə qədər olan bütün natural ədədlərin hasilini ifadə edir:
$$
n! = n \cdot (n-1) \cdot (n-2) \cdot \ldots \cdot 1
$$

Amma bu qayda **$0!$** üçün birbaşa işləmir, çünki $0$-dan geriyə saymaq mümkün deyil. Bunun üçün faktorialın başqa tərifindən istifadə edirik.

#### **Rekursiv Tərif**
Faktorial belə tərif olunur:
$$
n! = n \cdot (n-1)!
$$

Məsələn:
$$
5! = 5 \cdot 4! \quad \text{və} \quad 4! = 4 \cdot 3! \quad \text{və s.}
$$

Bu tərifi $n = 1$-ə qədər davam etdirib $0!$-ə gəlirik:
$$
1! = 1 \cdot 0!
$$
Buradan:
$$
0! = 1
$$

---

### **2. Kombinatorikadan Yanaşma**
Faktorial kombinatorikada mümkün düzülüşləri (kombinasiyaları) hesablamaq üçün istifadə olunur. Məsələn:
$$
n! \quad \text{= \(n\) elementin fərqli düzülüşlərinin sayı.}
$$

Amma **heç bir element olmadıqda (0 element)**, yalnız **bir yolla düzülüş** mövcuddur: **heç nə etməmək!**  
Bu halda:
$$
0! = 1
$$

---

### **3. Boş Hasil Məntiqi**
Riyaziyyatda boş hasil (heç bir ədədin hasilini götürmək) həmişə **1** olaraq qəbul edilir. Bu qayda faktorialın məntiqi davamlılığını təmin etmək üçün işləyir.  
Məsələn:
$$
1! = 1 \cdot 0! \quad \text{və bu, 0! = 1 olmasını tələb edir.}
$$

---

### **4. Gamma Funksiyası ilə Sübut**
Faktorialın ümumiləşdirilmiş tərifi **Gamma funksiyası** vasitəsilə verilir:
$$
n! = \Gamma(n+1)
$$

Gamma funksiyası belə tərif olunur:
$$
\Gamma(x) = \int_{0}^{\infty} t^{x-1} e^{-t} dt
$$

$n = 0$ üçün:
$$
0! = \Gamma(1) = \int_{0}^{\infty} e^{-t} dt = 1
$$

---

### **Nəticə**
Faktorialın 0 üçün **1** olması həm riyazi qaydalara əsaslanır, həm də praktiki kombinatorikaya və boş hasil anlayışına uyğun gəlir. Bu, faktorial funksiyasının məntiqli şəkildə genişləndirilməsi üçün vacibdir. 

Əlavə izaha ehtiyac varsa, çəkinmədən soruşun! 😊

## Function

Riyaziyyatda funksiya iki çoxluq arasında müəyyən bir qaydaya əsaslanan əlaqədir. Formal olaraq, funksiya belə tərif edilir:  

> **Funksiya**: Təyin olunma sahəsindəki hər bir elementə qiymət sahəsindən yalnız bir element uyğunlaşdıran qaydaya funksiya deyilir.

### Funksiyanın Tərkib Hissələri
1. **Təyin olunma sahəsi (Domain)**: Funksiyanın qəbul etdiyi bütün giriş dəyərlərinin çoxluğu. Məsələn, $ x $-in ala biləcəyi bütün mümkün qiymətlər.
2. **Qiymət sahəsi (Range)**: Funksiyanın qaytara biləcəyi çıxışların çoxluğu.
3. **Qayda**: Funksiyanın giriş dəyərlərinə uyğun olaraq çıxışları necə hesablamağını göstərir. Bu qayda riyazi ifadə, qrafik və ya cədvəl şəklində verilə bilər.

### Funksiya Necə İşləyir?
Funksiya ümumiyyətlə belə yazılır:
$$
f(x)
$$
Burada:
- $ f $ funksiyanın adıdır.
- $ x $ funksiyanın giriş (argument) dəyəridir.
- $ f(x) $ isə həmin giriş üçün çıxış dəyəridir.

#### Məsələn:
Verilmiş funksiya $ f(x) = 2x + 3 $ olsun.  
Burada:
- Təyin olunma sahəsi \( x \)-in ala biləcəyi dəyərlərdir (məsələn, bütün real ədədlər).
- Qiymət sahəsi \( 2x + 3 \)-ün qaytara biləcəyi bütün dəyərlərdir.

Girişlər və çıxışlar:
- $ x = 1 $ üçün: $ f(1) = 2 \cdot 1 + 3 = 5 $
- $ x = -2 $ üçün: $ f(-2) = 2 \cdot (-2) + 3 = -1 $

### Funksiyanın Qrafiki
Funksiya, təyin olunma sahəsindəki hər $ x $ üçün bir $ y $ dəyəri ilə əlaqə yaratdığı üçün bu əlaqə müstəvidə qrafik kimi göstərilə bilər. Məsələn:
- $ f(x) = x^2 $: Bu parabol şəklində bir qrafikdir.
- $ f(x) = 2x + 3 $: Bu isə bir xətt funksiyasıdır.

### Funksiyanın Xüsusiyyətləri
1. **Birdəfəliklik**: Təyin olunma sahəsindəki hər bir $ x $ dəyərinə qiymət sahəsindən yalnız **bir** $ y $ dəyəri uyğun gəlir.
2. **Münasibət**: Hər bir funksiya bir münasibətdir, amma hər münasibət funksiya deyil (çünki birdəfəlik şərt ödənməlidir).

#### Qeyri-müəyyən münasibət nümunəsi (Funksiya deyil):
$$
x = 2 \, \Rightarrow \, y = \{-1, 1\}
$$
Burada $ x = 2 $ üçün iki fərqli $ y $ dəyəri mövcuddur. Bu səbəbdən bu, funksiya deyil.

### Funksiya Riyaziyyatın Harasında Tətbiq Olunur?
Funksiyalar riyaziyyatın demək olar ki, bütün sahələrində istifadə edilir:
- **Cəbr**: Düzxətli və kvadrat funksiyalar.
- **Analiz**: Limiti, törəməsi və inteqralı hesablanan funksiyalar.
- **Həndəsə**: Əyrilər və qrafiklər.
- **Statistika və Tətbiqi Riyaziyyat**: Qeyri-xətti modellər, ehtimallar.

### Yekun
Funksiya, riyazi olaraq, "bir şeyi başqa bir şeylə əlaqələndirən qayda" kimi başa düşülə bilər. Hər $ x $-ə yalnız bir $ y $ uyğun gəlməlidir. Bu, həm gündəlik həyatda, həm də elmi tədqiqatlarda məlumatları analiz etmək üçün əsas vasitələrdən biridir.

## Logarithm

**Ali Riyaziyyatda Loqarifm** (loqarifma) funksiyası, əsasən **qüvvətlərin tərsi** kimi təsvir edilə bilər. Loqarifm, bir ədədin bir əsas üçün hansı qüvvətə bərabər olduğunu tapmağa kömək edir. Başqa sözlə, loqarifm bir ədədin, müəyyən bir əsasla hansı qüvvətə yüksəldilərək verilən ədədə çatacağını müəyyən edir.

**Loqarifm** riyaziyyatda bir ədədin, müəyyən bir əsas üçün, qüvvət şəklində ifadə olunmasını təyin edən bir funksiyadır. 
> **Loqarifm**: Bir ədədin verilmiş əsasın hansı qüvvətinə bərabər olduğunu müəyyən edir.

### **Loqarifmin Tərifi**
Əgər $a > 0$, $a \neq 1$ və $x > 0$ olarsa, o zaman loqarifm belə təyin edilir:

$$
y = \log_a(x)
$$
Bu tərifi $a$-nın **əsas** olduğu loqarifm olaraq oxuyuruq. Bu, deməkdir ki:

$$
a^y = x
$$

Burada:
- **$a$**: Loqarifmin əsasını təşkil edir (müsbət, amma 1-dən fərqli olmalıdır).
- **$x$**: Loqarifm dəyəri tapılacaq ədəd.
- **$y$**: $x$-in $a$-ya əsasən loqarifmi olan nəticədir, yəni $x$-in $a$-ya hansı qüvvətə yüksəldilməli olduğunu göstərir.

### **Loqarifmin Xüsusi Əsasları**
1. **Adi Loqarifm (Common Logarithm)**:
   - Əsas $10$-dur, yəni $\log_{10}(x)$.
   - Ən çox istifadə olunan loqarifm növüdür.
   - Məsələn: $\log_{10}(1000) = 3$, çünki $10^3 = 1000$.

2. **Təbii Loqarifm (Natural Logarithm)**:
   - Əsas $e$-dir, yəni $\ln(x)$ (təqribən $e \approx 2.718$).
   - Təbiətdə bir çox təbii proseslər, məsələn, əhali artımı, radiasiya, və sair təbii proseslər təbii loqarifm ilə modelləşdirilir.
   - Məsələn: $\ln(e^3) = 3$, çünki $e^3 = e^3$.

3. **Loqarifm Əsasları Arasında Əlaqə (Change of Base Formula)**:
   Əgər $a$-dan $b$-yə loqarifmi hesablamaq istəyiriksə, onda **əsas dəyişdirmə** qaydası belədir:
   $$
   \log_a(x) = \frac{\log_b(x)}{\log_b(a)}
   $$
   Bu qayda, loqarifmi istənilən əsasdan başqa bir əsasda hesablamağa imkan verir.

### **Loqarifmin Əsas Xassələri**
Loqarifm funksiyasının bəzi mühüm xassələri və tərifləri:
1. **Çarpmanın Loqarifmi**:
   $$
   \log_a(xy) = \log_a(x) + \log_a(y)
   $$
   Yəni, iki ədədin çarpımının loqarifmi, bu ədədlərin loqarifmlərinin cəmindən ibarətdir.

2. **Bölmənin Loqarifmi**:
   $$
   \log_a\left(\frac{x}{y}\right) = \log_a(x) - \log_a(y)
   $$
   Yəni, iki ədədin bölməsinin loqarifmi, bu ədədlərin loqarifmlərinin fərqinə bərabərdir.

3. **Qüvvətin Loqarifmi**:
   $$
   \log_a(x^n) = n \cdot \log_a(x)
   $$
   Yəni, bir ədədin qüvvətinin loqarifmi, həmin ədədin loqarifminin həmin qüvvətlə vurulmuş halıdır.

4. **Əsas və Argüment Uyğunluğu**:
   $$
   \log_a(a) = 1, \quad \log_a(1) = 0
   $$
   Bu xüsusiyyətlər, $a$-nın əsas olaraq qəbul edildiyi halda çox istifadə olunur.

---

### **Loqarifmin Qrafiki**
Loqarifm funksiyasının qrafiki əsas $a > 1$ olduqda **artandır** və $0 < a < 1$ olduqda **azalır**. Qrafiklərin digər əsas xüsusiyyətləri:
- **Təyin olunma sahəsi**: Loqarifmin yalnız müsbət ədədlər üçün təyin olunduğunu unutmamalıyıq, yəni $x > 0$.
- **Y oxu ilə yaxınlıq**: Loqarifm funksiyası $y$-oxuna yaxınlaşır, amma onu keçmir (asimptotik davranış göstərir).

### **Loqarifmin Riyazi və Elmi Əhəmiyyəti**
Loqarifm, riyaziyyatın, xüsusilə də **analiz**, **cəbri** və **eksponensial funksiyalar** sahələrində mühüm rol oynayır. Loqarifmin tətbiq olunduğu bəzi sahələr:
1. **Fizika və mühəndislik**: Zəlzələ gücü, səs intensivliyi (decibel), radiasiya və digər təbii proseslərdə.
2. **Kompleks Hesablama və Algoritm Analizi**: Algoritmlərin mürəkkəbliyini ifadə etməkdə (məsələn, $O(\log n)$).
3. **İnformasiya nəzəriyyəsi**: Məlumat sıxlığı və şifrələmə nəzəriyyələrində.
4. **Maliyyə**: Faiz hesablamaları və kapitalın eksponensial böyüməsi ilə bağlı məsələlərdə.

### **Nəticə**
Loqarifm, riyaziyyatın əsas funksiyalarından biridir və onun müxtəlif sahələrdə tətbiqləri çox genişdir. O, say sistemlərində, hesablamalarda, fizika və təbiət elmlərində fundamental rol oynayır. Loqarifmin əsası və xassələri, onun müxtəlif sahələrdə təbii olaraq istifadə olunmasına imkan verir.

## $\log_a x = b\$  
- \$a\$ həmişə \$a > 0\$ və \$a \neq 1\$ olmalıdır.

Logaritmik ifadədə baza $a = 1$ ola bilməməsinin əsas səbəbləri aşağıda daha dəqiq və ətraflı izah edilir:

---

### 1. **Logaritmin tərifi ilə ziddiyyət**  
Logaritmin tərifinə görə,  
$$\log_a x = b \implies a^b = x$$  
Bu ifadədə baza $a$ sıfırdan böyük $a > 0$ və 1-dən fərqli $a \neq 1$ bir real ədəd olmalıdır. Əgər baza $a = 1$ olsa, bu zaman:  
$$1^b = x$$  
İstənilən $b$ üçün $1^b = 1$ olduğuna görə, bu ifadə yalnız $x = 1$ halında keçərli olur. Bu isə logaritmin tərifini pozur, çünki logaritm hər hansı müsbət real $x > 0$ üçün müəyyən olmalıdır, amma burada $x = 1$-lə məhdudlaşır.

---

### 2. **Monotonluq pozulur**  
Logaritmik funksiyalar ya monoton **artan** (\(a > 1\)) olmalı, ya da monoton **azalan** (\(0 < a < 1\)) olmalıdır. Əgər baza \(a = 1\) olsa, logaritmik funksiya sabit olur:  
$$f(x) = \log_1 x \implies f(x) = 0 \quad (\text{bütün } x > 0 \text{ üçün})$$  
Bu halda logaritmik funksiyanın əsas xüsusiyyətlərindən biri olan monotonluq tamamilə pozulur. Sabit bir funksiya olaraq, dəyişənlər arasındakı əlaqəni ifadə edə bilmir.

---

### 3. **Real həyat tətbiqləri ilə uyğunsuzluq**  
Logaritm, eksponent funksiyanın inveri kimi istifadə olunur və əsasən:  
- Dəyişiklikləri ölçmək (böyümə və azalma),  
- Eksponent fenomenləri təsvir etmək (məsələn, əhali artımı, faiz hesablamaları) üçün istifadə edilir.  

Əgər baza $a = 1$ olsa, hər hansı eksponent dəyişiklik mümkün olmur, çünki $1^b$ həmişə sabit olaraq $1$-ə bərabərdir. Bu vəziyyət real həyatda istifadə üçün mənasızdır.

---

### 4. **Logaritmik şkalanın qeyri-mümkünlüyü**  
Logaritmik şkala müxtəlif dəyərlər arasında müqayisə aparmağa imkan verir. Lakin baza $a = 1$ olsa, logaritmik şkala mümkün olmaz, çünki:  
$\log_1 x = b \implies 1^b = x$  
və bu yalnız $x = 1$ üçün məna kəsb edir. Bu halda logaritm funksiyası digər dəyərləri əhatə etmir.

---

### Qısaca:  
- Əgər $a = 1$ olsa, logaritm funksiyası yalnız $x = 1$ üçün müəyyən olur.  
- Logaritm sabit funksiya olur və dəyişiklikləri təsvir edə bilmir.  
- Real dünyadakı tətbiqlər üçün mənasız hala gəlir.  

Bu səbəblərə görə baza $a$ həmişə $ > 0$ və $a \neq 1$ şərtlərini ödəməlidir.

## $ \log_a x = b $ 
- ($ x > 0 $) olmalıdır.

Logaritmik ifadə $ \log_a x = b $ olduğu zaman $ x $ müsbət bir ədəd olmalıdır ($ x > 0 $). Bunun əsas səbəbləri aşağıdakılardır:

---

### 1. **Logaritmin tərifi ilə əlaqə**

Logaritmin tərifinə görə:
$$ \log_a x = b \implies a^b = x $$

Buradan göründüyü kimi, $ x $ logaritmik ifadənin nəticəsidir və o, $ a^b $ ilə bağlıdır. Çünki baza $ a $ müsbət bir ədəd ($ a > 0 $) və $ a \neq 1 $ olduğu halda, $ a^b $ hər zaman müsbət bir ədəd olacaq. Başqa sözlə, $ a^b $ heç vaxt mənfi və ya sıfır ola bilməz. Buna görə də $ x $ da yalnız müsbət olmalıdır ($ x > 0 $).

---

### 2. **Əks eksponent funksiyası ilə əlaqə**

Logaritm eksponent funksiyasının tərsidir. Əgər eksponent funksiyasını $ y = a^x $ kimi düşünürüksə, $ a^x $ yalnız müsbət ədədlər ($ x > 0 $) verə bilər. Çünki:

- $ a^x > 0 $ hər zaman, istənilən $ x $-də olacaq (müsbət baza $ a > 0 $).

Bu, logaritmik funksiyanın yalnız müsbət $ x $ dəyərləri üçün işlədiyini göstərir.

---

### 3. **Real həyat tətbiqləri**

Logaritmik funksiyalar, çox vaxt real həyatda çoxluğa və ya miqyaslara əsaslanan hesablamalar üçün istifadə olunur (məsələn, əhali artımı, faiz hesablamaları). Bu cür tətbiqlərdə də $ x $ yalnız müsbət olmalıdır, çünki əhali və ya faiz kimi miqyaslar mənfi dəyərlər ala bilməz.

---

### 4. **Sıfırın logaritmi qeyri-mümkündür**

Əgər $ x = 0 $ olarsa, logaritmik ifadə $ \log_a 0 $ kimi bir şeyə çevrilər. Bu vəziyyətdə isə:

- $ a^b = 0 $ heç bir halda doğru ola bilməz, çünki heç bir müsbət $ a $ və $ b $-də $ a^b = 0 $ əldə edilə bilməz.

Buna görə də sıfırın logaritmi qeyri-mümkündür.

---

### Nəticə

Bu səbəblərə görə $ x $ həmişə müsbət bir ədəd olmalı, yəni $ x > 0 $ şərtini ödəməlidir.

---

## $ \log_a x = b$
- $a > 0$ olmalıdır.

Baza $a$ logaritmik ifadələrdə $a > 0$ olması niyə vacibdir? Bu sualın cavabını aşağıda izah edirəm:

---

### 1. **Logaritmin Tərifi ilə Ziddiyyət**
Logaritmin tərifinə görə:  
$$ \log_a x = b \implies a^b = x $$

Buradan aydın olur ki, baza $a$ müsbət bir ədəd olmalıdır. Əgər baza sıfır və ya mənfi bir ədəd olarsa, bu zaman logaritmik ifadənin tərifi məna kəsb etməz. Çünki:

- Əgər $a \leq 0$, $a^b$ ifadəsi real ədəd dünyasında müəyyən edilə bilməz. Mənfi ədədlərin və sıfırın qüvvətləri mənfi və ya qeyri-real ədəd dəyərləri verə bilər. Məsələn:
  - $(-1)^b$ qeyri-real nəticələr verə bilər, çünki mənfi bir ədədin qeyri-bütöv qüvvətləri (məsələn, $-1^{1/2}$) real ədəd yaratmaz.
  - $0^b$ qeyri-müəyyən olur, çünki sıfırın sıfırdan fərqli bir qüvvəti sıfırdır, amma sıfırın başqa bir qüvvəti müəyyən edilə bilməz.

Buna görə baza $a$ həmişə $a > 0$ olmalıdır.

---

### 2. **Eksponent Funksiyasının Tərsi Olması**
Logaritm, eksponent funksiyasının tərsidir və eksponent funksiyasının yalnız müsbət bazada ($a > 0$) mənalı nəticələr verməsi ilə bağlıdır. Əgər baza $a$ sıfır və ya mənfi olarsa, eksponent funksiyasının inverzini (logaritma) müəyyən etmək mümkün olmaz.

Məsələn:  
$$ a^b = x $$  
ifadəsi yalnız $a > 0$ və $a \neq 1$ olduqda düzgün və ardıcıl işləyir.

---

### 3. **Real Həyat Tətbiqləri**
Logaritmik funksiyalar real dünyada çoxluq, ölçü, artım və azalma kimi konseptləri təsvir etmək üçün istifadə olunur. Baza $a > 0$ olduqda bu tətbiqlər mənalı olur:

- Məsələn, əhali artımı və faiz artımı kimi modellərdə baza $a > 1$-dir və burada logaritm artımın sürətini və dəyişiklikləri ölçmək üçün istifadə edilir.
- Əgər baza $a \leq 0$ olsa, belə bir artım və azalma ölçülə bilməz, çünki mənfi ədədlərin və sıfırın qüvvətləri bu cür dəyişiklikləri düzgün göstərmir.

---

### 4. **Matematik Funksiyaların Xüsusiyyətləri**
Logaritmik funksiyalar monotonluq xüsusiyyətinə sahibdir:

- Əgər $a > 1$, logaritmik funksiya monoton artan olur.
- Əgər $0 < a < 1$, logaritmik funksiya monoton azalan olur.

Əgər baza $a$ sıfır və ya mənfi olarsa, bu monotonluq pozular və logaritmik funksiyanın düzgün işləməsi çətinləşər.

---

### Nəticə
Baza $a$ müsbət bir ədəd olmalıdır ($a > 0$), çünki:  
1. $a^b = x$ tərifinin işləməsi üçün $a > 0$ olmalıdır.  
2. Eksponent funksiyasının inverzi logaritm yalnız müsbət baza ilə mənalıdır.  
3. Real həyat tətbiqlərində dəyişiklikləri təsvir etmək və ölçmək üçün müsbət baza tələb olunur.  
4. Matematik funksiyaların düzgün işləməsi üçün $a > 0$ şərti vacibdir.  

---

## Eyler ədədi $e \approx 2.718281828459045...$

**Eyler ədədi** (Euler's number), $e$, riyaziyyatın bir çox sahəsində istifadə olunan xüsusi bir irrasiya (sonsuz və dövri olmayan onluq inkişafı olan) ədəddir. Bu ədədin təxmini dəyəri:

$$
e \approx 2.718281828459045...
$$

Eyler ədədi eksponent funksiyalar, logaritmlər, ehtimal nəzəriyyəsi, diferensial tənliklər, kompleks analiz, statistik modellər və bir çox digər riyazi və tətbiqi sahələrdə mühüm əhəmiyyətə malikdir.

---

### **Eyler ədədinin əsas tərifləri**
1. **Limit şəklində tərif**:
   Əylər ədədi aşağıdakı limit vasitəsilə təyin edilir:
   $$
   e = \lim_{n \to \infty} \left( 1 + \frac{1}{n} \right)^n
   $$
   Bu tərif, bir kapitalın faizlə böyümə prosesində istifadə edilir. Məsələn, faizlər sonsuz sayda kiçik dövrlərə bölündükdə məhz bu limit əldə olunur.

2. **Cəmin sonsuz sırası şəklində tərif**:
   $$
   e = \sum_{n=0}^\infty \frac{1}{n!} = 1 + \frac{1}{1!} + \frac{1}{2!} + \frac{1}{3!} + \cdots
   $$
   Burada $n!$ faktorialı bildirir ($n! = 1 \cdot 2 \cdot 3 \cdots n$).

3. **Eksponent funksiyasının tərifi**:
   Əylər ədədi, təbii eksponent funksiyanın bazasıdır:
   $$
   e^x = \sum_{n=0}^\infty \frac{x^n}{n!}
   $$
   Bu funksiyanın törəməsi və öz-özünə bərabərdir:
   $$
   \frac{d}{dx} e^x = e^x
   \$$

---

### **Eyler ədədinin tarixi**
- $e$ ədədini ilk dəfə 1618-ci ildə **John Napier** (logaritmlərin ixtiraçısı) öz işlərində istifadə etmişdir, lakin o zaman bu ədəd birbaşa $e$ kimi təyin edilməmişdi.
- $e$ ədədini müstəqil bir riyazi sabit kimi 1690-cı ildə **Jacob Bernoulli** faiz hesabatlarında kəşf etmişdir.
- Adını **Leonhard Euler**-dən almışdır, çünki Euler bu ədədin xüsusiyyətlərini dərin tədqiq etmiş və bu simvolu ($e$) istifadə etmişdir.

---

### **Eyler ədədinin xüsusiyyətləri**
1. **İrrasional və Transsendent ədəd**:
   - $e$ irrasional ədəddir, yəni onu iki tam ədədin nisbəti kimi ifadə etmək mümkün deyil.
   - O, həm də transsendent ədəddir, yəni onu heç bir cəbr tənliyinin kökü kimi ifadə etmək olmaz.

2. **Eksponent Funksiyanın Bazası**:
   $e^x$ funksiyası, eksponent funksiyaların əsasıdır və təbii logaritm bu baza əsaslanır:
   $$
   \ln(x) = \log_e(x)
   $$

3. **Fasiləsiz artım və faiz hesabı**:
   $e$ faiz və eksponent artım proseslərini təsvir edir. Məsələn, faiz hesablamada $e$ baza kimi istifadə olunur.

---

### **Tətbiq sahələri**
1. **Riyazi Modellər**:
   Eksponent artım və azalma proseslərini modelləşdirmək üçün (məsələn, əhali artımı, radioaktiv parçalanma).

2. **Fizika**:
   Eksponential funksiyalar dalğa, elektrik, və istilik ötürülməsi tənliklərində geniş istifadə olunur.

3. **Statistika və ehtimal nəzəriyyəsi**:
   Normal paylama, Poisson prosesi kimi statistik modellərdə $e$-nin xüsusiyyətləri istifadə edilir.

4. **Kompüter Elmləri**:
   Algoritmlərin analizi, xüsusilə $O(n \log n)$ kimi kompleksliklərin təhlili zamanı.

5. **Maliyyə**:
   Kompound faiz hesablamasında, $e$ fasiləsiz faizlə bağlıdır.

---

### **Nəticə**
Əylər ədədi ($e$) riyaziyyat və elmlərin bir çox sahəsində fundamental rola malikdir. Onun unikal xüsusiyyətləri təbii proseslərdə artım və dəyişikliklərin təsviri üçün əlverişlidir.

## Eyler ədədi və Fasiləsiz faiz 

Əylər ədədi ($e$) faiz hesablamasında "fasiləsiz faiz" prinsipi ilə bağlıdır. Bu anlayışı faizlərin hesablanma metodlarından izah edərək açıqlayaq:

---

### **Bank faizi və faiz hesablama üsulları**
Bank faizləri, bir kapitalın (pulun) müəyyən faizlə böyüməsini təsvir edir. Ümumi faiz hesablaması iki əsas üsuldan istifadə edir:

1. **Sadə faiz**:
   Faiz yalnız ilkin kapital üzərindən hesablanır. 
   $$
   M = P(1 + rt)
   $$
   Burada:
   - $M$: Yekun məbləğ,
   - $P$: İlkin kapital,
   - $r$: Faiz dərəcəsi (illik),
   - $t$: Vaxt (il ilə).

2. **Mürəkkəb faiz**:
   Faiz hər dövrün sonunda ilkin kapitala əlavə olunur, beləliklə, növbəti dövrdə faiz artan məbləğ üzərində hesablanır.
   $$
   M = P\left(1 + \frac{r}{n}\right)^{nt}
   $$
   Burada:
   - $n$: Bir il ərzində faiz dövrlərinin sayı.

---

### **Fasiləsiz faiz və Əylər ədədi**
Fasiləsiz faiz, faiz dövrlərinin sonsuz sayda olması deməkdir. Yəni faiz hər an hesablanır. Belə faiz üçün yekun məbləğ bu formul ilə təyin olunur:
$$
M = Pe^{rt}
$$
Burada:
- $P$: İlkin kapital,
- $r$: Faiz dərəcəsi (illik),
- $t$: Vaxt (il ilə),
- $e$: Əylər ədədi ($e \approx 2.718$).

---

### **Niyə $e$ istifadə olunur?**
Fasiləsiz faiz konsepti limitdən istifadə edir. Gəlin bunu izah edək:

Mürəkkəb faiz formulunda dövr sayını ($n$) artırdıqda, bu ifadə yaxınlaşır:
$$
M = P\left(1 + \frac{r}{n}\right)^{nt} \quad \text{və} \quad \lim_{n \to \infty} \left(1 + \frac{r}{n}\right)^{nt} = e^{rt}
$$
Yəni, faiz dövrlərinin sayı sonsuz dərəcədə artırılarsa, nəticə $e^{rt}$-yə yaxınlaşır. Bu, təbii artım proseslərinin təsviri üçün ən uyğun modeldir.

---

### **Praktiki nümunə**
Tutaq ki, bankda $P = 1000$ manat ilkin kapital yatırılıb və illik faiz dərəcəsi $r = 5\%$ ($0.05$) təşkil edir. Pul 3 il ərzində bankda saxlanılır.

#### **Mürəkkəb faizlə hesablanma**:
Əgər faiz hər rübdə ($n = 4$) hesablanırsa:
$$
M = 1000\left(1 + \frac{0.05}{4}\right)^{4 \cdot 3} = 1000 \cdot (1.0125)^{12} \approx 1161.62
$$

#### **Fasiləsiz faizlə hesablanma**:
Fasiləsiz faizlə:
$$
M = 1000 \cdot e^{0.05 \cdot 3} = 1000 \cdot e^{0.15} \approx 1000 \cdot 1.161834 \approx 1161.83
$$

Nəticə göstərir ki, fasiləsiz faizlə əldə olunan yekun məbləğ, faiz dövrü çoxaldıqca, mürəkkəb faizə çox yaxınlaşır, lakin həmişə bir qədər artıq olur.

---

### **Fasiləsiz faiz harada istifadə olunur?**
1. **Bank sektoru**:
   - Uzunmüddətli depozit və kredit hesablamalarında daha dəqiq nəticələr əldə etmək üçün.
2. **Maliyyə bazarları**:
   - Mürəkkəb maliyyə alətlərinin qiymətləndirilməsində.
3. **Təbii artım prosesləri**:
   - Əhali artımı, radioaktiv parçalanma, mikroorqanizmlərin böyüməsi və s.

---

### **Nəticə**
Əylər ədədi ($e$), fasiləsiz faiz kimi proseslərdə təbii artımın modelləşdirilməsi üçün idealdır. Bank faizlərinin hesablanmasında, xüsusilə çox kiçik dövrlərlə faiz artımı olduqda, \(e\) baza kimi istifadə edilir. Bu, daha dəqiq və təbii nəticələr verir.

## $M = P e^{rt}$

${e}$ ədədinin necə ortaya çıxdığını və $M = P e^{rt}$ düsturuna necə daxil olduğunu addım-addım izah edək.

---

### 1. Əsas düstur: Limit ilə faiz artımı
Mürəkkəb faiz artımı üçün başlanğıc düstur belədir:

$$
M = P \left(1 + \frac{r}{n}\right)^{nt}
$$

Burada:
- $P$: Başlanğıc məbləğ,
- $r$: İllik faiz dərəcəsi (desimal formada),
- $n$: Faizlərin bir ildə neçə dəfə tətbiq edilməsi,
- $t$: Müddət (illərlə).

---

### 2. Sonsuz mürəkkəbləşməyə keçid
Əgər faizlərin tətbiq edilməsi sayı ($n$) sonsuza yaxınlaşırsa ($n \to \infty$), faiz hesablaması davamlı olur. Bu halda yuxarıdakı ifadənin limiti hesablanır:

\$$
M = P \lim_{n \to \infty} \left(1 + \frac{r}{n}\right)^{nt}
$$

---

### 3. $\text{e}$ ədədinin rolu
Burada $\text{e}$ ədədi bu hissədən yaranır:

$$
e = \lim_{n \to \infty} \left(1 + \frac{1}{n}\right)^n
$$

Yəni $\text{e}$, faiz artımının sonsuz kiçik dövrlərə bölünməsi halında təbii olaraq yaranan bir sabitdir ($e \approx 2.718$).

---

### 4. Düsturun $\text{e}$ ilə əlaqəsi
Faiz artımı düsturunda $\frac{r}{n}$ varsa, bu $\text{e}$-yə əsaslanır. Gəlin bunu əvəz edək:

$$
\lim_{n \to \infty} \left(1 + \frac{r}{n}\right)^{nt} = \left[\lim_{n \to \infty} \left(1 + \frac{1}{n}\right)^n\right]^{rt}
$$

Burada:
- $\lim_{n \to \infty} \left(1 + \frac{1}{n}\right)^n = e$,
- Ona görə də bu ifadə sadələşir:

$$
M = P e^{rt}
$$

---

### 5. Əsas məqam: $\text{e}$-nin olduğu yer
$\text{e}$ məhz faizlərin sonsuz tezliklə tətbiqi zamanı yaranır. Bu limit $\text{e}$-nin hesablanması ilə əlaqədardır və **davamlı artım** prinsiplərini modelləşdirir. 

**Son nəticə** isə belə olur:

$$
M = P e^{rt}
$$

Burada $\text{e}^{rt}$, faizlərin davamlı olaraq eksponential şəkildə artımını göstərir.

## Limit 

### **Limit haqqında geniş məlumat**

Riyaziyyatda **limit** anlayışı bir funksiya və ya ardıcıllığın müəyyən bir nöqtədə və ya sonsuzluqda yaxınlaşdığı dəyəri təsvir edir. Limit anlayışı **diferensial və inteqral hesabın** təməlini təşkil edir və bir çox riyazi, fiziki və mühəndislik tətbiqlərində istifadə olunur.

> **Limit**: Bir funksiya və ya ardıcıllığın dəyişəninin müəyyən bir nöqtəyə və ya sonsuzluğa yaxınlaşdığı zaman yaxınlaşdığı dəyəri təsvir edən anlayışdır.


## **1. Limitin Tərifi**

### a) **Funksiyanın limiti**  
Funksiyanın $\lim_{x \to a} f(x) = L$ olması bu deməkdir:  
- $\forall \epsilon > 0$ (kiçik bir müsbət ədəd),  
- $\exists \delta > 0$ (başqa bir müsbət ədəd) belə ki,  
- Əgər $0 < |x - a| < \delta$, o zaman $|f(x) - L| < \epsilon$.  

Bu o deməkdir ki, $x$ dəyəri $a$-ya yaxınlaşdıqca $f(x)$ funksiyası $L$-ə yaxınlaşır.

---

### b) **Ardıcıllığın limiti**  
Bir ardıcıllığın limiti belə təsvir olunur:  
$$ 
\lim_{n \to \infty} a_n = L \quad \Leftrightarrow \quad \forall \epsilon > 0, \, \exists N \in \mathbb{N} \text{ belə ki, } n > N \Rightarrow |a_n - L| < \epsilon. 
$$  

Burada ardıcıllıq ($a_n$) nə qədər irəli getsə, $L$-ə o qədər yaxınlaşır.

---

## **2. Limitin Təsviri**  
Limit konsepti vasitəsilə dəyişənlərin bir-birinə yaxınlaşması araşdırılır. Bu yaxınlaşma həm sonsuz ($x \to \infty$) zamanlarda, həm də kiçik dəyişikliklər zamanı ($x \to a$) tədqiq edilir.

Məsələn:  
- Funksiyanın davranışını müəyyən etmək üçün ($\lim_{x \to 0} \frac{\sin x}{x} = 1$),  
- Sonsuz kiçik və ya böyük dəyərlərə yaxınlaşmada ($\lim_{x \to \infty} \frac{1}{x} = 0$) istifadə edilir.

---

## **3. Limitin Növləri**

### a) **Sonsuzluqda limit ($x \to \infty$):**  
Əgər $x$ sonsuzluğa yaxınlaşdıqca $f(x)$ müəyyən bir dəyərə yaxınlaşırsa:  
$$ 
\lim_{x \to \infty} f(x) = L 
$$  
Məsələn:  
$$ 
\lim_{x \to \infty} \frac{1}{x} = 0 
$$  

---

### b) **Sonsuzluğa yaxınlaşma ($\lim_{x \to \infty} f(x) = \infty$):**  
Əgər $f(x)$ qeyri-məhdud şəkildə artırsa:  
$$ 
\lim_{x \to \infty} f(x) = \infty 
$$  
Məsələn:  
$$ 
\lim_{x \to \infty} x^2 = \infty 
$$  

---

### c) **Sol və sağ limitlər:**  
Əgər limit yalnız bir tərəfdən araşdırılırsa:  
- **Sol limit:** $\lim_{x \to a^-} f(x)$,  
- **Sağ limit:** $\lim_{x \to a^+} f(x)$.  

Əgər:  
$$ 
\lim_{x \to a^-} f(x) = \lim_{x \to a^+} f(x) = L, 
$$  
o zaman limit mövcuddur və $\lim_{x \to a} f(x) = L$.

---

## **4. Əsas Limit Xassələri**

1. **Cəmin limiti:**  
   $$ 
   \lim_{x \to a} [f(x) + g(x)] = \lim_{x \to a} f(x) + \lim_{x \to a} g(x) 
   $$  

2. **Hasilin limiti:**  
   $$ 
   \lim_{x \to a} [f(x) \cdot g(x)] = \lim_{x \to a} f(x) \cdot \lim_{x \to a} g(x) 
   $$  

3. **Nisbin limiti:**  
   $$ 
   \lim_{x \to a} \frac{f(x)}{g(x)} = \frac{\lim_{x \to a} f(x)}{\lim_{x \to a} g(x)}, \quad \text{əgər } \lim_{x \to a} g(x) \neq 0. 
   $$  

---

## **5. Əsas Limitlər**

### a) **Triviyal nümunələr:**  
1. $\lim_{x \to 0} c = c$, harada $c$ sabitdir.  
2. $\lim_{x \to \infty} \frac{1}{x} = 0$.  

---

### b) **Əsas trigonometrik limit:**  
$$

## Limitin hesablanma qaydaları və hər qaydaya uyğun bir real nümunə misalı



### 1. **Cəmin limiti**

**Qayda:**  
Əgər $ \lim_{x \to a} f(x) $ və $ \lim_{x \to a} g(x) $ mövcuddursa, o zaman:  
$$
\lim_{x \to a} [f(x) + g(x)] = \lim_{x \to a} f(x) + \lim_{x \to a} g(x)
$$

**Misal:**  
Funkisiyalar: $ f(x) = 2x $ və $ g(x) = x^2 $, $ a = 1 $.
$$
\lim_{x \to 1} \left( 2x + x^2 \right) = \lim_{x \to 1} 2x + \lim_{x \to 1} x^2 = 2(1) + 1^2 = 2 + 1 = 3
$$

---

### 2. **Hasilin limiti**

**Qayda:**  
Əgər $ \lim_{x \to a} f(x) $ və $ \lim_{x \to a} g(x) $ mövcuddursa, o zaman:  
$$
\lim_{x \to a} [f(x) \cdot g(x)] = \lim_{x \to a} f(x) \cdot \lim_{x \to a} g(x)
$$

**Misal:**  
Funkisiyalar: $ f(x) = x $ və $ g(x) = x + 1 $, $ a = 3 $.  
$$
\lim_{x \to 3} (x \cdot (x + 1)) = \lim_{x \to 3} x \cdot \lim_{x \to 3} (x + 1) = 3 \cdot (3 + 1) = 3 \cdot 4 = 12
$$

---

### 3. **Nisbinin limiti**

**Qayda:**  
Əgər $ \lim_{x \to a} f(x) $ və $ \lim_{x \to a} g(x) $ mövcuddursa və $ \lim_{x \to a} g(x) \neq 0 $, o zaman:  
$$
\lim_{x \to a} \frac{f(x)}{g(x)} = \frac{\lim_{x \to a} f(x)}{\lim_{x \to a} g(x)}
$$

**Misal:**  
Funkisiyalar: $ f(x) = x^2 + 2x $ və $ g(x) = x - 1 $, $ a = 2 $.  
$$
\lim_{x \to 2} \frac{x^2 + 2x}{x - 1} = \frac{\lim_{x \to 2} (x^2 + 2x)}{\lim_{x \to 2} (x - 1)} = \frac{2^2 + 2(2)}{2 - 1} = \frac{4 + 4}{1} = 8
$$

---

### 4. **L'Hopital Qaydası**

**Qayda:**  
Əgər limit $ 0/0 $ və ya $ \infty/\infty $ formasında olarsa, L'Hopital qaydasını tətbiq edirik:  
$$
\lim_{x \to a} \frac{f(x)}{g(x)} = \lim_{x \to a} \frac{f'(x)}{g'(x)}
$$

**Misal:**  
Funkisiyalar: $ f(x) = \sin x $ və $ g(x) = x $, $ a = 0 $.  

Bu funksiya $ \frac{0}{0} $ formasında olduğu üçün:  
$$
\lim_{x \to 0} \frac{\sin x}{x} = \lim_{x \to 0} \frac{\cos x}{1} = 1
$$

## Derivative

### Törəmənin Tərifi

> **Törəmə**: Bir funksiyanın dəyişmə sürətini və ya qrafikinin müəyyən bir nöqtədəki toxunma xəttinin meylini ifadə edir.  
- **Funksiyanın Törəməsi**: Funksiyanın törəməsi, onun hər bir nöqtədə dəyişmə sürətini hesablayan yeni bir funksiyadır.Bu, bir funksiyanın xəttinin meylini və ya əyrinin bucağını tapmaq üçün istifadə olunur.

Riyaziyyatda, törəmə bir funksiyanın müəyyən bir nöqtədəki dəyişmə sürətini ifadə edir.  Törəmə həmçinin fizikada və mühəndislikdə müxtəlif tətbiqlərə malikdir, məsələn, sürət, sürətlənmə və s. kimi anlayışları hesablamaq üçün.

---

### **Törəmənin Rəsmi Tərifi**

Bir $f(x)$ funksiyasının $a$ nöqtəsində törəməsi aşağıdakı sərhəd olaraq müəyyən edilir:

$$
f'(a) = \lim_{h \to 0} \frac{f(a+h) - f(a)}{h}
$$

Burada:
- $f(a)$: Funksiyanın $a$ nöqtəsindəki qiyməti,
- $f(a+h)$: Funksiyanın $a$-dan $h$ qədər uzaqlıqda olan nöqtədəki qiyməti,
- $h$: İnkişafın kiçik addımıdır.

Bu ifadə bir nöqtədəki ani dəyişmə sürətini hesablayır.

---

### **Törəmənin Əsas İdeyası**

Törəmə, bir əyrinin toxunma xəttinin meylini tapmağa imkan verir. Əgər əyri düz bir xəttdirsə, bu xətlərin meyli sabit olur və törəmə bu sabit meyli verir. Əgər əyri dəyişkəndirsə, o zaman hər nöqtədəki meyl fərqli olur və törəmə bunu göstərir.

---

### **Törəmənin Əsas Qaydaları**

1. **Sabit funksiyanın törəməsi**:  
   Əgər $f(x) = c$ (burada $c$ sabitdir), onda:  
   $$
   f'(x) = 0
   $$

2. **Xətti funksiyanın törəməsi**:  
   Əgər $f(x) = ax + b$, onda:  
   $$
   f'(x) = a
   $$

3. **Çoxhədli funksiyanın törəməsi**:  
   Əgər $f(x) = x^n$, onda:  
   $$
   f'(x) = n \cdot x^{n-1}
   $$

4. **Cəmin törəməsi**:  
   $$
   [f(x) + g(x)]' = f'(x) + g'(x)
   $$

5. **Vurmanın törəməsi**:  
   $$
   [f(x)g(x)]' = f'(x)g(x) + f(x)g'(x)
   $$

6. **Bölmənin törəməsi**:  
   $$
   \left[\frac{f(x)}{g(x)}\right]' = \frac{f'(x)g(x) - f(x)g'(x)}{g(x)^2}
   $$

---

### **Törəmənin Geometrik İzahı**

Törəmə, bir əyrinin bir nöqtəsində çəkilən toxunma xəttinin meylini (bu xətdən $x$-oxuna olan bucağı) tapmaqdır.

- Əgər $f'(x) > 0$, funksiya həmin nöqtədə artandır.
- Əgər $f'(x) < 0$, funksiya azalandır.
- Əgər $f'(x) = 0$, bu nöqtə ya maksimum, ya minimum, ya da sabit nöqtə ola bilər.

---

### **Törəmənin Tətbiq Sahələri**  
1. **Fizika**: Hərəkət (sürət və sürətlənmə) hesablamaları.  
2. **Optimallaşdırma**: Maksimum və minimum dəyərləri tapmaq.  
3. **İqtisadiyyat**: Mənfəətin və ya istehsal funksiyasının analizi.  
4. **Maşın Öyrənməsi**: Gradient Descent və s.

Törəmə bir funksiyanın dəyişmə sürətini və ya meylini təyin etmək üçün istifadə olunan əsas riyazi anlayışdır. Törəmənin hesablanması qaydaları və üsulları müəyyən edilmiş qaydalara əsaslanır. Gəlin əsas qaydaları ətraflı izah edək:

---

## **1. Əsas Törəmə Qaydaları:**

### **a) Sabitin Törəməsi**
Sabit bir ədədin törəməsi həmişə 0-dır.
$$
\frac{d}{dx}[c] = 0
$$
Məsələn:
$$
\frac{d}{dx}[5] = 0
$$

---

### **b) X-in Törəməsi**
$$
\frac{d}{dx}[x] = 1
$$
Məsələn:
$$
\frac{d}{dx}[3x] = 3
$$

---

### **c) Güc Qaydası (Power Rule)**
Əgər $f(x) = x^n$, onda törəmə:
$$
\frac{d}{dx}[x^n] = n \cdot x^{n-1}
$$
Məsələn:
$$
\frac{d}{dx}[x^3] = 3x^2
$$

---

### **d) Cəmin və Fərqin Törəməsi**
Törəmə cəmləri və fərqləri ayrı-ayrılıqda hesablana bilər:
$$
\frac{d}{dx}[f(x) + g(x)] = f'(x) + g'(x)
$$
$$
\frac{d}{dx}[f(x) - g(x)] = f'(x) - g'(x)
$$

---

### **e) Sabit Çarpanın Qaydası**
Sabit bir ədədi funksiya ilə çarpdıqda, törəmə sabiti qoruyur:
$$
\frac{d}{dx}[c \cdot f(x)] = c \cdot f'(x)
$$
Məsələn:
$$
\frac{d}{dx}[5x^2] = 5 \cdot \frac{d}{dx}[x^2] = 5 \cdot 2x = 10x
$$

---

## **2. Məxsusi Qaydalar:**

### **a) Çarpmanın Qaydası (Product Rule)**
İki funksiyanın hasili üçün:
$$
\frac{d}{dx}[f(x) \cdot g(x)] = f'(x) \cdot g(x) + f(x) \cdot g'(x)
$$
Məsələn:
$$
\frac{d}{dx}[x^2 \cdot \sin(x)] = 2x \cdot \sin(x) + x^2 \cdot \cos(x)
$$

---

### **b) Bölmə Qaydası (Quotient Rule)**
İki funksiyanın bölünməsi üçün:
$$
\frac{d}{dx}\left[\frac{f(x)}{g(x)}\right] = \frac{f'(x) \cdot g(x) - f(x) \cdot g'(x)}{[g(x)]^2}
$$
Məsələn:
$$
\frac{d}{dx}\left[\frac{x}{x^2+1}\right] = \frac{1 \cdot (x^2+1) - x \cdot 2x}{(x^2+1)^2} = \frac{x^2+1 - 2x^2}{(x^2+1)^2} = \frac{1-x^2}{(x^2+1)^2}
$$

---

### **c) Zəncir Qaydası (Chain Rule)**
Kompozit funksiyalar üçün törəmə:
$$
\frac{d}{dx}[f(g(x))] = f'(g(x)) \cdot g'(x)
$$
Məsələn:
$$
\frac{d}{dx}[\sin(x^2)] = \cos(x^2) \cdot 2x
$$

## Partial Differentiation (Qismən Törəmə)

### **Qismən Törəmə (Partial Differentiation)**

Qismən törəmə (partial differentiation), çox dəyişənli funksiyalar üçün tətbiq olunan bir törəmə üsuludur. Bir funksiya bir neçə dəyişənə bağlı olduqda, qismən törəmə yalnız bir dəyişənin dəyişməsi ilə digər dəyişənlərin sabit tutulduğu halda funksiyanın dəyişmə sürətini hesablamağa imkan verir.

Qismən törəmənin məqsədi, çox dəyişənli funksiyanı hər bir dəyişənə görə ayrı-ayrılıqda törəməkdir.

---

### **Tərif:**

> **Partial Differentiation (Qismən Törəmə)**: Bir funksiyanın bir neçə dəyişəni olduqda yalnız bir dəyişənə görə törəməsini hesablama metodudur. Digər dəyişənlər sabit kimi qəbul edilir.

Əgər $ f(x_1, x_2, \dots, x_n) $ funksiyası $ n $ dəyişənə bağlıdırsa və biz yalnız bir dəyişənin təsirini nəzərdən keçirmək istəyiriksə, o zaman o dəyişənə görə qismən törəmə (partial derivative) hesablamağı tətbiq edirik. Məsələn, \( x \)-ə görə qismən törəmə belə göstərilir:

$$
\frac{\partial f}{\partial x} = \lim_{\Delta x \to 0} \frac{f(x + \Delta x, x_2, \dots, x_n) - f(x_1, x_2, \dots, x_n)}{\Delta x}
$$

Bu formulda $ x $ dəyişəninin təsiri, digər dəyişənlər ($x_2, x_3, \dots, x_n\$) sabit tutulur.

---

### **Qismən Törəmə Hesablama Qaydaları:**

1. **Əsas Funksiya Qaydası:**
   - Törəmə alarkən, digər dəyişənlər sabit qalır.
   - Məsələn, $ f(x, y) = x^2 + y^3 $ funksiyası üçün:
     $$
     \frac{\partial}{\partial x} (x^2 + y^3) = 2x, \quad \frac{\partial}{\partial y} (x^2 + y^3) = 3y^2
     $$

2. **Güc Qaydası (Power Rule):**
   - Funksiyanın gücü ilə bağlı törəmə qaydası, bir dəyişən üçün olduğu kimi çox dəyişənli funksiyalarda da tətbiq olunur.
   $$
   \frac{\partial}{\partial x} (x^n) = n \cdot x^{n-1}
   $$
   Məsələn:
   $$
   \frac{\partial}{\partial x} (x^3) = 3x^2
   $$

3. **Cəmin və Fərqin Törəməsi:**
   - Çox dəyişənli funksiyanın cəmi və fərqi üçün törəmə qaydaları belədir:
   $$
   \frac{\partial}{\partial x} [f(x) + g(x)] = \frac{\partial f}{\partial x} + \frac{\partial g}{\partial x}
   $$
   $$
   \frac{\partial}{\partial x} [f(x) - g(x)] = \frac{\partial f}{\partial x} - \frac{\partial g}{\partial x}
   $$

4. **Çarpmanın Qaydası (Product Rule):**
   - Çox dəyişənli funksiyanın çarpılmasında:
   $$
   \frac{\partial}{\partial x} [f(x) \cdot g(x)] = \frac{\partial f}{\partial x} \cdot g(x) + f(x) \cdot \frac{\partial g}{\partial x}
   $$

5. **Bölmənin Qaydası (Quotient Rule):**
   - Funksiyalar bölündükdə:
   $$
   \frac{\partial}{\partial x} \left[ \frac{f(x)}{g(x)} \right] = \frac{g(x) \cdot \frac{\partial f}{\partial x} - f(x) \cdot \frac{\partial g}{\partial x}}{[g(x)]^2}
   $$

6. **Zəncir Qaydası (Chain Rule):**
   - Qismən törəmə üçün, funksiyanın kompozisiyası olduqda bu qayda tətbiq edilir:
   $$
   \frac{\partial}{\partial x} [f(g(x))] = f'(g(x)) \cdot g'(x)
   $$

---

### **Qismən Törəmənin İstifadə Sahələri:**

1. **Optimizasiya Problemləri:**
   Qismən törəmə, çox dəyişənli funksiyaların maksimum və ya minimum nöqtələrini tapmaqda geniş istifadə olunur. Məsələn, şirkətlərin mənfəət funksiyalarını optimallaşdırarkən bir neçə dəyişənin təsirini hesablamaq üçün istifadə edilir.

2. **Fizika və Mühəndislik:**
   Fərqli fiziki sistemlərdə və mühəndislik problemlərində, məsələn, temperaturun dəyişməsi, sıxlıq, təzyiq və s. kimi faktorlar arasında əlaqəni tapmaq üçün qismən törəmələr istifadə olunur.

3. **Təsadüfi Proseslər və İqtisadiyyat:**
   Qismən törəmələr iqtisadi modellərin qurulmasında, marketinq strategiyalarının və risk analizi məsələlərinin həllində də tətbiq olunur.

---

### **Misallar:**

1. **Misal 1:**
   $$
   f(x, y) = x^3 y^2 + 5x^2 y
   $$
   - $x$-ə görə qismən törəmə:
   $$
   \frac{\partial}{\partial x}(x^3 y^2 + 5x^2 y) = 3x^2 y^2 + 10x y
   $$
   - $y$-yə görə qismən törəmə:
   $$
   \frac{\partial}{\partial y}(x^3 y^2 + 5x^2 y) = 2x^3 y + 5x^2
   $$

2. **Misal 2:**
   $$
   f(x, y) = \sin(xy)
   $$
   - $x$-ə görə qismən törəmə:
   $$
   \frac{\partial}{\partial x}(\sin(xy)) = y \cos(xy)
   $$
   - $y$-yə görə qismən törəmə:
   $$
   \frac{\partial}{\partial y}(\sin(xy)) = x \cos(xy)
   $$

---

### **Nəticə:**

Qismən törəmə çox dəyişənli funksiyaların analizində vacib bir alətdir və tətbiq sahələri genişdir. Bu üsul, bir neçə dəyişənin təsirini müstəqil şəkildə təhlil etməyə imkan verir və funksiyaların daha mürəkkəb analizini asanlaşdırır.

## Chain Rule (Zəncir Qaydası)

**Zəncir Qaydası (Chain Rule)**: Mürəkkəb funksiyaların törəməsini tapmaq üçün istifadə olunur. Əgər funksiya bu formada verilirsə:  

$$
y = f(g(x)),
$$

onun törəməsi belə hesablanır:  

$$
\frac{dy}{dx} = f'(g(x)) \cdot g'(x).
$$

**Sadə izah:**
1. Xarici funksiyanı ($f$) daxili funksiya ($g(x)$) üzrə törəmələyirsən.
2. Daha sonra daxili funksiyanı (\(g(x)\)) \(x\) üzrə törəmələyirsən.
3. Hər ikisini vurursan.

**Misal:**  
$$
y = (\sin(x))^2
$$  
1. Xarici funksiya: $(u)^2$ → Törəməsi: $2u$, burada $u = \sin(x)$.  
2. Daxili funksiya: $\sin(x)$ → Törəməsi: $\cos(x)$.  

Nəticə:  
$$
\frac{dy}{dx} = 2\sin(x) \cdot \cos(x).
$$

---
Əgər $ y = \sin(x^2) $ funksiyasının törəməsini hesablamaq istəyirsinizsə, **zəncir qaydasını** tətbiq edirik.



### Funksiya:
$$
y = \sin(x^2)
$$



### Addımlar:

1. **Xarici funksiya**:  
   $\sin(u)$, burada $ u = x^2 $.  
   Törəməsi:  
   $$
   \frac{d}{du} \sin(u) = \cos(u)
   $$

2. **Daxili funksiya**:  
   $ u = x^2 $.  
   Törəməsi:  
   $$
   \frac{d}{dx} x^2 = 2x
   $$

---

### Zəncir Qaydası:
Zəncir qaydasına əsasən:  
$$
\frac{dy}{dx} = \cos(u) \cdot \frac{du}{dx}.
$$

Burada $ u = x^2 $, ona görə:

$$
\frac{dy}{dx} = \cos(x^2) \cdot 2x.
$$

---

### Nəticə:
$$
\frac{d}{dx} \sin(x^2) = 2x \cos(x^2).
$$

---
$y=\sin(2x)$-in törəməsini hesablamaq üçün **zəncir qaydasından** istifadə edirik.

$$
\frac{d}{dx} \sin(2x) = \cos(2x) \cdot \frac{d}{dx}(2x)
$$

$\frac{d}{dx}(2x) = 2$ olduğuna görə:

$$
\frac{d}{dx} \sin(2x) = 2 \cos(2x)
$$

Beləliklə, $sin(2x)$-in törəməsi:

$$
2 \cos(2x)
$$

## Piecewise Differentiation

**Piecewise differentiation** (hissə-hissə diferensiallama) bir funksiyanın fərqli aralıqlarda müxtəlif ifadələrlə təyin edildiyi hallarda törəməni tapmaq metodudur. Bu yanaşma **piecewise-defined functions** üçün tətbiq olunur, yəni funksiya fərqli aralıqlarda müxtəlif qanunauyğunluqlarla göstərilir.

### **Piecewise-defined functions nədir?**
Bu, aşağıdakı kimi yazılan funksiyalardır:

$$
f(x) =
\begin{cases} 
x^2 & \text{əgər } x < 0 \\ 
2x + 1 & \text{əgər } x \geq 0
\end{cases}
$$

Bu funksiya göstərir ki:
- $x < 0$ üçün funksiya $x^2$-dir.
- $x \geq 0$ üçün isə funksiya $2x + 1$-dir.

### **Piecewise differentiation necə işləyir?**
Hissə-hissə diferensiallama üçün aşağıdakı addımlar tətbiq olunur:

#### 1. **Hər bir hissədə fərdi törəmə hesabla**
Hər bir aralıq üçün funksiyanın ifadəsini ayrı-ayrılıqda diferensialla:

- $x < 0$ üçün $f(x) = x^2$-dir. Bu halda törəmə:
  $$
  f'(x) = 2x
  $$

- $x \geq 0$ üçün $f(x) = 2x + 1$-dir. Bu halda törəmə:
  $$
  f'(x) = 2
  $$

#### 2. **Törəməni yenidən piecewise olaraq yaz**
Törəmə də **piecewise** formada yazılır:

$$
f'(x) =
\begin{cases} 
2x & \text{əgər } x < 0 \\ 
2 & \text{əgər } x \geq 0
\end{cases}
$$

#### 3. **Keçid nöqtəsini (junction) yoxla**
Hər bir hissənin arasında keçid nöqtəsində törəmənin uyğun olub-olmamasını yoxlamaq vacibdir. Keçid nöqtəsi burada $x = 0$-dır.

- **Sol tərəfdən törəmə:** 
  $$
  \lim_{{x \to 0^-}} f'(x) = 2(0) = 0
  $$

- **Sağ tərəfdən törəmə:** 
  $$
  \lim_{{x \to 0^+}} f'(x) = 2
  $$

Keçid nöqtəsində törəmə **uyğun deyil** (0 ilə 2 fərqlidir). Bu halda funksiya törəməlidir, amma törəmə keçid nöqtəsində kəsilməyə məruz qalır.

### **Praktik nümunə**
Məsələn, parça-parça verilən funksiyanı differensiallayaq:

$$
f(x) =
\begin{cases} 
x^3 & \text{əgər } x \leq 1 \\ 
\sin(x) & \text{əgər } x > 1
\end{cases}
$$

**Addım 1: Hər hissədə törəməni hesablayırıq**
- $x \leq 1$ üçün $f(x) = x^3$, törəmə:
  $$
  f'(x) = 3x^2
  $$

- $x > 1$ üçün $f(x) = \sin(x)$, törəmə:
  $$
  f'(x) = \cos(x)
  $$

**Addım 2: Törəmə ifadəsini yazırıq**
$$
f'(x) =
\begin{cases} 
3x^2 & \text{əgər } x \leq 1 \\ 
\cos(x) & \text{əgər } x > 1
\end{cases}
$$

### **Xülasə**
**Piecewise differentiation**:
1. Funksiyanı hər aralıqda ayrıca diferensiallayır.
2. Keçid nöqtələrində törəmənin uyğunluğunu yoxlayır (əgər varsa).
3. Törəməni fərqli aralıqlarda **piecewise** olaraq təyin edir.

Bu metod, məsələn, hissə-hissə təyin olunan funksiyaların modelləşdirilməsi və analizi zamanı geniş istifadə olunur. Əlavə sualınız varsa, məmnuniyyətlə kömək edərəm! 😊

**Piecewise Differentiation** (hissə-hissə diferensiallama), bir funksiyanın müxtəlif aralıqlarda fərqli ifadələrlə təyin olunduğu hallarda törəməsini hesablamaq üçün tətbiq olunan metoddur. Bu, əsasən **piecewise-defined functions** (hissə-hissə təyin olunmuş funksiyalar) üçün istifadə edilir.

#### **Tərif:**
Piecewise Differentiation, bir funksiyanın hər aralıqda ayrıca törəməsinin tapılmasını, sonra bu törəmələrin həmin aralıqlar üçün şərtlərə uyğun birləşdirilməsini nəzərdə tutur. 

Funksiyanın ifadəsi belədir:

$$
f(x) =
\begin{cases} 
f_1(x) & \text{əgər } a_1 \leq x < a_2 \\ 
f_2(x) & \text{əgər } a_2 \leq x < a_3 \\ 
\vdots \\
f_n(x) & \text{əgər } a_{n-1} \leq x \leq a_n
\end{cases}
$$

Hər bir hissədə ($f_1(x), f_2(x), \dots, f_n(x)$) törəmə ayrıca hesablanır.

#### **Əsas xüsusiyyətləri:**
1. **Hissələr üzrə törəmə:** Hər bir ifadə (hissə) öz intervalları daxilində diferensiallanır.
2. **Keçid nöqtələrinin yoxlanışı:** Törəmənin keçid nöqtələrində uyğun olub-olmadığını yoxlamaq lazımdır (sol tərəfli və sağ tərəfli törəmələr eyni olmalıdır).
3. **Törəmənin birləşdirilməsi:** Hər bir aralıqdakı törəmə ifadələri parça-parça birləşdirilərək tam törəmə təyin olunur.

#### **Tətbiq sahələri:**
- **Hissə-hissə təyin olunmuş funksiyalar:** Məsələn, addım funksiyaları, modifikasiya edilmiş sinus funksiyaları, və ya məhdudlaşdırılmış çoxhədli funksiyalar.
- **Mühəndislik və fizika:** Şok dalğaları, elektrik dövrələri və ya gərginliklərin dəyişməsi ilə bağlı modellər.
- **Maşın öyrənmə:** Aktivasiya funksiyalarında hissə-hissə analiz.

Qısacası, **piecewise differentiation**, mürəkkəb funksiyaları daha kiçik, idarə oluna bilən hissələrə ayıraraq onları analiz etməyə imkan verir. 😊$$

## $\mathbf{\frac{d}{dx}}$ Törəmə Notası nə deməkdir?

**$\frac{d}{dx}$** notası törəməni simvolik şəkildə yazmaq üçün istifadə olunan standart riyazi işarədir. Bu, törəmənin mənasını daha aydın ifadə edir və "funksiyanın $x$-ə görə dəyişmə sürəti"ni göstərir.

---

### **Törəmə Notasının İzahı**

1. **$\mathbf{\frac{d}{dx}}$ nə deməkdir?**
   - **$d$**: "Differensial" deməkdir. Funksiyanın sonsuz kiçik dəyişməsini ifadə edir.
   - **$dx$**: $x$-dəki sonsuz kiçik dəyişməni göstərir.

   Yəni $\frac{d}{dx}$, funksiyanın $x$-ə görə dəyişmə sürətini hesablayan operator kimi başa düşülür.

---

2. **Tam yazılışı:**
   $$
   \frac{d}{dx} f(x)
   $$
   Bu ifadə, $f(x)$-in $x$-ə görə törəməsini tapmaq deməkdir.

---

3. **Sadələşdirilmiş formada:**
   Əgər $y = f(x)$ verilirsə, onun törəməsini belə də yaza bilərlər:
   $$
   \frac{dy}{dx}.
   $$
   Burada:
   - **$dy$**: Funksiyanın dəyişməsi.
   - **$dx$**: Müvafiq olaraq $x$-dəki dəyişmə.

---

### Niyə belə yazılır?
Bu işarə **Liebniz notası** adlanır və törəmə konsepsiyasını çox dəqiq göstərir:
1. Törəmənin **məqsədi**: Bir funksiyanın giriş dəyəri ($x$) dəyişdikcə çıxış dəyərinin ($y$) necə dəyişdiyini ölçməkdir.
2. **$\frac{dy}{dx}$**: Dəyişmələr arasındakı nisbəti göstərir.

---

### Digər notalar:
Fərqli yazılışlar da mövcuddur:
1. Newton notası: $\dot{y} $ (zamanla dəyişmə üçün istifadə olunur, fizika kontekstində).
2. Lagrange notası: $ f'(x) $ (x-ə görə törəməni göstərir).
3. Liebniz notası: $ \frac{d}{dx} f(x) $ (diferensial dəyişmələrə fokuslanır). 

Hər biri eyni məna versə də, **$\frac{d}{dx}$** diferensial hesabın əsas anlayışını daha geniş izah edir.

## $\text{f(x,y)} = 4x^4 + 5y^5$ Funksiya və Onun Törəməsi

Verilən sualı analiz edək:

### Funksiya və Onun Törəməsi
$$
\text{f(x,y)} = 4 \cdot x^4 + 5 \cdot y^5
$$

Burada, $ \text{diff(f(x,y), x, y)} $ ifadəsi ikinci dərəcəli törəməni hesablayır:
- Əvvəlcə $ \text{f(x,y)} $ ifadəsinin $x$-ə görə törəməsi alınır.
- Sonra alınan nəticənin $y$-yə görə törəməsi alınır.

---

### Funksiyanın Strukturu
Verilən funksiya:
$$
\text{f(x,y)} = 4x^4 + 5y^5
$$
Bu, $x$ və $y$ arasında **heç bir qarşılıqlı əlaqə (interaction)** olmayan sadə bir funksiyadır. Yəni:
- $x$ ilə bağlı olan hissə $4x^4$-dür.
- $y$ ilə bağlı olan hissə $5y^5$-dir.

Buna görə:
1. $4x^4$ ifadəsinin $y$-yə görə törəməsi **0** olacaq, çünki $y$-yə bağlı heç nə yoxdur.
2. Oxşar şəkildə, $5y^5$ ifadəsinin $x$-ə görə törəməsi də **0** olacaq.

---

### İkinci Törəmə (Hesablama)
Birinci addım:
$$
\frac{\partial \text{f(x,y)}}{\partial x} = \frac{\partial (4x^4 + 5y^5)}{\partial x} = 16x^3 + 0 = 16x^3
$$

İkinci addım:
$$
\frac{\partial}{\partial y} (16x^3) = 0, \quad \text{çünki burada } y \text{ yoxdur.}
$$

Nəticə:
$$
\text{diff( f(x,y), x, y )} = 0
$$

---

### Ümumi Qaydalar
Əgər bir funksiya $x$ və $y$ arasında qarşılıqlı təsirə malik deyilsə (yəni $x \cdot y$, $x + y$, və s. kimi ifadələr yoxdur), qarışıq törəmə həmişə **0** olur. Bu sizin funksiyanızda da baş verir.

## Törəmə (Artma, Azalma aralığı, Böhran nöqtələri , Ekstremium nöqtələr, Parçada ən böyük qiymət və ən kiçik qiymət)

### Törəmə və Onun Əsas Tətbiqləri

Törəmə, funksiyanın dəyişmə sürətini və dəyişmə istiqamətini müəyyən edən mühüm bir riyazi alətdir. Bu anlayışın tətbiqləri çox genişdir, xüsusən **artma** və **azalma** aralıqları, **böhran nöqtələri**, **ekstremum nöqtələri** və **parçada ən böyük qiymət** məsələlərində.

Aşağıda bu anlayışların hər birini misallarla izah edəcəyəm:

---

### 1. **Törəmə və Artma, Azalma Aralıqları**

Bir funksiyanın artma və ya azalma aralıqlarını tapmaq üçün onun törəməsindən istifadə edirik.

- **Artma Aralığı**: Əgər funksiyanın törəməsi müsbət bir dəyər alırsa ($ f'(x) > 0 $), bu halda funksiyanın qiymətləri artır. Bu, funksiyanın artma olduğu aralıqdır.
  
- **Azalma Aralığı**: Əgər funksiyanın törəməsi mənfi bir dəyər alırsa ($ f'(x) < 0 $), bu halda funksiyanın qiymətləri azalır. Bu, funksiyanın azalma olduğu aralıqdır.

#### Misal:  
Verilən funksiya:
$$
f(x) = x^3 - 3x^2 + 2
$$

Törəməsini hesablayaq:
$$
f'(x) = 3x^2 - 6x
$$

Törəməni sıfıra bərabər edirik və tənliyin köklərini tapırıq:
$$
3x^2 - 6x = 0 \quad \Rightarrow \quad 3x(x - 2) = 0
$$
Buradan $ x = 0 $ və $ x = 2 $ olan kökləri tapırıq.

Bu kökləri istifadə edərək funksiyanın artma və azalma aralıqlarını müəyyən edirik:
- $ f'(x) > 0 $ olduğu aralıqlar (artma): $ (-\infty, 0) \cup (2, \infty) $
- $ f'(x) < 0 $ olduğu aralıqlar (azalma): $ (0, 2) $

---

### 2. **Böhran Nöqtələri**

Böhran nöqtəsi, funksiyanın törəməsinin sıfır olduğu və ya mövcud olmadığı nöqtədir. Bu nöqtələr həm də ekstremum nöqtələri (maksimum və minimum) ola bilər.

Böhran nöqtələrini tapmaq üçün funksiyanın törəməsini sıfıra bərabər edirik.

#### Misal:  
Yuxarıdakı funksiyanı ( $ f(x) = x^3 - 3x^2 + 2 $) təhlil etdik. Burada $ x = 0 $ və $ x = 2 $ olan nöqtələr **böhran nöqtələridir**.

---

### 3. **Ekstremum Nöqtələri**

**Ekstremum nöqtələri**, funksiyanın maksimum və ya minimum qiymətlərini aldığı nöqtələrdir. Ekstremum nöqtələri iki növə bölünür:
- **Maksimum**: Funksiyanın həmin nöqtədəki qiyməti, ətrafdakı nöqtələrdən böyükdür.
- **Minimum**: Funksiyanın həmin nöqtədəki qiyməti, ətrafdakı nöqtələrdən kiçikdir.

Ekstremum nöqtələrinin mövcudluğunu təyin etmək üçün bir neçə üsul mövcuddur:
- **Birinci törəmə testi**: Əgər $ f'(x) = 0 $ və $ f'(x) $ işarəsi dəyişirsə, bu nöqtə ekstremum nöqtəsidir.
- **İkinci törəmə testi**: Əgər $ f''(x) > 0 $ isə nöqtə minimum, $ f''(x) < 0 $ isə nöqtə maksimumdur.

#### Misal:
Törəməsindən əldə etdiyimiz $ f'(x) = 3x^2 - 6x $, köklərini tapdıq: $ x = 0 $ və $ x = 2 $.

**İkinci törəmə**:
$$
f''(x) = 6x - 6
$$
- $f''(0) = -6 $, bu zaman $ x = 0 $ nöqtəsində **maksimum** var.
- $ f''(2) = 6 $, bu zaman $ x = 2 $ nöqtəsində **minimum** var.

Beləliklə, funksiyanın $ x = 0 $ nöqtəsində maksimum və $ x = 2 $ nöqtəsində minimum qiymətləri var.

---

### 4. **Parçada Ən Böyük Qiymət və Ən Kiçik Qiymət**

Bir funksiyanın parçada ən böyük qiymətini tapmaq üçün, funksiyanın qiymətləri həm də **böhran nöqtələrində** və ya **sərhəd nöqtələrində** (funksiyanın təyin olduğu intervalın sərhədləri) nəzərdən keçirilməlidir.

- **Parçada ən böyük qiymət**: Funksiyanın təyin olduğu intervalda ən yüksək qiymət aldığı nöqtədir.
- **Parçada ən kiçik qiymət**: Funksiyanın təyin olduğu intervalda ən aşağı qiymət aldığı nöqtədir.

#### Misal:
Verilən funksiya:
$$
f(x) = -x^2 + 4x
$$
Funksiya $ [0, 4] $ intervalında təyin olunub.

Funksiyanın törəməsi:
$$
f'(x) = -2x + 4
$$

Törəməni sıfıra bərabər edirik:
$$
-2x + 4 = 0 \quad \Rightarrow \quad x = 2
$$

**Sərhəd nöqtələrini** yoxlayırıq:
- $ f(0) = -(0)^2 + 4(0) = 0 $
- $ f(4) = -(4)^2 + 4(4) = 0 $
- $ f(2) = -(2)^2 + 4(2) = 4 $

Bu zaman, parçada ən böyük qiymət $x = 2 $nöqtəsindədir və ən kiçik qiymət $ x = 0 $ və $ x = 4 $ nöqtələrində eynidir.

---

### Yekun:

- **Törəmə** funksiyanın dəyişmə sürətini və istiqamətini göstərir.
- **Artma və Azalma aralıqları** funksiyanın nə vaxt artırıldığını və ya azaldığını təyin edir.
- **Böhran nöqtələri** funksiyanın törəməsinin sıfır olduğu nöqtələrdir.
- **Ekstremum nöqtələri** funksiyanın maksimum və ya minimum qiymətini aldığı nöqtələrdir.
- **Parçada ən böyük qiymət** funksiyanın təyin olduğu intervalda ən yüksək və ya ən aşağı qiymətləri tapmağa kömək edir. 

Bu anlayışların hər biri funksiyanın davranışını təhlil etmək və optimallaşdırma məsələlərini həll etmək üçün istifadə olunur.

### Törəmə və Ekstremum Nöqtələri (Maksimum və Minimum)  genis izah



#### Törəmə nədir?
Törəmə, bir funksiyanın dəyişmə sürətini göstərir. Funksiyanın hansı nöqtələrdə artdığını, azaldığını və ya sabit olduğunu müəyyən etmək üçün istifadə olunur. Riyazi olaraq, bir funksiyanın törəməsi belə ifadə edilir:

$$
f'(x) = \lim_{\Delta x \to 0} \frac{f(x+\Delta x) - f(x)}{\Delta x}
$$

---

#### Maksimum və Minimum nöqtələr nədir?
Bir funksiyanın **maksimum** və ya **minimum** nöqtələri (ümumi olaraq **ekstremum nöqtələri**) həmin funksiyanın ən yüksək və ya ən aşağı dəyərə çatdığı nöqtələrdir.

- **Maksimum nöqtə:** Funksiya bu nöqtədə digər yaxın nöqtələrə nisbətən daha yüksək dəyərə malikdir.
- **Minimum nöqtə:** Funksiya bu nöqtədə digər yaxın nöqtələrə nisbətən daha aşağı dəyərə malikdir.

---

### Ekstremum nöqtələrini tapma qaydası

1. **Törəməni tapın:** 
   $$
   f'(x)
   $$

2. **Kritik nöqtələri tapın:** 
   Törəməni 0-a bərabər edin və bu tənlikdən $x$-ləri hesablayın:
   $$
   f'(x) = 0
   $$

3. **Törəmənin işarə diaqramını qurun:**
   Kritik nöqtələrin solunda və sağında törəmənin işarəsini ($+$ və ya $-$) yoxlayın:
   - $ f'(x) > 0 $: Funksiya artır.
   - $ f'(x) < 0 $: Funksiya azalır.

4. **İkinci törəməni yoxlayın (opsional):** 
   $$
   f''(x)
   $$
   - $ f''(x) > 0 $: Minimum nöqtə.
   - $ f''(x) < 0 $: Maksimum nöqtə.

---

### Məsələlər

#### Misal 1: Sadə funksiya
Tapın: $
f(x) = x^2 - 4x + 3$

**Həll:**
1. **Törəməni tapırıq:**
   $$
   f'(x) = 2x - 4
   $$

2. **Kritik nöqtələri tapırıq:**
   \$$
   2x - 4 = 0 \implies x = 2
   $$

3. **Törəmənin işarəsini yoxlayırıq:**
   - $x < 2$ üçün $f'(x) = 2x - 4 < 0$, yəni funksiya azalır.
   - $x > 2$ üçün $f'(x) = 2x - 4 > 0$, yəni funksiya artır.

4. **Nəticə:**
   $ = 2$ nöqtəsində **minimum** var.

5. **Minimum dəyəri tapırıq:**
   $$
   f(2) = (2)^2 - 4(2) + 3 = -1
   $$

**Cavab: Minimum nöqtə $x = 2$, minimum dəyər $-1$-dir.**

---

#### Misal 2: Maksimum və Minimum
Tapın:
$$
f(x) = -x^3 + 3x^2 + 1
$$

**Həll:**
1. **Törəməni tapırıq:**
   $$
   f'(x) = -3x^2 + 6x
   $$

2. **Kritik nöqtələri tapırıq:**
   $$
   -3x^2 + 6x = 0 \implies x(-3x + 6) = 0 \implies x = 0 \text{ və } x = 2
   $$

3. **Törəmənin işarəsini yoxlayırıq:**
   - $x < 0$: $f'(x) > 0$, funksiya artır.
   - $0 < x < 2$: $f'(x) < 0$, funksiya azalır.
   - $x > 2$: $f'(x) > 0$, funksiya artır.

4. **Nəticə:**
   - $x = 0$: **Maksimum** nöqtə.
   - $x = 2\$: **Minimum** nöqtə.

5. **Maksimum və minimum dəyərləri tapırıq:**
   - $f(0) = -0^3 + 3(0)^2 + 1 = 1$
   - $f(2) = -(2)^3 + 3(2)^2 + 1 = -8 + 12 + 1 = 5$

**Cavab:**
- Maksimum nöqtə $x = 0$, maksimum dəyər $1$.
- Minimum nöqtə $x = 2$, minimum dəyər $5$. 

---

#### Qısa xülasə
1. **Törəmə $f'(x)$ artır/azalan intervalı təyin edir.**
2. **$f'(x) = 0$ kritik nöqtələr ekstremumların yerini tapmağa kömək edir.**
3. **İkinci törəmə ilə maksimum və ya minimum olduğu müəyyən edilir.**

## İkinci Törəmə İlə Maksimum və Minimum Tapılması

İkinci törəmənin maksimum və ya minimum nöqtəni müəyyən etməsi üsulunu **misal** üzərində izah edək:

---

#### Misal:
Funksiyanı nəzərdən keçirək:
$$
f(x) = x^3 - 6x^2 + 9x + 1
$$

**Həll:**

1. **Törəməni tapırıq:**
   $$
   f'(x) = 3x^2 - 12x + 9
   $$

2. **Kritik nöqtələri tapırıq (birinci törəməni 0-a bərabər edirik):**
   $$
   3x^2 - 12x + 9 = 0
   $$

   Bu kvadrat tənliyi həll edirik:
   $$
   3(x^2 - 4x + 3) = 0 \implies 3(x - 1)(x - 3) = 0
   $$
   $$
   x = 1 \quad \text{və} \quad x = 3
   $$

   **Kritik nöqtələr:** $x = 1$ və $x = 3$.

3. **İkinci törəməni tapırıq:**
   İkinci törəmə:
   $$
   f''(x) = 6x - 12
   $$

4. **Hər kritik nöqtə üçün ikinci törəmənin dəyərini yoxlayırıq:**

   - **$x = 1$:**
     $$
     f''(1) = 6(1) - 12 = -6
     $$
     $f''(1) < 0$, bu səbəbdən $x = 1$ nöqtəsi **maksimum nöqtədir**.

   - **$x = 3$:**
     $$
     f''(3) = 6(3) - 12 = 6
     $$
     $f''(3) > 0$, bu səbəbdən $x = 3$ nöqtəsi **minimum nöqtədir**.

5. **Maksimum və minimum dəyərləri tapırıq:**
   Funksiyada bu nöqtələrin dəyərlərini hesablayaq:
   - Maksimum nöqtə üçün:
     $$
     f(1) = (1)^3 - 6(1)^2 + 9(1) + 1 = 1 - 6 + 9 + 1 = 5
     $$

   - Minimum nöqtə üçün:
     $$
     f(3) = (3)^3 - 6(3)^2 + 9(3) + 1 = 27 - 54 + 27 + 1 = 1
     $$

---

### Cavab:
- **Maksimum nöqtə:** $x = 1$, maksimum dəyər $f(1) = 5$.
- **Minimum nöqtə:** $x = 3$, minimum dəyər $f(3) = 1$.

---

### İkinci Törəmənin Məntiqi:

- **$f''(x) > 0$:** Funksiya bu nöqtədə **konkav yuxarı**dır (qabarıq açıq yuxarı). Bu, **minimum** nöqtədir.
- **$f''(x) < 0$:** Funksiya bu nöqtədə **konkav aşağı**dır (qabarıq açıq aşağı). Bu, **maksimum** nöqtədir. 

Bu qayda ikinci törəmənin funksiyanın əyriliyini təsvir etməsindən irəli gəlir.

## x² funksiyasının və onun törəməsinin (2x) qrafiki və animasiya 

````python
%matplotlib notebook

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Funksiya və onun törəməsi
def f(x):
    return x**2

def f_prime(x):
    return 2*x

# X oxu üçün aralıq
x_vals = np.linspace(-10, 10, 200)
y_vals = f(x_vals)

# Animasiya üçün hazırlıq
fig, ax = plt.subplots()
ax.set_xlim(-30, 30)
ax.set_ylim(-80, 80)
ax.set_title("Funksiya və Törəməsi")
ax.set_xlabel("x")
ax.set_ylabel("y")

function_line, = ax.plot([], [], color="blue", label="f(x) = x^2")
tangent_line, = ax.plot([], [], color="red", linestyle="--", label="Tangent")
point, = ax.plot([], [], 'ro', label="Point on f(x)")

def update(frame):
    x_tangent = x_vals[frame]
    y_tangent = y_vals[frame]
    slope = f_prime(x_tangent)
    tangent_y = slope * (x_vals - x_tangent) + y_tangent
    
    function_line.set_data(x_vals, y_vals)
    tangent_line.set_data(x_vals, tangent_y)
    point.set_data([x_tangent], [y_tangent])  # Dəyişiklik burada
    
    return function_line, tangent_line, point


ani = FuncAnimation(fig, update, frames=len(x_vals), interval=20, blit=True)
plt.legend()
plt.show()
````

f(x) = x^2

Qısa olaraq, **x²**-nin törəməsi **x = -2** nöqtəsində **-4** olacaq, amma **x = -2**-də **x²** funksiyasının özü **4**-dür.

## indi bu  (-4)  neyi ifade edir?

**x²** funksiyasının törəməsi **x = -2** nöqtəsində **-4** olması, bu nöqtədə **funksiyanın meylini** (və ya əyilməsini) ifadə edir. Başqa sözlə, törəmə **funksiyanın necə dəyişdiyini** və **x = -2** nöqtəsində funksiyanın qrafikinin necə "qıvrıldığını" göstərir.

2. **Meyl (Slope)**: Əgər **törəmə mənfi**dirsə, funksiyanın qrafiki həmin nöqtədə **azalma meyli** göstərir. Bu halda, **x = -2** nöqtəsində **törəmə -4** olduğuna görə, qrafik həmin nöqtədə **aşağıya doğru meyillənir**.

3. **Praktiki açıqlama**: **Törəmə -4** deməkdir ki, **x = -2** nöqtəsində **funksiya 1 vahidlik dəyişiklik üçün 4 vahid azalma** baş verir. Yəni, **x**-in kiçik bir dəyişməsi (məsələn, **0.1** qədər artması) **y**-in 0.4 qədər azalmasına səbəb olacaq.

### Qrafik üzərindəki təsiri:
- Funksiyanın qrafikində **x = -2** nöqtəsi ətrafında **meyl mənfi** olduğuna görə, qrafik **sola doğru** azaldığı bir təpə nöqtəsinə yaxındır. Bu nöqtə funksiyanın **minimum nöqtəsinə** yaxın ola bilər, çünki qrafikin meyli azalmağa başlayır.

### Nümunə:
Məsələn, əgər **x = -2**-də qrafiki kəsən bir təyyarə düşünsək, onun meyli **-4** olacaq. Yəni, **x**-in kiçik artımı ilə **y** çox sürətlə azalacaq.

### Nəticə:
**Törəmə -4** bu nöqtədə funksiyanın **təzahürünü və dəyişmə sürətini** göstərir. Bu halda, **x = -2** nöqtəsində **qrafikdə meyl azalma yönündədir və sürətlidir**.

## $x^2$ funksiyasının artma və azalma aralıqlar

$x^2$ funksiyasının artma və azalma aralıqları, funksiyanın törəməsinə əsaslanaraq müəyyən edilir.

1. **Funksiyanın törəməsi**:
   
   $f(x) = x^2$ funksiyasının törəməsi:
   $$
   f'(x) = 2x
   $$

2. **Artma və Azalma**:
   - Funksiya **artır** (yəni $f'(x) > 0$) olduğunda, funksiyanın dəyəri nöqtədən nöqtəyə artacaq.
   - Funksiya **azalır** (yəni $f'(x) < 0$) olduğunda, funksiyanın dəyəri nöqtədən nöqtəyə azalacaq.

3. **Törəmənin işarəsi**:
   - **$f'(x) = 2x$ olduğu üçün**:
     - Funksiya **azalır** $x < 0$ (negativ $x$-də), çünki $f'(x) < 0$.
     - Funksiya **artar** $x > 0$ (müsbət $x$-də), çünki $f'(x) > 0$.
     - Funksiya $x = 0$-da sabitdir, çünki burada $f'(x) = 0$.

4. **Artma və Azalma Aralıqları**:
   - **Artma aralığı**: $ (0, +\infty) $
   - **Azalma aralığı**: $ (-\infty, 0) $

Yəni, $x^2$ funksiyası:
- $x = 0$-dan əvvəl azalan (negativ $x$-də),
- $x = 0$-dan sonra artan (müsbət $x$-də) bir funksiyadır.

## $ f(x) = x^2 $ funksiyasının **minimum** və **maksimum** nöqtələri

$ f(x) = x^2 $ funksiyasının **minimum** və **maksimum** nöqtələrini tapmaq üçün əvvəlcə funksiyanın təhlilini edək:

1. **Funksiyanın törəməsi**:
   $$
   f'(x) = 2x
   $$
   Törəmənin sıfır olduğu nöqtə, funksiya üçün mümkün bir ekstremum nöqtəsi ola bilər. Törəməni sıfıra bərabərləyirik:
   $$
   2x = 0 \quad \Rightarrow \quad x = 0
   $$
   Bu nöqtə $x = 0$ funksiyasının ekstremum nöqtəsi ola bilər.

2. **Minimum və Maksimum**:
   - **Minimum nöqtəsi**: Funksiya $x = 0$-da ən aşağı dəyəri alır. Çünki $x^2$ funksiyası, $x$ artdıqca və ya azaldıqca müsbət dəyər alır. Bu, funksiyanın bir **minimum** nöqtəsidir.
     - **Minimum**: $ f(0) = 0^2 = 0 $
   
   - **Maksimum nöqtəsi**: $f(x) = x^2$ funksiyası təkcə aşağıya doğru azaldıqdan sonra yenidən artmağa başlayır. Yəni, funksiyanın maksimum nöqtəsi yoxdur, çünki $x \to \infty$ və ya $ \to -\infty$ olduqda funksiya daha böyük müsbət qiymətlər alır.

**Nəticə**:
- **Minimum** nöqtəsi $x = 0$ və $f(0) = 0$-dır.
- Funksiyanın **maksimum** nöqtəsi yoxdur (sonsuz dəyərə qədər böyüyə bilər).

## İkinci törəmə ilə minimum və maksimum nöqtələrini yoxlamaq

İkinci törəmə ilə minimum və maksimum nöqtələrini yoxlamaq üçün funksiyanın ikinci törəməsini tapmalıyıq.

Funksiya: 
$$
f(x) = x^2
$$

1. **Birinci törəmə**: 
$$
f'(x) = 2x
$$

2. **İkinci törəmə**:
$$
f''(x) = 2
$$

İkinci törəmənin işarəsi funksiyanın necə davranacağına (yəni, minimum və maksimum nöqtələrini) müəyyən edir.

- Əgər $ f''(x) > 0 $ isə, funksiyada **minimum** nöqtəsi var.
- Əgər $ f''(x) < 0 $ isə, funksiyada **maksimum** nöqtəsi var.
- Əgər $ f''(x) = 0 $ isə, bu nöqtə haqqında daha çox təhlil tələb olunur.

Bu halda:
$$
f''(x) = 2 > 0
$$

Deməli, $ f(x) = x^2 $ funksiyasının hər yerdə **minimum** nöqtəsi olduğunu göstərir. Bu, funksiyanın açılan parabola (yuxarıya doğru açılan) olması ilə əlaqədardır. Həmçinin, minimum nöqtəsi $ x = 0 $-da yerləşir, yəni:
$$
f(0) = 0^2 = 0
$$

**Nəticə**:
- **Minimum nöqtəsi**: $ x = 0 $ və $ f(0) = 0 $.
- **Maksimum nöqtəsi yoxdur** (çünki $ f''(x) > 0 $ funksiyanın hər yerdə minimum olduğunu göstərir).

## $ f(x) = x^3 $ funksiyasını təhlil edək.

### 1. **Funksiya və birinci törəmə**
Funksiya $ f(x) = x^3 $ verilmişdir. İndi birinci törəməni hesablayaq:
$$
f'(x) = 3x^2
$$
Bu, funksiyanın meylini göstərir: yəni, funksiyanın artma və azalma hissələrini müəyyən edəcəkdir.

### 2. **Birinci törəmənin sıfır olduğu nöqtə**
İndi $ f'(x) = 0 $ olduğu nöqtəni tapmalıyıq:
$$
3x^2 = 0 \Rightarrow x = 0
$$
Bu, $ f(x) = x^3 $ funksiyasının birinci törəməsinin sıfır olduğu nöqtədir.

### 3. **İkinci törəmə və nöqtənin təhlili**
İndi ikinci törəməni hesablayaq:
$$
f''(x) = 6x
$$
İkinci törəmə $ f''(x) = 6x $ olduğuna görə, bu nöqtənin təhlilini edək:

- $ f''(x) > 0 $ olduğu zaman funksiya bu nöqtədə yuxarıya doğru əyilir və bu, **minimum** nöqtəsini göstərir.
- $ f''(x) < 0 $ olduğu zaman funksiya bu nöqtədə aşağıya doğru əyilir və bu, **maksimum** nöqtəsini göstərir.

Lakin, burada $ x = 0 $-da $ f''(0) = 0 $, buna görə də bu nöqtə nə minimum, nə də maksimum nöqtəsidir. Bu nöqtə **sadalama nöqtəsidir** (inflection point).

### 4. **Funksiyanın artma və azalma aralıqları**
Birinci törəmə $ f'(x) = 3x^2 $ olduğuna görə:

- $ f'(x) > 0 $ olduğu zaman funksiyanın artma hissəsi olacaqdır. Bu, $ x \neq 0 $ üçün doğrudur.
- $ f'(x) = 0 $ yalnız $ x = 0 $ nöqtəsindədir.
- Funksiya hər zaman artmaqdadır, çünki $ f'(x) = 3x^2 $ müsbət olduğu üçün bütün $ x \neq 0 $ nöqtələrində artma mövcuddur.

### 5. **Minimum və maksimum**
$ f(x) = x^3 $ funksiyası heç bir **minimum** və ya **maksimum** nöqtəsinə malik deyil. Çünki bu funksiya nə yuxarıya, nə də aşağıya əyilir; sadəcə bütün $ x $-aralıqlarında artmaqdadır, amma nöqtə üzrə maksimum və minimum yoxdur.

### 6. **Qrafik**:
Bu funksiyanın qrafiki bir kubik funksiyadır. $ x = 0 $-da bir **sadalama nöqtəsi** var, amma nə minimum, nə də maksimum nöqtəsi vardır. Funksiya $ x < 0 $-da aşağıya doğru gedir və $ x > 0 $-da yuxarıya doğru gedir.

### Yekun nəticə:
- **Sadalama nöqtəsi**: $ x = 0 $, burada$ f(x) = 0 $
- Funksiya bütün $ x $-aralıqlarında artmaqdadır.


### **Toxunanın tənliyi və onun törəmə ilə fərqi**
Toxunanın tənliyi və törəmə bir-birilə sıx əlaqəli olsa da, funksiyanı izah etməkdə fərqli məqsədlərə xidmət edir:

---

#### **1. Törəmə:**
- **Tərif:** Funksiyanın müəyyən bir nöqtədəki dəyişmə sürətini və əyriliyini təsvir edir.
- **Forması:** \( f'(x_0) \), yəni həmin nöqtədəki meyildir.
- **Rol:** Yalnız bir rəqəm (meyil) verir, amma funksiyanın başqa nöqtələrdə necə davrandığını birbaşa göstərmir.

---

#### **2. Toxunanın tənliyi:**
- **Tərif:** Əyri bir funksiyanın verilmiş nöqtəsində onu təmsil edən düz xəttin tənliyidir. 
- **Forması:**  
  $$
  f(x) = f(x_0) + f'(x_0)(x - x_0)
  $$
  Burada:
  - $ f(x_0) $ – nöqtənin y-dəki koordinatı (funksiyanın həmin nöqtədəki qiyməti),
  - $ f'(x_0) $ – nöqtədəki törəmə (meyil),
  - $ (x_0, f(x_0)) $ – toxunmanın keçdiyi nöqtədir.

- **Rol:** Bu tənlik həmin nöqtəyə yaxın digər nöqtələrdə funksiyanı xətti şəkildə təxmini təsvir edir.

---

#### **Fərq:**

| **Törəmə ($ f'(x_0) $)**                            | **Toxunanın tənliyi ($ f(x) = f(x_0) + f'(x_0)(x - x_0) $)** |
|-------------------------------------------------------|---------------------------------------------------------------|
| Yalnız funksiyanın dəyişmə sürətini ($ m $) verir.  | Verilmiş nöqtədə funksiyanı düz xətt kimi təmsil edir.         |
| Tək bir qiymət olaraq istifadə edilir.                | Bir tənlik olaraq funksiyanın yaxınlaşmasını göstərir.         |
| **Dinamik dəyişmə fokusundadır.**                     | **Yaxınlaşma və proqnoz üçün istifadə olunur.**                |

---

### **Linear Regression-da törəmə və toxunanın tənliyi necə istifadə olunur?**

Linear regression prosesində həm törəmə, həm də toxunanın tənliyinin fərqli rolları var:

---

#### **1. Linear Regression-da törəmə:**
Törəmə əsasən modeli öyrətmək üçün optimallaşdırma mərhələsində istifadə olunur:
- **Gradient Descent:** Regression xəttinin \( y = mx + b \) tənliyindəki \( m \) (meyil) və \( b \) (kəsişmə) parametrlərini öyrətmək üçün itki funksiyasının törəməsi hesablanır.
  - İtki funksiyası:
    $$
    \text{Loss} = \frac{1}{n} \sum_{i=1}^n (y_i - \hat{y}_i)^2
    $$
    Burada \( \hat{y}_i = mx_i + b \) modelin proqnoz qiymətləridir.
  - Törəmə ilə yenilənmə qaydası:
    $$
    m \gets m - \alpha \frac{\partial (\text{Loss})}{\partial m}, \quad b \gets b - \alpha \frac{\partial (\text{Loss})}{\partial b}.
    $$
  - **Funksiya yaxınlaşması:** Bu proses tədricən itki funksiyasını azaldaraq \( m \) və \( b \)-nin optimal dəyərlərini tapır.

---

#### **2. Linear Regression-da toxunanın tənliyi:**
- Regression xətti əslində verilmiş məlumatların ən yaxşı uyğunlaşma xəttidir. Bu xətt hər bir nöqtədə təxmini olaraq "toxunan" kimi qəbul edilə bilər.
- **Forması:**
  $$
  y = mx + b
  $$
  Burada:
  - $m = f'(x) $: Törəmə nəticəsində tapılmış optimal meyildir.
  - $ b $: Kəsişmə nöqtəsidir.
- **Rol:** Bu xətt, məlumatdakı hər hansı bir $ x $ üçün $ y $-nin proqnozlaşdırılmasını təmin edir.

---

### **Törəmə və toxunanın tənliyinin Linear Regression-dakı ardıcıl rolu:**

1. **Törəmə:** 
   - Gradient Descent ilə $ m $ və $ b $-ni optimallaşdırmaq üçün istifadə olunur.
   - Məlumatdakı nümunələrdən ən yaxşı uyğunlaşmanı təmin edən parametrləri tapır.

2. **Toxunanın tənliyi:**
   - Model hazır olduqdan sonra məlumatların xətti əlaqəsini ifadə edir:
     $$
     y = mx + b.
     $$

---

### **Toxunanın tənliyi forması ilə Linear Regression əlaqəsi:**

Toxunanın ümumi tənliyi:
$$
f(x) = f(x_0) + f'(x_0)(x - x_0)
$$

Linear regression xətti də bu formada yazıla bilər, sadəcə:
- $ x_0 = 0 $-ı qəbul edir (xüsusi hallarda),
- $ f(x_0) = b $,
- $ f'(x_0) = m $.

Beləliklə, regression xətti **toxunanın sadələşmiş tənliyi** kimi başa düşülə bilər:
$$
y = b + mx.
$$

## Törəmə və toxunanın tənliyinin praktiki fərqləri

Törəmə və toxunanın tənliyinin praktiki fərqlərini anlamaq üçün onların istifadə sahələrinə nəzər salaq:

---

### **1. Törəmə (f'(x)) – Funksiyanın dəyişmə sürətini ölçmək:**
#### **Praktiki məqsədi:**
- Törəmə bir funksiya necə sürətlə dəyişir və hansı istiqamətdə dəyişir, onu göstərir.
- **İstifadə halları:**
  - **Optimal qərarlar:** Maksimum və ya minimum nöqtələri tapmaq üçün (məsələn, qazancı artırmaq və ya xərcləri azaltmaq).
  - **Fizika:** Sürət $v = s'(t) $ və təcil $ a = v'(t) $ kimi dəyişmələrin hesablanması.
  - **Maliyyə:** Əyrinin meylini taparaq qiymətlərin və ya gəlirlərin dəyişmə sürətini anlamaq.

#### **Məsələn:**
  $$
  y = x^2 \ \text{funksiyasının} \ f'(x) = 2x \ \text{törəməsi, verilmiş nöqtədə meyli verir.}
  $$
  - $ x = 3 $-də $ f'(3) = 2(3) = 6 $, yəni bu nöqtədə əyrinin "qalxma sürəti" $ 6 $-dır.

---

### **2. Toxunanın tənliyi – Düz xətt olaraq bir nöqtəyə yaxınlaşma:**
#### **Praktiki məqsədi:**
- Əyri bir funksiyanı **sadə bir düz xətt ilə təxmini təsvir etmək**.
- **İstifadə halları:**
  - **Təxmini hesablama:** Çətin və əyri funksiyaları asanlaşdırmaq üçün istifadə edilir (məsələn, mühəndislikdə, fizikada və statistik modellərdə).
  - **Yerində analiz:** Funksiyanın müəyyən bir nöqtədə davranışını düz xətt ilə izah etmək.
  - **Proqnozlar:** Qısa məsafədə funksiyanın qiymətlərini proqnozlaşdırmaq üçün (çünki toxunan əyriyə yaxın bir xəttdir).

#### **Məsələn:**
 $$
  y = x^2 \ \text{üçün toxunanın tənliyi} \ x = 3 \ \text{nöqtəsində:}
  $$
  - $ f'(3) = 6 $, nöqtə isə $ (3, 9) $-dur.
  - Tənlik:
   $$
    y = 6(x - 3) + 9 = 6x - 18 + 9 = 6x - 9.
    $$
  Bu xətt $ x = 3 $-ə yaxın əyrini təmsil edir.

---

### **Praktiki Fərq:**
| **Törəmə**                                           | **Toxunanın tənliyi**                                  |
|------------------------------------------------------|-------------------------------------------------------|
| Funksiyanın dəyişmə sürətini (meyli) izah edir.      | Verilən nöqtədə əyrini düz xətt kimi təsvir edir.      |
| Meyil dəyəri ilə məhdudlaşır ($ m $).              | Tam xətt tənliyi verir ($ y = mx + b $).            |

**Birlikdə istifadəsi:** Törəmə toxunanın meylini (meyl = $ m $) verir və bu meyl toxunanın tənliyini qurmaq üçün istifadə olunur.

## $x^2$ Funskiya, Törəməsi və Toxunanın tənliyi

````python
from sympy import *

# Funksiyanı təyin edirik
x = symbols('x')
f = x**2

# Törəməni hesablayırıq
f_prime = diff(f, x)

# Toxunma nöqtəsini seçirik
x0 = 2  # Toxunanın keçdiyi nöqtə
y0 = f.subs(x, x0)  # Nöqtədə funksiyanın qiyməti
slope = f_prime.subs(x, x0)  # Nöqtədə meyil (törəmə)

# Toxunanın tənliyi
tangent_line = slope * (x - x0) + y0

# Qrafiki çəkirik
p = plot(
    f, 
    f_prime, 
    tangent_line, 
    show=False, 
    legend=True  # Legend-i aktiv edirik
)

# Hər bir element üçün adlar təyin edirik
p[0].label = f           # Funksiyanın adı
p[1].label = f_prime           # Törəmənin adı
p[2].label = tangent_line       # Toxunan xəttin adı

# Qrafik parametrləri
p[0].line_color = 'blue'
p[1].line_color = 'red'
p[2].line_color = 'green'

# Oxları tənzimləyirik
p.ylim = [-10, 10]           # Ordinat oxu üçün məhdudiyyətlər təyin edirik
p.xlim = [-10, 10]           # Absis oxu üçün məhdudiyyətlər təyin edirik

p.show()
````

## **Ən Kiçik Kvadratlar Metodu**

**Ən Kiçik Kvadratlar Metodu** ($ \text{Ordinary Least Squares} $, OLS) xətti regresiyanın əsas prinsipidir. Bu metodun məqsədi verilən məlumat dəstinə ən yaxşı uyğun gələn xətti tapmaqdır. "Ən yaxşı xətt" isə faktiki (real) dəyərlərlə proqnozlaşdırılmış dəyərlər arasındakı fərqlərin kvadratlarının cəmini minimuma endirən xəttdir.

---

### **Əsas Fikir:**
Məlumat dəstində hər bir məlumat nöqtəsi üçün faktiki dəyər ($ y_i $) ilə modelin proqnozlaşdırdığı dəyər ($ \hat{y}_i $) arasında fərq (qalıq və ya səhv) olur. Bu fərqlərin kvadratlarının cəmi minimum olan xətti tapmaq üçün OLS metodundan istifadə edilir.

#### **Formul:**
OLS metodunun məqsədi aşağıdakı itki funksiyasını minimallaşdırmaqdır:
$$
\text{Loss Function} = \sum_{i=1}^n (y_i - \hat{y}_i)^2
$$
Burada:
- $ y_i $: Faktiki (real) dəyərlər,
- $ \hat{y}_i $: Proqnozlaşdırılmış dəyərlər ($ \hat{y}_i = \beta_0 + \beta_1 x_i $),
- $ n $: Məlumat dəstindəki nöqtələrin sayı.

---

### **Metodun Addımları:**

1. **Xətti Model Tənliyi:**
   Ən sadə xətti regresiya modeli belə görünür:
   $$
   y = \beta_0 + \beta_1 x + \epsilon
   $$
   Burada:
   - $ \beta_0 $: Intercept (y oxunu kəsdiyi nöqtə),
   - $ \beta_1 $: Meyil (slope),
   - $ \epsilon $: Qalıq (səhv).

2. **Hədəf:**
   $\beta_0 $ və $\beta_1 $ parametrlərini elə seçmək lazımdır ki, faktiki və proqnozlaşdırılmış dəyərlər arasındakı fərqlərin kvadratlarının cəmi minimum olsun:
   $$
   \min_{\beta_0, \beta_1} \sum_{i=1}^n \left( y_i - (\beta_0 + \beta_1 x_i) \right)^2
   $$

3. **Hesablamalar:**
   $ \beta_0 $ və $ \beta_1 $-nin analitik həlləri **törəmə** vasitəsilə hesablanır:
   $$
   \beta_1 = \frac{\sum_{i=1}^n (x_i - \bar{x})(y_i - \bar{y})}{\sum_{i=1}^n (x_i - \bar{x})^2}
   $$
   $$
   \beta_0 = \bar{y} - \beta_1 \bar{x}
   $$
   Burada:
   - $ \bar{x} $: $ x $-lərin orta qiyməti,
   - $\bar{y} $: $ y $-lərin orta qiyməti.

4. **Proqnozlaşdırma:**
   Tapılan $ \beta_0 $ və $ \beta_1 $ ilə model belə yazılır:
   $$
   \hat{y} = \beta_0 + \beta_1 x
   $$

---

### **Metodun İdeyası Qrafikdə:**

1. **Verilmiş məlumat nöqtələri:** Bunlar (x, y) koordinatlarında qrafikdə işarələnir.
2. **Xətti uyğunlaşma xətti:** Bu xətt $ \hat{y} = \beta_0 + \beta_1 x $ tənliyinə uyğun gəlir.
3. **Qalıqlar ($y_i - \hat{y}_i $:** Faktiki dəyərlə proqnozlaşdırılmış dəyər arasındakı şaquli məsafədir.
   - Bu qalıqların kvadratlarının cəmi minimum olan xətt seçilir.

---

### **Praktiki Məsələn:**

Tutaq ki, bir insanın çəkisi ($ y $) ilə boyu ($ x $) arasında xətti bir əlaqə qurmaq istəyirsiniz. Məlumat belədir:

| Boy (x) (m) | Çəki (y) (kg) |
|-------------|---------------|
| 1.5         | 50            |
| 1.6         | 55            |
| 1.7         | 65            |
| 1.8         | 70            |

Ən kiçik kvadratlar metodu istifadə edərək $ \beta_0 $ və $\beta_1 $-ni hesablamaqla, çəki ilə boy arasındakı ən uyğun xətti tapacaqsınız.

---

### **Əsas Üstünlükləri:**

1. **Sadəlik:** Asan başa düşülür və tətbiq edilir.
2. **Analitik həll:** OLS metodu parametr təxminləri üçün dəqiq analitik ifadələr verir.
3. **Optimal uyğunlaşma:** Verilən məlumatlara ən yaxşı uyğun gələn xətti təmin edir.

---

Əgər qrafiklə izah etmək və ya nümunə kod vermək istəyirsinizsə, əlavə soruşa bilərsiniz.

## Gradient Descent (Eniş Metodu)

### Gradient Descent (Eniş Metodu)

**Gradient Descent**, funksiyanın minimum dəyərini tapmaq üçün istifadə edilən iterativ optimizasiya metodudur. Bu metod, verilmiş funksiyanın ən sürətli enmə istiqamətində hərəkət edərək minimum nöqtəyə yaxınlaşır. Xüsusilə **məhdudiyyətsiz optimizasiya** və **Machine Learning** modellərinin öyrənilməsi üçün geniş istifadə olunur.

---

### Əsas Prinsiplər

1. **İterasiya ilə Yaxınlaşma**:
   - Başlanğıc nöqtəsindən ($ x_0 $) başlayaraq, hər iterasiyada funksiyanın **gradienti** ilə müəyyən olunan istiqamətdə hərəkət edilir.
   - Gradient, funksiyanın ən böyük artım istiqamətini göstərir. **Gradient Descent** isə bu istiqamətin əksinə hərəkət edərək funksiyanın minimumuna yaxınlaşır.

2. **Tənlik**:
   $$
   x_{n+1} = x_n - \eta \cdot \nabla f(x_n)
   $$
   Burada:
   - $ x_n $: Cari nöqtə.
   - $\eta $: **Learning Rate** (öyrənmə sürəti), addım ölçüsünü müəyyən edir.
   - $ \nabla f(x_n) $: Funksiyanın gradienti, istiqaməti və sürəti göstərir.

3. **Learning Rate** ($ \eta $):
   - Çox böyük dəyər olarsa, minimum nöqtədən "ötüb keçmək" mümkündür.
   - Çox kiçik dəyər olarsa, optimizasiya prosesi çox yavaş olar.

---

### İki Əsas Versiya

1. **Batch Gradient Descent**:
   - Tam datasetdən istifadə edərək gradient hesablanır.
   - Daha sabit və dəqiq, lakin böyük datalar üçün yavaşdır.

2. **Stochastic Gradient Descent (SGD)**:
   - Gradient hər dəfə datasetdən təsadüfi seçilmiş bir nümunə ilə hesablanır.
   - Daha sürətlidir, amma çox hallarda funksiyanın ətrafında "titrəmə" effekti yarada bilər.

3. **Mini-batch Gradient Descent**:
   - Dataset kiçik qruplara bölünür və hər dəfə bir qrup ilə gradient hesablanır.
   - Batch və SGD metodlarının yaxşı tərəflərini birləşdirir.

---

### Gradient Descent ilə Funksiya Minimumunun Tapılması

#### Məsələn:
Funksiya: 
$$
f(x) = x^2 + 4x + 4
$$

**Addımlar**:

1. Gradient Hesabı:
   $$
   f'(x) = 2x + 4
   $$

2. İterativ Tənlik:
   $$
   x_{n+1} = x_n - \eta \cdot (2x_n + 4)
   $$

3. Addım-addım Təkrarla:
   Başlanğıc nöqtəsi $ x_0 = 10 $, learning rate $ \eta = 0.1 $.
   - $ x_1 = 10 - 0.1 \cdot (2 \cdot 10 + 4) = 8.6 $
   - $ x_2 = 8.6 - 0.1 \cdot (2 \cdot 8.6 + 4) = 7.28 $
   - Davam edərək, $ x $-in minimum nöqtəyə yaxınlaşdığı müşahidə olunur.

2. **Riyazi Problemlər**:

   - Resursların optimal bölgüsü və ya sistem dizaynının optimallaşdırılması.

Bu metod həm nəzəri olaraq güclüdür, həm də praktiki tətbiqlərdə effektiv nəticələr verir.

````python
### Əyani İzahlı Python Misalı

import numpy as np
import matplotlib.pyplot as plt

# Funksiya və onun törəməsi
def f(x):
    return x**2 + 4*x + 4

def grad_f(x):
    return 2*x + 4

# Parametrlər
x = 10  # Başlanğıc nöqtəsi
eta = 0.1  # Learning rate
iterations = 50  # İterasiya sayı
x_values = [x]

# Gradient Descent İterasiyası
for _ in range(iterations):
    x = x - eta * grad_f(x)
    x_values.append(x)

# Nəticələri çap edək
print(f"Minimum nöqtə: x = {x}, f(x) = {f(x)}")

# Funksiyanın qrafiki
x_plot = np.linspace(-10, 2, 100)
y_plot = f(x_plot)

plt.plot(x_plot, y_plot, label="f(x) = x^2 + 4x + 4")
plt.scatter(x_values, [f(val) for val in x_values], color='red', label="Gradient Descent Yolları")
plt.title("Gradient Descent İterasiyası")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()
````

## Gradient nə deməkdir?

Gradient, çoxölçülü funksiyanın **ən dik eniş** və ya **ən dik yüksəliş istiqamətini** göstərən bir vektorudur. Başqa sözlə, gradient bir nöqtədə funksiyanın dəyərlərinin ən sürətli dəyişdiyi istiqaməti göstərir. Bu anlayış bir funksiyanın meylini və necə dəyişəcəyini ifadə etmək üçün istifadə olunur.

### **Gradient nə deməkdir?**
Təsəvvür edin ki, dağlıq bir ərazidə dayanırsınız. Əgər dağın zirvəsinə qalxmaq istəyirsinizsə, ən dik yuxarı istiqamətdə hərəkət edəcəksiniz. Əgər dağın ətəyinə enmək istəyirsinizsə, ən dik aşağı istiqamətə doğru hərəkət edəcəksiniz. **Gradient** bu vəziyyətdə dağın hündürlük funksiyasının ən dik dəyişdiyi istiqaməti və həmin istiqamətdəki meylin miqdarını göstərir.

Gradient:
- İstiqaməti ilə dəyişimin ən sürətli olduğu xətti göstərir.
- Böyüklüyü ilə həmin istiqamətdəki dəyişimin sürətini göstərir.

---

### **Gradient riyazi olaraq nədir?**
Gradient, skalyar bir funksiyanın (məsələn, $ f(x, y) $) bütün dəyişənlərinə görə qismən törəmələrin vektorudur:

$$
\nabla f(x, y) = \left( \frac{\partial f}{\partial x}, \frac{\partial f}{\partial y} \right)
$$

Burada:  
-$ \frac{\partial f}{\partial x} $: $ x $-ə görə qismən törəmədir (funksiyanın $ x $-də necə dəyişdiyini göstərir).
- $ \frac{\partial f}{\partial y} $: $ y $-ə görə qismən törəmədir (funksiyanın \$ y $-də necə dəyişdiyini göstərir).

---

### **Məna: Gradient nə üçün istifadə olunur?**
1. **İstiqamət göstərir:** Gradient funksiyanın dəyərlərinin **ən sürətli artdığı** istiqaməti göstərir.  
2. **Optimallaşdırma:** Gradient descent kimi metodlarda gradientdən istifadə edərək bir funksiyanın minimum nöqtəsinə yaxınlaşırıq. Bu, çox vaxt **maksimum və ya minimum tapmaq** üçün istifadə edilir.  
3. **Əlverişli sürət:** Gradientin böyüklüyü həmin istiqamətdəki dəyişimin nə qədər sürətli olduğunu ifadə edir.

---

### **Sadə bir nümunə:**
Verilən funksiya:  
$$
f(x, y) = x^2 + y^2
$$

Gradienti hesablayırıq:  
$$
\nabla f(x, y) = \left( \frac{\partial f}{\partial x}, \frac{\partial f}{\partial y} \right)
$$

Qismən törəmələr:
- $ \frac{\partial f}{\partial x} = 2x $
- $ \frac{\partial f}{\partial y} = 2y $

Beləliklə:
$$
\nabla f(x, y) = (2x, 2y)
$$

### Gradientin məntiqi:  
- Nöqtədə $(x, y) = (1, 2) $:  
$$
\nabla f(1, 2) = (2 \cdot 1, 2 \cdot 2) = (2, 4)
$$

Bu vektor:
- İstiqamət olaraq $ (2, 4) $-ü göstərir (daha yüksək $ f(x, y) $ dəyərlərinə gedən istiqamət).
- Böyüklük olaraq $ \sqrt{2^2 + 4^2} = \sqrt{20} $, yəni dəyişimin nə qədər sürətli olduğunu göstərir.

---

### **Gradientin istifadəsi:**
- **Machine Learning**: Neyron şəbəkələrin parametrlərinin optimallaşdırılması üçün gradient descent ilə itki funksiyasının (loss function) minimum nöqtəsi tapılır.
- **Fizika və mühəndislik**: Çoxölçülü sahələrdə qüvvələrin təsirini anlamaq üçün istifadə olunur.
- **Qrafiklər**: Funksiyanın səthində hansı istiqamətdə ən dik dəyişmə baş verdiyini tapmaq.

---

### **Qrafik izah:**
Təsəvvür edin, $ f(x, y) = x^2 + y^2 $ bir parabolik səthdir. Gradientin istiqaməti səthin ən yüksək nöqtəyə və ya ən aşağı nöqtəyə doğru necə dəyişdiyini göstərir.  
Əgər minimuma getmək istəyirsinizsə, gradientin **əksi istiqamətində** hərəkət edirsiniz (gradient descent).
