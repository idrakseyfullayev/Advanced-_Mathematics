## Binomial

### Mesele 1

```python
from scipy.stats import binom

# Verilənlər
n = 22    # İş günlərinin sayı
p = 0.7  # İş vaxtında çatma ehtimalı

# Bütün mümkün k dəyərləri üçün ehtimallar
for k in range(n + 1):
    probability = binom.pmf(k, n, p)*100
    print(f"P(X = {k}) = {probability:.1f}")
```

### Mesele 2

```python
from scipy.stats import binom
import matplotlib.pyplot as plt

n = 10 # tercube sayi
p = 0.9 #  ugurlu olmaq

p_list = []
k_list = []

for k in range(0, n+1):
    pr = binom.pmf(k, n, p)
    print(f"{k} - {pr:.2f}")
    p_list.append(pr)
    k_list.append(k)
    
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

# Histogram çək
plt.hist(x=k_list, weights=p_list, bins=len(k_list), edgecolor="black")

# Y eksenində eksponent formatını dəyiş
plt.gca().yaxis.set_major_formatter(FuncFormatter(lambda x, _: f"{x:.4f}"))

# Başlıqlar
plt.title("Binomial Histogram")
plt.xlabel("Uğur Sayı (k)")
plt.ylabel("Ehtimal")

# Qrafiki göstər
plt.show()

plt.plot(k_list, p_list)
plt.show()

print(sum(p_list[:-2]))
print(sum(p_list[-2:]))
```
