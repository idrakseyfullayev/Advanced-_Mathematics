**Asimptot** (Asymptote) riyaziyyatda bir funksiyanın qrafiki ilə əlaqəli bir anlayışdır və müəyyən bir xəttə yaxınlaşan, amma bu xətti heç vaxt keçməyən bir əyri üçün istifadə edilir. 

Başqa sözlə, **asimptot**, funksiyanın qrafikinin müəyyən bir nöqtəyə yaxlaştıqca daha da yaxınlaşdığı, lakin heç vaxt bu xətti kəsişməyən bir xətdir.

Asimptotlar əsasən üç formada olur:

1. **Yatay Asimptot (Horizontal Asymptote):**
   - Bu tip asimptotda funksiya $x \to \infty $ (və ya $ x \to -\infty $) yaxınlaşdıqca müəyyən bir sabit dəyərə (yəni, $ y = c $) yaxınlaşır.
   - Məsələn, $ f(x) = \frac{1}{x} $ funksiyasının $ y = 0 $ olan yatay asimptotu vardır. $ x \to \infty $ olduqda $ f(x) \to 0 $.
   
   **Nümunə:**
  $$
   f(x) = \frac{1}{x}
   $$
   $ x \to \infty $ olduqda \( f(x) $ sıfıra yaxınlaşır, amma heç vaxt sıfır olmaz.

2. **Şaquli Asimptot (Vertical Asymptote):**
   - Bu tip asimptotda funksiya müəyyən bir $ x = c $ nöqtəsinə yaxınlaşdıqca sonsuzluğa doğru getməkdədir. Bu halda, $ x = c $ şaquli asimptot xətti olur.
   - Məsələn, $ f(x) = \frac{1}{x-2} $ funksiyasının $ x = 2 $ nöqtəsində şaquli asimptotu vardır, çünki bu nöqtəyə yaxınlaşdıqda funksiyanın dəyəri sonsuzluğa gedir.
   
   **Nümunə:**
  $$
   f(x) = \frac{1}{x-2}
   $$
   Bu funksiya $ x = 2 $ nöqtəsində şaquli asimptot yaradır, çünki $ x \to 2 $ olduqda $ f(x) $ sonsuz olur.

3. **Əyri Asimptot (Oblique Asymptote / Slant Asymptote):**
   - Bu, xüsusilə bir funksiya əyri asimptot ilə təyin edildikdə görülür. Əyri asimptotlar, funksiyanın qrafiki müəyyən bir x-in sonsuza yaxınlaşdığı zaman bir doğruya yaxınlaşır, amma bu doğru nə şaquli, nə də yatay olur.
   - Əyri asimptotlar, ən çox polinom funksiyalarında, xüsusilə qismən bölünmə zamanı rast gəlinir.
   
   **Nümunə:**
   $$
   f(x) = \frac{x^2 + 1}{x}
   $$
   Bu funksiyanın $ x \to \infty $ olduğu zaman $ y = x $ bir əyri asimptotu vardır.

### Asimptotların Tapılması:
Asimptotları tapmaq üçün funksiyanın limitlərinə baxmaq lazımdır:
- **Yatay asimptot üçün**: Funksiyanın $ x \to \infty $ və $ x \to -\infty $ limitlərinə baxılır.
- **Şaquli asimptot üçün**: Funksiyanı müəyyən bir $ x = c $ nöqtəsinə yaxınlaşdırırıq və orada funksiyanın qiymətinin sonsuz olub olmadığını yoxlayırıq.
- **Əyri asimptot üçün**: Funksiyanın uzun müddətli davranışını tədqiq edirik və əgər əyri bir xəttə yaxınlaşırsa, bu asimptot olur.

**Qısaca**: Asimptotlar, bir funksiya üçün o qədər vacibdir ki, funksiyanın qrafiki heç vaxt bu xətləri keçməsə də, onlar funksiyanın uzun müddətli davranışını göstərir.

### y = (ax + b)  / (cx + d)

```python
from sympy import *
from sympy.plotting import plot

# Simvolik dəyişənləri təyin edirik
x, a, b, c, d = symbols('x a b c d')

# Riyazi ifadəni müəyyənləşdiririk
y = (a*x + b) / (c*x + d)
# a və b üçün nümunəvi dəyərləri təyin edirik
# a < 0 şərtini təmin etmək üçün a = -1 təyin edirik
a_val = 2  # a mənfi dəyər
b_val = 2   # b üçün bir müsbət dəyər
c_val = 2
d_val = 2

# y ifadəsinə təyin edilmiş dəyərləri daxil edirik
y_val = y.subs({a: a_val, b: b_val, c: c_val, d:d_val})

# Qrafiki çəkirik
p = plot(y_val, show=False, title=f"Qrafik: a={a_val}, b={b_val}", ylabel="y", xlabel="x")

# Oxları tənzimləyirik
p[0].line_color = 'blue'  # Xəttin rəngini dəyişirik
p.ylim = [-10, 10]           # Ordinat oxu üçün məhdudiyyətlər təyin edirik
p.xlim = [-10, 10]         # Absis oxu üçün məhdudiyyətlər təyin edirik
p.show()
```
