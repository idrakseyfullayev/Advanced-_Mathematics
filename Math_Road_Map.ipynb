Data science sahəsində uğurlu olmaq üçün bəzi əsas riyazi biliklər lazımdır. Aşağıda mövzular ardıcıllıqla verilmişdir. Bu ardıcıllıq mövzuların asanlıqdan çətinliyə doğru keçidini əks etdirir.  

### **1. Əsas Riyazi Biliklər**
   - Ədədlər və Əməliyyatlar  
   - Funksiyalar və qrafiklər $ + $
   - Əsas cəbr: tənliklər və bərabərsizliklər  

### **2. Statistik və Ehtimal**
   - **Statistika**  
     - Descriptive statistics: Orta, Median, Mod, Variasiya, Standart sapma  
     - Dağılımlar: Normal, Binomial, Poisson  
     - Outlier-lərin aşkarlanması  
     - Korelyasiya və Kovariasiya  
   - **Ehtimal nəzəriyyəsi**  
     - Əsas anlayışlar: hadisələr, ehtimalların toplanması/məhsulu qaydası  
     - Şərti ehtimal  
     - Bayes nəzəriyyəsi  

### **3. Cəbr və Lineer Cəbr**
   - Matrislər və vektorlar  
   - Matris əməliyyatları: toplama, çıxma, vurma  
   - Ters matris və determinant  
   - Xətti tənliklər sistemləri  
   - Özvektorlar və özqiymətlər (eigenvectors, eigenvalues)  
   - Singular Value Decomposition (SVD)  

### **4. Kalkulyus (Hesablama Riyaziyyatı)**
   - Törəmə və onun tətbiqləri (maksimum və minimum nöqtələrin tapılması)  $ + $
   - İnteqral və onun tətbiqləri (sahə hesablamaları)  $ + $
   - Qismən törəmələr (multivariat funksiyalar üçün)  $ + $
   - Optimizasiya metodları  

### **5. Diskret Riyaziyyat**
   - Kombinatorika  
   - Qraf nəzəriyyəsi (Graph Theory)  
   - Loqika və dəstlər nəzəriyyəsi  

### **6. Riyazi Statistikalar**
   - Hipotez yoxlamaları (t-test, z-test, chi-square test)  
   - Güvən intervalları  
   - Regressiya analizi: sadə və çoxlu lineer regressiya  

### **7. Ehtimal Dağılımları və Statistik Modellər**
   - Əsas ehtimal paylanmaları (Normal, Exponential, Uniform, Bernoulli və s.)  
   - Maksimum ehtimal qiymətləndirilməsi (Maximum Likelihood Estimation, MLE)  

### **8. Optimallaşdırma və Riyazi Modellər**
   - Gradient Descent  
   - Convex Optimization  
   - Lagrange Multipliers  

### **9. Dərinləşmiş Statistik və Machine Learning Məkanları**
   - Neyron şəbəkələr üçün aktivasiya funksiyaları (Sigmoid, ReLU, Softmax)  
   - Loss Functions: MSE, Cross-Entropy və s.  
   - Tənzimləmə (L1 və L2 normları, Ridge və Lasso)  
   - Feature Engineering və PCA (Principal Component Analysis)  

### **10. Diferensial Tənliklər (İrəliləmiş Mövzu)**
   - Sadə diferensial tənliklər  
   - Lineer və qeyri-lineer tənliklər  

### **11. Zaman Sırası Analizi (Time Series Analysis)**  
   - Autocorrelation və Partial Autocorrelation  
   - ARIMA modelləri  
   - Trend, Mövsümi və Residual analizləri  

### **12. Qrafik Statistikalar və Visualization**
   - Histogramlar, Boxplot, Heatmaplər  
   - Distribution və Relationship analizləri  

Hər bir mövzunu öyrənərkən praktiki nümunələr üzərində işləmək, Python, R və ya digər data science alətlərindən istifadə etmək öyrənməni asanlaşdıracaqdır. Əvvəlcə əsaslardan başlayın və tədricən irəliləyin!
