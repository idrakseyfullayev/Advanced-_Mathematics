## Derivative

### $f(x) = x^2$ $\quad$ $\Longrightarrow$ $\quad$ $f'(x) = 2x$

```python
from sympy import *
from sympy.plotting import plot

# Simvolik dəyişəni təyin edirik
x = symbols('x')

# Riyazi ifadəni müəyyənləşdiririk
y1 = x**2
y2 = 2*x 

# Qrafiki çəkirik
p = plot(y1, y2, show=False)

# Oxları tənzimləyirik
p[0].line_color = 'blue'
p[0].label = 'y1 = x^2'
p[1].line_color = 'red'
p[1].label = 'y2 = 2x'
p.legend = True
p.ylim = [-10, 10]           # Ordinat oxu üçün məhdudiyyətlər təyin edirik
p.xlim = [-10, 10]         # Absis oxu üçün məhdudiyyətlər təyin edirik

p.show()
```

## $ f(x) = x^2 $ Limit_Differentiation

```python
from sympy import *
a1, h, x = symbols('a1 h x')
a2 = a1 + h

# def f(x): return x**2
f_x = x**2

formula = ( f_x.subs(x, a2) - f_x.subs(x, a1) ) / (a2 - a1)
print(formula)
diff_lim = limit(formula, h, 0)

print(diff_lim, type(diff_lim), sep='\n')

print(diff_lim.subs(a1, 4))
```

## $ f(x) = x^2 $ Left_limit and Right_limit Differentiation

```python
from sympy import *
a1, h, x = symbols('a1 h x')
a2 = a1 + h

# def f(x): return x**2
f_x = x**2

formula = ( f_x.subs(x, a2) - f_x.subs(x, a1) ) / (a2 - a1)
diff_lim_left = limit(formula, h, -0.0000000001)
diff_lim_right = limit(formula, h, 0.0000000001)

print(diff_lim_left, diff_lim_right, sep='\n')
print(diff_lim_left.subs(a1, 4), diff_lim_right.subs(a1, 4), sep='\n')
print( ( diff_lim_left.subs(a1, 4) + diff_lim_right.subs(a1, 4) ) / 2 )
```

## With def(x)

```python
from sympy import *
a1, h, x = symbols('a1 h x')
a2 = a1 + h

def f(x): return x**2

formula = ( f(a2) - f(a1) ) / (a2 - a1)
print(formula)
diff_lim = limit(formula, h, 0)

print(diff_lim)

print(diff_lim.subs(a1, 4))
```

## $ f(x) = x^2 $ Differentiation without limit

```python
def der_f_x(func, x, h):
    return ( func(x + h) - func(x)) / ( (x+h) - x )

def f_x(x):
    return x**2

der_f_x(f_x, 2, 0.0000000001)
```

```python
def der_f_x(func, x, h):
    return ( func(x + h) - func(x)) / ( (x+h) - x )

def f_x(x):
    return x**2

der_f_x(f_x, 2.1, 0.0000000000001), der_f_x(f_x, 2.1, 0.000000000001)
```

## $ f(x) = x^2 $ Differentiation

```python
from sympy import *

# Simvolik dəyişənlər
x = symbols('x')

# Funksiya təyin edilməsi
f_x = x**2

# Törəmənin hesablanması
f_diff = diff(f_x, x)
# f_diff = diff(f_x)

# Nəticələrin çapı
print(f_diff)  # Törəmənin ümumi ifadəsi
print(f_diff.subs(x, 4))  # x = 4 üçün törəmənin qiyməti
```

## Chain Rule (Zəncir Qaydası)

- $y = x^2$ 
- $z = y^3$

```python
from sympy import *
x, y, z = symbols('x y z')

y = x**2
z = y**3

z
```

```python
from sympy import *
x, y, z = symbols('x y z')

y = x**2
z = y**3

dz_dx = diff(z, x)
dz_dx
```

- $y = x^3 - 2x +1$ 
- $z = y^5 - 7y + 6$

```python
from sympy import *
x, y, z = symbols('x y z')

y = x**3 - 2*x + 1
z = y**5 - 7*y + 6

dz_dx = diff(z, x)
dz_dx
```

```python
dz_dx.subs(x, 7)
```

```python
from sympy import *
x, y, z = symbols('x y z')

f_y = x**3 - 2*x + 1
dy_dx = diff(f_y, x)

f_z = y**5 - 7*y + 6
dz_dy = diff(f_z, y)

dz_dx = (dy_dx * dz_dy)
dz_dx
```

```python
dz_dx.subs(y, f_y)
```

```python
dz_dx.subs(y, f_y).subs(x, 7)
```

```python
from sympy import *
x, y, z = symbols('x y z')

f_y = x**3 - 2*x + 1
f_z = y**5 - 7*y + 6

dz_dx = diff(f_z.subs(y, f_y))
dz_dx
```

```python
dz_dx.subs(x, 7)
```
