## $ f(x) = x^2 $ Limit_Differentiation

```python
from sympy import *
a1, h, x = symbols('a1 h x')
a2 = a1 + h

# def f(x): return x**2
f_x = x**2

formula = ( f_x.subs(x, a2) - f_x.subs(x, a1) ) / (a2 - a1)
print(formula)
diff_lim = limit(formula, h, 0)

print(diff_lim, type(diff_lim), sep='\n')

print(diff_lim.subs(a1, 4))
```

## $ f(x) = x^2 $ Left_limit and Right_limit Differentiation

```python
from sympy import *
a1, h, x = symbols('a1 h x')
a2 = a1 + h

# def f(x): return x**2
f_x = x**2

formula = ( f_x.subs(x, a2) - f_x.subs(x, a1) ) / (a2 - a1)
diff_lim_left = limit(formula, h, -0.0000000001)
diff_lim_right = limit(formula, h, 0.0000000001)

print(diff_lim_left, diff_lim_right, sep='\n')
print(diff_lim_left.subs(a1, 4), diff_lim_right.subs(a1, 4), sep='\n')
print( ( diff_lim_left.subs(a1, 4) + diff_lim_right.subs(a1, 4) ) / 2 )
```

## Törəmə (Artma, Azalma aralığı, Böhran nöqtələri ,Ekstremium nöqtələr,Parçada ən böyük qiymət)

### Artma, Azalma aralığı

### $f(x) = 2x^2 - 12x$

```python
from sympy import symbols, diff, solve, solve_univariate_inequality, oo

# Funksiyanı təyin et
x = symbols('x')
f_x = 2*x**2 - 12*x

# Törəməni tap
f_prime = diff(f_x, x)

# Törəmənin sıfır nöqtələrini tap
critical_points = solve(f_prime, x)

# Törəmənin azalma aralığı üçün analiz (f'(x) < 0)
decreasing_interval = solve_univariate_inequality(f_prime < 0, x)

# Törəmənin artma aralığı üçün analiz (f'(x) > 0)
increasing_interval = solve_univariate_inequality(f_prime > 0, x)

# Nəticələri uyğun formatda çap et
print(f"Kritik noqteler: {critical_points}")
print(f"Azalma aralığı: {decreasing_interval}")
print(f"Artma aralığı: {increasing_interval}")
```

### $f(x) = 3x^2 - x^3 + 3, \quad x \in [-1, 1]$


```python
from sympy import symbols, diff, solve, Interval, Min

# Funksiyanı təyin et
x = symbols('x')
f_x = 3*x**2 - x**3 + 3

# Funksiyanın törəməsini tap
f_prime = diff(f_x, x)

# Böhran nöqtələrini tap (törəməni sıfıra bərabərləşdir)
critical_points = solve(f_prime, x)
print('Bohran noqteleri', critical_points)

# İntervalda funksiyanın son nöqtələrdə və böhran nöqtələrindəki qiymətlərini tap
interval = Interval(-1, 1)
values = [f_x.subs(x, point) for point in critical_points if point in interval]
print(values)
values.extend([f_x.subs(x, interval.start), f_x.subs(x, interval.end)])
print(values)

# Ən kiçik qiyməti tap
min_value = min(values)
max_value = max(values)

# Nəticəni çap et

print("Ən kiçik qiymət:", min_value)
print("Ən böyük qiymət:", max_value)
```

## $x^2$ Funskiya, Törəməsi və Toxunanın tənliyi

```python
from sympy import *

# Funksiyanı təyin edirik
x = symbols('x')
f = x**2

# Törəməni hesablayırıq
f_prime = diff(f, x)

# Toxunma nöqtəsini seçirik
x0 = 2  # Toxunanın keçdiyi nöqtə
y0 = f.subs(x, x0)  # Nöqtədə funksiyanın qiyməti
slope = f_prime.subs(x, x0)  # Nöqtədə meyil (törəmə)

# Toxunanın tənliyi
tangent_line = slope * (x - x0) + y0

# Qrafiki çəkirik
p = plot(
    f, 
    f_prime, 
    tangent_line, 
    show=False, 
    legend=True  # Legend-i aktiv edirik
)

# Hər bir element üçün adlar təyin edirik
p[0].label = f           # Funksiyanın adı
p[1].label = f_prime           # Törəmənin adı
p[2].label = tangent_line       # Toxunan xəttin adı

# Qrafik parametrləri
p[0].line_color = 'blue'
p[1].line_color = 'red'
p[2].line_color = 'green'

# Oxları tənzimləyirik
p.ylim = [-10, 10]           # Ordinat oxu üçün məhdudiyyətlər təyin edirik
p.xlim = [-10, 10]           # Absis oxu üçün məhdudiyyətlər təyin edirik

p.show()
```

## Mesele helli


**1. Tam səthinin sahəsi $ 24\pi \, \text{sm}^2 $ olan silindrin oturacağının radiusu nə qədər olmalıdır ki, onun həcmi ən böyük olsun?**

A) 1 sm  
B) 4 sm  
C) 2 sm  
D) 1,5 sm  
E) 3 sm

```python
from sympy import symbols, Eq, solve, diff, pi

# Parametrləri müəyyən edək
r, h = symbols('r h', real=True, positive=True)

# Verilmiş tam səth sahəsi
S = 2 * pi * r**2 + 2 * pi * r * h
S_value = 24 * pi

# Həcm ifadəsi
V = pi * r**2 * h

# h-ni S = 24*pi-dən tapırıq:
# 1. Eq(S, S_value): Burada S = S_value şəklində bir tənlik yaradılır. Eq() funksiyası bərabərlik ifadəsi qurur.
# 2. solve(Eq(...), h): solve() funksiyası bu tənliyi h dəyişəni üzrə həll edir.
#-  solve() funksiyası həmişə ifadəni sıfıra bərabər hesablayaraq dəyişənləri tapır.
# 3. [0]: solve() funksiyası həllər siyahısı qaytarır. [0] ilə bu siyahının ilk həllini seçirik.
# h_expr: h üçün tapılan ifadə bu dəyişənə mənimsədilir.
h_expr = solve(Eq(S, S_value), h)[0]

# h-ni həcm funksiyasına yerləşdiririk
V_subs = V.subs(h, h_expr)

# Həcmin törəməsini radiusa görə alıb sıfıra bərabər edirik
V_prime = diff(V_subs, r)
critical_r = solve(V_prime, r)

# Real və müsbət olan radiusu seçirik
radius = [r_val for r_val in critical_r if r_val > 0][0]

# Nəticə
radius.evalf()
```

13. Bir tərəfi divar olan düzbucaqlı şəklində olan əkin yerinin divar olmayan ətrafına 60 metr uzunluğunda məftil tor çəkilib. Bu əkin yerinin sahəsi ən çox neçə m² olar?

```python
from sympy import symbols, Eq, solve

# Dəyişənləri təyin edirik
x, y = symbols('x y')

# Məftil uzunluğu tənliyi: 2x + y = 60
L = 2*x + y
L_value = 60

# Sahə funksiyasını tərtib edirik: S = x * y
S = x * y

# x-i tənlikdən y-nin ifadəsi ilə əvəz edirik
y_expr = solve(Eq(L, L_value), y)[0]  # y-ni tapırıq

# Sahə funksiyasını sadələşdiririk
S_expr = S.subs(y, y_expr)

# Sahəni maksimum etmək üçün derivativini tapırıq
S_prime = S_expr.diff(x)

# Derivativini sıfıra bərabər edirik
optimal_x = solve(S_prime, x)[0]

# optimal y-ni tapırıq
optimal_y = y_expr.subs(x, optimal_x)

# Maksimum sahəni hesablama
max_area = S_expr.subs(x, optimal_x).subs(y, optimal_y)

# Nəticələri çap edirik
print(f"Əkin yerinin maksimum sahəsi: {max_area} m²")
print(f"Optimal uzunluqlar: x = {optimal_x} m, y = {optimal_y} m")
```

### Tapşırıq mətni:¶

3. 48 ədədini elə iki müsbət ədədin cəmi şəklində göstərin ki, onlardan birinin kubu ilə digərinin kvadratı cəmi ən kiçik olsun.

A) 15; 33  
B)  $\frac{14}{3}$, $\frac{130}{3}$  
C) $\frac{17}{3}$, $\frac{127}{3}$  
D) 16; 32  
E) \$\frac{16}{3}$, $\frac{128}{3}$

```python
from sympy import symbols, Eq, solve, diff, Rational

# Dəyişənləri təyin edirik
x = symbols('x')

#  x + y = 48
# y-ni x üzərindən ifadə edirik: y = 48 - x
y = 48 - x

# Hədəf funksiyasını təyin edirik: x^3 + y^2
f = x**3 + (48 - x)**2

# Funksiyanın derivativini alırıq
f_prime = diff(f, x)

# Derivativi sıfıra bərabər edirik
optimal_x = solve(f_prime, x)

# Müsbət həlləri seçirik və nəticəni kesr formasında əldə edirik
optimal_x = [val.evalf() for val in optimal_x if val > 0]

# Kesr formasında optimal x və y-ni tapırıq
optimal_x_rational = [Rational(val).limit_denominator() for val in optimal_x]
optimal_y_rational = [Rational(48 - val).limit_denominator() for val in optimal_x_rational]

# Minimum qiyməti hesablayırıq və kesr şəklində göstəririk
min_value_rational = [Rational(f.subs(x, val).subs(y, 48 - val)).limit_denominator() for val in optimal_x_rational]

# Nəticələri çap edirik
for i in range(len(optimal_x_rational)):
    print(f"Optimal x: {optimal_x_rational[i]}, Optimal y: {optimal_y_rational[i]}, \
          Minimum qiymət: {min_value_rational[i]}")
```
