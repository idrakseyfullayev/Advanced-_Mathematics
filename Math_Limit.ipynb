## Limit

###  $y=\frac{x + 100}{x + 500}$

```python
from sympy import *

x = symbols('x')

y = (x+100) / (x+500)

# Qrafiki çəkirik
p = plot(y, show=False)

# Oxları tənzimləyirik
# p[0].line_color = 'blue'  # Xəttin rəngini dəyişirik
p.ylim = [0, 1]           # Ordinat oxu üçün məhdudiyyətlər təyin edirik
p.xlim = [-50, 50]         # Absis oxu üçün məhdudiyyətlər təyin edirik
p.show()
```

### $lim_{x \to \infty} \frac{x + 100}{x + 500}$
 
- limitini hesablamaq üçün aşağıdakı kodu yaza bilərsiniz:

```python
from sympy import symbols, limit, oo

# Dəyişən və funksiyanı təyin et
x = symbols('x')
y = (x + 100) / (x + 500)

# Sonsuzluqda limit
limit_inf = limit(y, x, oo)

# Nəticələri göstər
limit_inf
```

```python
import numpy as np
import matplotlib.pyplot as plt

# x üçün dəyərlər
x_vals = np.linspace(1, 1000, 500)  # x-ın dəyərləri 1-dən 1000-ə qədər

# y funksiyasının hesablanması
y_vals = (x_vals + 100) / (x_vals + 500)

# Qrafiki qurmaq
plt.plot(x_vals, y_vals, label=r"$y = \frac{x + 100}{x + 500}$")
plt.axhline(1, color='r', linestyle='--', label="Limit: y=1")  # Limit xətti
plt.title("Qrafik: y = (x + 100) / (x + 500)")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

### $lim_{x \to \infty} \frac{1}{x}$

```python
from sympy import symbols, limit, oo

# Dəyişən və funksiyanı təyin et
x = symbols('x')

y = 1/x

# Sonsuzluqda limit
limit_inf = limit(y, x, oo, dir='+-')

# Nəticələri göstər
limit_inf
```

### $lim_{x \to 0} \frac{1}{x}$

```python
from sympy import symbols, limit, oo

# Dəyişən və funksiyanı təyin et
x = symbols('x')

y = 1/x

# Sonsuzluqda limit
limit_0 = limit(y, x, 0, dir='+-')

# Nəticələri göstər
limit_0
```

$$ \lim_{n \to \infty} \left( 1 + \frac{1}{n} \right)^n = e \approx 2.71828 $$
$$ \quad \text{where} \quad n = \frac{1}{m}, \quad m = \frac{1}{n} \quad \Rightarrow \quad  \quad m = \frac{1}{n\to \infty} \quad  \text{and} \quad m \to 0 $$
$$ \quad \Rightarrow \quad \lim_{m \to 0} \left( 1 + m \right)^{\frac{1}{m}} = e $$


### $ \lim_{n \to \infty} \left( 1 + \frac{1}{n} \right)^n$

```python
from sympy import *

# Dəyişən və funksiyanı təyin et
x = symbols('x')

y = (1 + 1/x)**x

# Sonsuzluqda limit
limit_inf = limit(y, x, oo, dir='+-')

# Nəticələri göstər
limit_inf, limit_inf.evalf()
```

### $\lim_{m \to 0} \left( 1 + m \right)^{\frac{1}{m}}$

```python
from sympy import *

# Dəyişən və funksiyanı təyin et
x = symbols('x')

y = (1 + x)**(1/x)

# Sonsuzluqda limit
limit_0 = limit(y, x, 0, dir='+-')

# Nəticələri göstər
limit_0, limit_0.evalf()
```

### $ y = \left( 1 + \frac{1}{n} \right)^n$

```python
from sympy import *

x = symbols('x')

y = (1 + 1/x)**x

# Qrafiki çəkirik
p = plot(y, show=False)

# Oxları tənzimləyirik
# p[0].line_color = 'blue'  # Xəttin rəngini dəyişirik
p.ylim = [-1, 10]           # Ordinat oxu üçün məhdudiyyətlər təyin edirik
p.xlim = [-10, 10]         # Absis oxu üçün məhdudiyyətlər təyin edirik
p.show()
```

```python
import sympy as sp
from sympy.plotting import plot

# n dəyişəni
n = sp.symbols('n')

# funksiyanı təyin et
y_expr = (1 + 1/n)**n

# Qrafiki qurmaq
plot(y_expr, (n, 1, 1000), title=r"$y = \left(1 + \frac{1}{n} \right)^n$", ylabel="y")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# n üçün dəyərlər
n = np.linspace(-12, 12, 500)

# funksiyanı təyin et
y_expr = (1 + 1/n)**n


# Qrafiki qurmaq
plt.plot(n, y_expr, label=r"$y = \left(1 + \frac{1}{n} \right)^n$")
plt.axhline(2.71828, color='r', linestyle='--', label="Limit: e = 2.71828")  # Limit xətti
plt.title(r"Qrafik: $y = \left( 1 + \frac{1}{n} \right)^n$")
plt.xlim(-2, 2)
plt.ylim(-4, 6)
plt.xlabel("n")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```
