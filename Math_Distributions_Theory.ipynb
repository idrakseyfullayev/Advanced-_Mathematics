## Binomial Paylanma Nədir?

### **Binomial Paylanma**

**Binomial paylanma**, iki nəticəli (uğur və ya uğursuzluq) təcrübələrdən ibarət bir sıra nəticələrin ehtimalını təsvir edən statistik paylanmadır. Binomial paylanma adətən təkrarlanan təcrübələrdə istifadə olunur və hər təcrübədə müstəqil olaraq uğur və ya uğursuzluq baş verə bilər.

---

### **Əsas Xüsusiyyətlər:**
1. **Müəyyən edilmiş təcrübələrin sayı ($ n $):**
   Təcrübələr sabit sayda təkrarlanır.
   
2. **Uğur ehtimalı ($ p $):**
   Hər bir təcrübədə uğurun baş vermə ehtimalı sabitdir.
   
3. **Müstəqillik:**
   Təcrübələr bir-birindən müstəqildir, yəni bir təcrübənin nəticəsi digərinə təsir etmir.

4. **İki nəticə:**
   Hər bir təcrübənin yalnız iki mümkün nəticəsi var: uğur ($ k $) və uğursuzluq ($ n-k $).

---

### **Binomial Paylanma Formulu:**

Bir binomial paylanma təcrübəsində $X $-in $ k $ uğur almasının ehtimalı belə verilir:

$$
P(X = k) = \binom{n}{k} p^k (1-p)^{n-k}
$$

Burada:
- $ \binom{n}{k} = \frac{n!}{k! (n-k)!} $: Kombinasiyalar,
- $ p $: Uğur ehtimalı,
- $ 1-p $: Uğursuzluq ehtimalı,
- $ k $: Uğurların sayı,
- $ n $: Təcrübələrin sayı.

---

### **Binomial Paylanmanın Xüsusiyyətləri:**

1. **Orta (Mean):**
   $$
   \mu = n \cdot p
   $$

2. **Dispersiya (Variance):**
   $$
   \sigma^2 = n \cdot p \cdot (1-p)
   $$

3. **Standart Sapma (Standard Deviation):**
   $$
   \sigma = \sqrt{n \cdot p \cdot (1-p)}
   $$

---

### **Binomial Paylanmaya Nümunə:**

**Misal:** Bir tələbə 10 suallıq bir test edir. Hər suala doğru cavab vermə ehtimalı $ p = 0.7 $-dir. Testdə doğru cavabların sayını təsvir etmək üçün binomial paylanma istifadə edilə bilər.

1. **Verilənlər:**
   - $ n = 10 $ (təcrübələrin sayı),
   - $ p = 0.7 $ (uğur ehtimalı),
   - $ k $ (doğru cavabların sayı).

2. **Ehtimalın Hesablanması (Məsələn, $ k = 7 $):**
   $$
   P(X = 7) = \binom{10}{7} (0.7)^7 (1-0.7)^3
   $$
   Burada:
   $$
   \binom{10}{7} = \frac{10!}{7! \cdot (10-7)!} = 120
   $$
   
   
Hesablamanı davam etdirərək nəticə belədir:

1. Kombinasiya ($ \binom{10}{7} $):
   $$
   \binom{10}{7} = 120
   $$

2. Binomial ehtimal:
   $$
   P(X = 7) = 120 \cdot (0.7)^7 \cdot (0.3)^3 \approx 0.2668
   $$

**Cavab:**

**Bu o deməkdir ki, tələbənin  10 sualdan 7-nin doğru cavab verilməsi ehtimalı təqribən **0.2668** (və ya 26.68%) təşkil edir.**

---

### **Python ilə Binomial Paylanma Hesablaması:**

```python
from scipy.stats import binom

# Parametrlər
n = 10  # Təcrübə sayı
p = 0.7  # Uğur ehtimalı
k = 7  # Uğur sayı

# Binomial ehtimal
probability = binom.pmf(k, n, p)
print(f"P(X = {k}) = {probability:.4f}")
```

---

### **Binomial Paylanmanın İstifadə Sahələri:**

1. **Test və sınaq analizləri:** Məsələn, məhsul keyfiyyət testlərində müəyyən bir məhsulun uğurlu olub olmadığını yoxlamaq.
2. **Tibbi təcrübələr:** Müəyyən bir dərmanın uğurlu olub olmadığını müəyyənləşdirmək.
3. **Sosial araşdırmalar:** Seçki nəticələrinin proqnozlaşdırılması üçün.

Binomial paylanma real həyatda geniş istifadə olunan bir ehtimal paylanmasıdır. Əgər əlavə nümunə və ya xüsusi bir mövzu haqqında məlumat istəsəniz, əlavə izah edə bilərəm!

## Beta Paylanma

Beta paylanma, statistikada bir ehtimal paylanmasıdır və [0,1] intervalında bir təsadüfi dəyişənin ehtimallarını modelləşdirmək üçün istifadə olunur. Bu paylanma bir çox sahədə, o cümlədən bayes statistikası, qərar nəzəriyyəsi və simulyasiya modellərində istifadə olunur.

### Beta Paylanmanın Tərifi
Beta paylanma iki müsbət parametrə, $ \alpha $ və $ \beta $-ya bağlıdır və onların qiymətləri paylanmanın formasını təyin edir. Paylanmanın sıxlıq funksiyası (PDF) belə ifadə edilir:

$$
f(x; \alpha, \beta) = \frac{x^{\alpha-1}(1-x)^{\beta-1}}{B(\alpha, \beta)}, \quad 0 < x < 1,
$$

burada 
- $ x $: Təsadüfi dəyişən,
- $ \alpha > 0 $: Forma parametri (əvvəlki müvəffəqiyyətlər),
- $ \beta > 0 $: Forma parametri (əvvəlki uğursuzluqlar),
- $ B(\alpha, \beta) $: Beta funksiyası.

### Beta Funksiyası
Beta funksiyası $ B(\alpha, \beta) $ belə tərif edilir:

$$
B(\alpha, \beta) = \int_0^1 t^{\alpha-1}(1-t)^{\beta-1} \, dt,
$$

və bu, $ \Gamma $-funksiyası ilə əlaqəlidir:
$$
B(\alpha, \beta) = \frac{\Gamma(\alpha)\Gamma(\beta)}{\Gamma(\alpha + \beta)},
$$

burada $ \Gamma(x) $ qamma funksiyasıdır.

### Beta Paylanmanın Xüsusiyyətləri
1. **Orta**:
   $$
   \mu = \frac{\alpha}{\alpha + \beta}
   $$
2. **Dispersiya**:
   $$
   \sigma^2 = \frac{\alpha \beta}{(\alpha + \beta)^2 (\alpha + \beta + 1)}
   $$
3. **Dəyişkənlik Aralığı**: Beta paylanması həmişə [0,1] intervalında dəyərlər alır.

### Beta Paylanmanın Forması
Beta paylanmasının forması $ \alpha $ və $ \beta $-nın qiymətlərindən asılıdır:
- $ \alpha = \beta = 1 $: Uniform paylanma.
- $ \alpha > 1, \beta > 1 $: Belə vəziyyətdə paylanma simmetrik və ortada zirvəli olur.
- $ \alpha < 1, \beta < 1 $: Paylanma hər iki ucda zirvəlidir.
- $ \alpha > \beta $: Paylanma sola meylli.
- $ \beta > \alpha $: Paylanma sağa meylli.

### Beta Paylanmanın Tətbiqləri
1. **Bayes statistikası**: Beta paylanma tez-tez bir Bernoulli təsadüfi dəyişən üçün ilkin paylanma kimi istifadə olunur.
2. **Proseslərin simulyasiyası**: Ehtimal dəyərlərini [0,1] intervalında modelləşdirmək üçün istifadə olunur.
3. **Beta-binomial paylanma**: Binomial paylanma ilə birləşdirilərək istifadə olunur.

### Misal
Tutaq ki, bir sınaqda əvvəlki uğurların sayı $ \alpha = 2 $, uğursuzluqların sayı isə $ \beta = 3 $-dür. Bu halda paylanmanın sıxlıq funksiyası belə olur:
$$
f(x; 2, 3) = \frac{x^{2-1}(1-x)^{3-1}}{B(2, 3)} = \frac{6x(1-x)^2}{1},
$$
burada $ B(2, 3) = \frac{\Gamma(2)\Gamma(3)}{\Gamma(5)} = \frac{1 \cdot 2}{120} = \frac{1}{6} $.

Bu funksiyanı inteqrasiya etməklə müxtəlif ehtimal bölgələrini hesablamaq olar.
