##  İkidəyişənli Funksiya,  İkidəyişənli funksiyanın limiti, Xüsusi törəmələr (Partial Derivatives), Tam diferensial (Total Differential) 

### 0. **İkidəyişənli Funksiya**  
#### Tərif:  
İkidəyişənli funksiya $ f(x, y) $, iki müstəqil dəyişən $ x $ və $ y $-yə görə tərtib edilmiş funksiyadır. Belə bir funksiya, təyin olunmuş sahənin hər bir nöqtəsinə bir $ z = f(x, y) $ qiymətini təyin edir.  
Əsasən $ z $-nin qiyməti $ x $ və $ y $-nin dəyişməsindən asılı olaraq dəyişir.

#### Nümunə:  
$$
f(x, y) = x^2 + y^2
$$
Bu funksiya müstəvidə dairəvi paraboloid formasını ifadə edir.

---

### 1. **İkidəyişənli funksiyanın limiti**  
#### Tərif:  
Bir ikidəyişənli funksiya $ f(x, y) $ üçün, $(x, y) \to (a, b)$ nöqtəsinə yaxınlaşdıqca $ f(x, y) $ ifadəsinin sabit bir $ L $ qiymətinə yaxınlaşırsa, bu halda limit mövcuddur və yazılır:  
$$
\lim_{(x, y) \to (a, b)} f(x, y) = L.
$$  
Funksiyanın limiti yalnız o halda mövcuddur ki, fərqli yollarla $(a, b)$ nöqtəsinə yaxınlaşdıqda eyni $ L $ nəticəsini verək.

#### Nümunə:  
$$
f(x, y) = \frac{xy}{x^2 + y^2}.
$$  
$(x, y) \to (0, 0)$ üçün:  
- $y = 0$ boyunca: $\frac{xy}{x^2 + y^2} = 0$.  
- $x = y$ boyunca: $\frac{xy}{x^2 + y^2} = \frac{x^2}{2x^2} = \frac{1}{2}$.  
Limit fərqli yollardan fərqli olduğu üçün limit mövcud deyil.

---

### 2. **Xüsusi törəmələr (Partial Derivatives)**  
#### Tərif:  
İkidəyişənli funksiya $ f(x, y) $ üçün xüsusi törəmə bir dəyişənin sabit tutulması ilə digər dəyişənə görə törəmə almağa deyilir:  
$$
\frac{\partial f}{\partial x} = \lim_{h \to 0} \frac{f(x+h, y) - f(x, y)}{h},
$$
$$
\frac{\partial f}{\partial y} = \lim_{h \to 0} \frac{f(x, y+h) - f(x, y)}{h}.
$$

#### Nümunə:  
$$
f(x, y) = x^2y + y^3.
$$  
Xüsusi törəmələr:  
$$
\frac{\partial f}{\partial x} = 2xy, \quad \frac{\partial f}{\partial y} = x^2 + 3y^2.
$$

---

### 3. **Tam diferensial (Total Differential)**  
#### Tərif:  
Bir ikidəyişənli funksiya $ f(x, y) $ üçün tam diferensial aşağıdakı kimi hesablanır:  
$$
df = \frac{\partial f}{\partial x} dx + \frac{\partial f}{\partial y} dy.
$$  
Bu, $ f(x, y) $-nin kiçik dəyişməsinin $ x $ və $ y $-yə görə xüsusi törəmələrin dəyişimi ilə necə ifadə edildiyini göstərir.

#### Nümunə:  
$$
f(x, y) = x^2y + y^3.
$$  
Xüsusi törəmələrdən istifadə edərək tam diferensial:  
$$
df = (2xy)dx + (x^2 + 3y^2)dy.
$$

---

### Xülasə:
- **İkidəyişənli funksiyanın limiti**: Fərqli yollarla $(a, b)$-yə yaxınlaşdıqda limit mövcud olub-olmadığını yoxlamaq vacibdir.  
- **Xüsusi törəmələr**: Funksiyanın bir dəyişənə görə təsirini ölçür.  
- **Tam diferensial**: Funksiyanın hər iki dəyişənə görə ümumi dəyişimini ifadə edir.

## İkidəyişənli Funksiya, Təyin Oblastı ($ D $), Qiymətlər Oblastı ($ E $), Həndəsi Təsviri, Səviyyə Xətti (Level Curve)


### **İkidəyişənli Funksiya və Əsas Məfhumlar**

#### 1. **Təyin Oblastı (\( D \))**  
İkidəyişənli funksiya \$ f(x, y) \$ üçün təyin oblastı \$ D \$, \$(x, y)\$-lərin qəbul edə biləcəyi bütün mümkün qiymətlərin çoxluğudur.  
Başqa sözlə, \$ D \$ elə bir sahədir ki, bu sahədə \$ x \$ və \$ y \$-yə uyğun olan \$ f(x, y) \$ funksiyası müəyyən olunur.

#### Misal:  
Funksiya \$ f(x, y) = \sqrt{x^2 + y^2 - 1} \$ üçün:  
- Təyin oblastı \$ x^2 + y^2 \geq 1 \$ şərtini ödəyən nöqtələrdən ibarətdir.

---

#### 2. **Təyin Sahəsi (\( E \))**  
Təyin sahəsi \$ E \$, \$ z = f(x, y) \$ üçün alınan nəticələrin çoxluğudur. Başqa sözlə, \$ f(x, y) \$-nin bütün mümkün qiymətləri təyin sahəsini təşkil edir.  

#### Misal:  
Funksiya \$ f(x, y) = x^2 + y^2 \$ üçün:  
- Təyin oblastı \$ D \$: Bütün \$(x, y) \in \mathbb{R}^2 \$ (tam müstəvi).  
- Təyin sahəsi \$ E \$: \$ z \geq 0 \$, çünki kvadrat ifadələr mənfi ola bilmir.

---

#### 3. **Həndəsi Təsviri**  
İkidəyişənli funksiya \$ f(x, y) \$ üç ölçülü fəzada (3D müstəvidə) bir səth ilə təsvir olunur.  
- \$ x \$- və \$ y \$-lər müstəvinin iki üfüqi koordinatını,  
- \$ z = f(x, y) \$ isə hündürlüyü göstərir.

#### Misal:  
Funksiya \$ f(x, y) = x^2 + y^2 \$ üçün səth, müsbət \$ z \$-lərdə bir paraboloid təşkil edir.

---

#### 4. **Səviyyə Xətti (Level Curve)**  
Səviyyə xətləri \$ z \$-nin sabit olduğu nöqtələrdən ibarət əyrilərdir. Başqa sözlə, səviyyə xətləri \$ f(x, y) = k \$ bərabərliyini ödəyən nöqtələrin çoxluğudur.  
Səviyyə xətləri, funksiyanın sabit dəyərlərini ikidəyişənli müstəvidə təsvir edir.

#### Misal:  
Funksiya \$ f(x, y) = x^2 + y^2 \$ üçün səviyyə xətləri:  
\$ x^2 + y^2 = k. \$  
Bu, radiusu \$ \sqrt{k} \$ olan dairələrdir.

---

### **Xülasə:**  
- **Təyin Oblastı (\$ D \$)**: Funksiyanın tərtib olunduğu sahə.  
- **Təyin Sahəsi (\$ E \$)**: Funksiyanın mümkün nəticələrinin çoxluğu.  
- **Həndəsi Təsvir**: Üç ölçülü səthin təsviri.  
- **Səviyyə Xətti**: Sabit \$ z \$-yə görə müstəvidə alınan əyrilər.

## Funksiya $ f(x, y) = \sqrt{1 - x^2 - y^2} $ üçün səviyyə xətləri $ f(x, y) = c$

Funksiya $ f(x, y) = \sqrt{1 - x^2 - y^2} $ üçün səviyyə xətləri $ f(x, y) = c $ şəklində təsvir olunur. Bu bərabərliyi aşağıdakı kimi yazırıq:

$$
f(x, y) = \sqrt{1 - x^2 - y^2} = c.
$$

Hər iki tərəfin kvadratını alırıq ki, kökdən qurtulaq:

$$
1 - x^2 - y^2 = c^2.
$$

Daha sonra bu ifadəni sadələşdiririk:

$$
x^2 + y^2 = 1 - c^2.
$$

### Nəticə:
- Bu, radiusu $ \sqrt{1 - c^2} $ olan bir dairədir.
- Dairə yalnız $ c^2 \leq 1 $ şərtini ödəyən qiymətlər üçün mövcuddur, çünki $ 1 - c^2 $ ifadəsi mənfi ola bilməz. Başqa sözlə, \$ c \in [-1, 1] $.

### Həndəsi Təsvir:
- Funksiya səthi $ x^2 + y^2 + z^2 = 1 $ şəklində olan bir sfera hissəsidir (sferanın üst yarısı).
- Hər bir səviyyə xətti bu sfera üzərində sabit $ z = c $ hündürlüyünə uyğun olan dairəvi bir kəsiyə uyğundur.

## İkidəyişənli Funksiyanın Limiti(1)

>**Tərif:** İkidəyişənli funksiyanın limiti, dəyişənlər $x$ və $y$ verilən nöqtəyə yaxınlaşdıqca funksiyanın qiymətinin sabit bir $L\$ ədədinə yaxınlaşmasını ifadə edir.  

**Düstur:**  
$$
\lim_{(x, y) \to (a, b)} f(x, y) = L
$$ 
Burada:  
- $(x, y) \to (a, b)$: $(x, y)$ nöqtəsi $(a, b)$-yə yaxınlaşır,  
- $f(x, y)$: funksiya,  
- $L$: yaxınlaşdığı sabit qiymət.


İkidəyişənli funksiyanın limiti, iki dəyişəni olan funksiyanın müəyyən bir nöqtəyə yaxınlaşarkən dəyərinin nə cür bir qiymətə yaxınlaşdığını təyin edir. Bu anlayış daha çox çox dəyişənli analizdə tətbiq edilir və funksiyanın bir nöqtəyə yaxınlaşarkən nə cür davrandığını başa düşməyə kömək edir.

### İkidəyişənli Funksiyanın Limiti

Bir funksiyanı $f(x, y)$ kimi düşünək, burada $x$ və $y$ dəyişənləridir. Əgər bu funksiyanın $(a, b)$ nöqtəsinə yaxınlaşarkən bir limit mövcuddursa, onda:

$$
\lim_{(x, y) \to (a, b)} f(x, y) = L
$$

Bu, o deməkdir ki, $f(x, y)$ funksiyasının dəyəri $(a, b)$ nöqtəsinə yaxınlaşan zaman $L$-yə yaxınlaşır.

### İkidəyişənli Funksiyanın Limitinin Mövcud Olması

İkidəyişənli funksiyanın limitinin mövcud olub-olmaması, müxtəlif yollardan (həm düz, həm də qeyri-düz) yaxınlaşma nəticələrinə əsaslanır. Əgər bütün bu yollarla limitlər eyni olarsa, o zaman limit mövcuddur. Əks halda, limit mövcud deyil.

### Limitin Mövcudluğunu Yoxlama Yolları

1. **Düz Yollarla Yaxınlaşma:**
   - $x = a$ və $y \to b$
   - $y = b$ və $x \to a$

2. **Qeyri-Düz Yollarla Yaxınlaşma:**
   - Məsələn, $y = x$, $y = x^2$ kimi yollarla.

Əgər bu yollarla yaxınlaşma nəticələri eyni dəyəri verirsə, o zaman limit mövcuddur.

### Misal:
Funksiya:  
$$
f(x, y) = \frac{x^2 - y^2}{x^2 + y^2}
$$

Bu funksiyanın $(0, 0)$ nöqtəsinə yaxınlaşarkən limitini müxtəlif yollarla araşdıraraq qərar verək:

---

### **$y = x$ yolu ilə yaxınlaşma:**

1. $y = x$ olduğunu funksiyada əvəz edək:
   $$
   f(x, y) = f(x, x) = \frac{x^2 - x^2}{x^2 + x^2} = \frac{0}{2x^2} = 0
   $$

Beləliklə, $y = x$ yolu ilə yaxınlaşanda limit $0$-dır.

---

### **$y = -x$ yolu ilə yaxınlaşma:**

1. $y = -x$ olduğunu funksiyada əvəz edək:
   $$
   f(x, y) = f(x, -x) = \frac{x^2 - (-x)^2}{x^2 + (-x)^2} = \frac{x^2 - x^2}{x^2 + x^2} = \frac{0}{2x^2} = 0
   $$

Buradan görünür ki, $y = -x$ yolu ilə yaxınlaşanda da limit $0$-dır.

---

### **$y = x^2$ yolu ilə yaxınlaşma:**

1. $y = x^2$ olduğunu funksiyada əvəz edək:
   $$
   f(x, y) = f(x, x^2) = \frac{x^2 - (x^2)^2}{x^2 + (x^2)^2}
   $$

2. Sadələşdirək:
   $$
   f(x, x^2) = \frac{x^2 - x^4}{x^2 + x^4}
   $$

3. Payı və məxrəci $x^2$-yə bölək:
   $$
   f(x, x^2) = \frac{1 - x^2}{1 + x^2}
   $$

4. $x \to 0$ olduqda:
   $$
   f(x, x^2) \to \frac{1 - 0}{1 + 0} = 1
   $$

Beləliklə, $y = x^2$ yolu ilə yaxınlaşanda limit $1$-dir.

---

### **Nəticə:**

Fərqli yollarla yaxınlaşanda fərqli limitlər əldə edirik:  
- $y = x$ və $y = -x$ yollarında limit $0$-dır.  
- $y = x^2$ yolunda isə limit $1$-dir.

Bu o deməkdir ki, $f(x, y)$ funksiyasının $(0, 0)$ nöqtəsində limiti **mövcud deyil**.

## İkidəyişənli Funksiyanın Kəsilməzliyi (1)

### İkidəyişənli Funksiyanın Kəsilməzliyi

>**Tərif:** İkidəyişənli funksiya $(x, y)$ nöqtəsində kəsilməzdir, əgər həmin nöqtədə limit mövcuddur və funksiyanın limiti onun həmin nöqtədəki qiymətinə bərabərdir.  

**Düstur:**  
$$
\lim_{(x, y) \to (a, b)} f(x, y) = f(a, b)
$$  
Burada:  
- $f(a, b)$: funksiyanın $(a, b)$ nöqtəsindəki qiymətidir.  
- $(x, y) \to (a, b)$: $(x, y)$ nöqtəsi $(a, b)$-yə yaxınlaşır.


Bir funksiyanın kəsilməz olması, onun müəyyən bir nöqtədə və bu nöqtənin ətrafında kəsilməz şəkildə dəyərlər almasını təmin edir. İkidəyişənli bir funksiyanın kəsilməzliyi anlayışı, birdəyişənli funksiyalardakı kəsilməzlik anlayışının çox dəyişənlərə genişləndirilməsidir.

---

### **Tərif**

Bir $f(x, y)$ funksiyası, $(a, b)$ nöqtəsində kəsilməzdir, əgər aşağıdakı üç şərt ödənilirsə:

1. $f(a, b)$-nin dəyəri mövcuddur, yəni $f(a, b)$ təyin olunmuşdur.
2. $\lim_{(x, y) \to (a, b)} f(x, y)$ mövcuddur, yəni limit mövcuddur.
3. 
   $$
   \lim_{(x, y) \to (a, b)} f(x, y) = f(a, b).
   $$

Bu, funksiyanın $(a, b)$-də kəsilməz olması üçün funksiyanın limit dəyəri ilə həmin nöqtədəki dəyərinin bərabər olması deməkdir.

---

### **Ümumi Yanaşma**

Bir $f(x, y)$ funksiyası müəyyən bir bölgədə kəsilməzdirsə, bu bölgənin hər bir nöqtəsində yuxarıdakı şərtlər ödənir. Kəsilməzlik anlayışı, funksiyanın tərtib etdiyi səthin hamar olduğunu və müəyyən bir nöqtədə sıçrayış (discontinuity) olmadığını göstərir.

---

### **Kəsilməz Funksiyalara Misallar**

1. **Polinom Funksiyalar:**  
   Hər hansı bir polinom funksiyası (məsələn, $f(x, y) = x^2 + y^2 + 3xy + 2$) bütün $\mathbb{R}^2$-də kəsilməzdir.

2. **Trigonometrik Funksiyalar:**  
   Trigonometrik funksiyalar (məsələn, $f(x, y) = \sin(x) + \cos(y)$) təyin olunduğu bölgədə kəsilməzdir.

3. **Rasional Funksiyalar:**  
   $$
   f(x, y) = \frac{x^2 + y^2}{x^2 + y^2 + 1}
   $$
   Bu funksiya təyin olunduğu bütün nöqtələrdə kəsilməzdir, çünki məxrəc heç vaxt 0 olmur.

---

### **Kəsilməz Olmayan Funksiyalara Misal**

Funksiya:  
$$
f(x, y) =
\begin{cases} 
\frac{x^2 - y^2}{x^2 + y^2}, & \text{əgər } (x, y) \neq (0, 0) \\ 
1, & \text{əgər } (x, y) = (0, 0)
\end{cases}
$$

Bu funksiyanın $(0, 0)$-da kəsilməz olub-olmadığını yoxlayaq:

1. $f(0, 0) = 1$, yəni dəyər mövcuddur.
2. $(x, y) \to (0, 0)$ olduqda limiti araşdıraq:
   - $y = x$ yolunda:
     $$
     f(x, y) = \frac{x^2 - x^2}{x^2 + x^2} = 0
     $$
   - $y = -x$ yolunda:
     $$
     f(x, y) = \frac{x^2 - (-x)^2}{x^2 + (-x)^2} = 0
     $$

   Limiti müxtəlif yollarla hesablasaq, $0$-a yaxınlaşır, amma $f(0, 0) = 1$. Bu səbəbdən $\lim_{(x, y) \to (0, 0)} f(x, y) \neq f(0, 0)$.

Bu funksiyanın $(0, 0)$-da kəsilməz olmadığını görürük.

---

### **Misal 1: Kəsilməzlik Analizi**

Funksiya:  
$$
f(x, y) = x^2 + y^2
$$

1. $f(a, b) = a^2 + b^2$, yəni dəyər mövcuddur.
2. $\lim_{(x, y) \to (a, b)} f(x, y) = \lim_{(x, y) \to (a, b)} (x^2 + y^2) = a^2 + b^2$, yəni limit mövcuddur.
3. 
   $$
   \lim_{(x, y) \to (a, b)} f(x, y) = f(a, b)
   $$

Bu funksiya hər bir nöqtədə kəsilməzdir.

---

### **Misal 2: Məxrəc Sıfır Olarsa**

Funksiya:  
$$
f(x, y) = \frac{x + y}{x^2 + y^2}
$$

Bu funksiya $(0, 0)$-da təyin olunmur, çünki məxrəc $0$ olur. Bu səbəbdən $f(x, y)$ həmin nöqtədə kəsilməzdir. Təyin olunduğu digər nöqtələrdə isə kəsilməzdir.

## İkidəyişənli funksiyanın xüsusi törəmələri (1)

> **Tərif:** İki dəyişənli funksiyanın qismən törəməsi, bir dəyişən sabit saxlanaraq digər dəyişənə görə funksiyanın dəyişmə sürətini ölçür.  

**Düstur:**  
- $ x $-ə görə qismən törəmə:  
$$
\frac{\partial f}{\partial x} = \lim_{\Delta x \to 0} \frac{f(x + \Delta x, y) - f(x, y)}{\Delta x}
$$  
- $ y $-ə görə qismən törəmə:  
$$
\frac{\partial f}{\partial y} = \lim_{\Delta y \to 0} \frac{f(x, y + \Delta y) - f(x, y)}{\Delta y}
$$



İkidəyişənli funksiyanın **xüsusi törəmələri** çox dəyişənli funksiyanın analitik təhlilində vacib rol oynayır. Xüsusi törəmələr, bir dəyişənin sabit saxlanıldığı halda funksiyanın digər dəyişən üzrə necə dəyişdiyini təyin edir. Aşağıda bu anlayışın təfərrüatları və misallar verilmişdir.

---

### Xüsusi Törəmənin Anlayışı
Bir $z = f(x, y)$ funksiyasını nəzərdən keçirək, burada $x$ və $y$ müstəqil dəyişənlərdir.

- **$X$ dəyişəni üzrə xüsusi törəmə:**  
  $y$ sabit qəbul edilir və $f(x, y)$-nin $x$ üzrə törəməsi belə təyin olunur:  
  $$
  \frac{\partial f}{\partial x} = \lim_{\Delta x \to 0} \frac{f(x + \Delta x, y) - f(x, y)}{\Delta x}
  $$

- **$Y$ dəyişəni üzrə xüsusi törəmə:**  
  $x$ sabit qəbul edilir və $f(x, y)$-nin $y$ üzrə törəməsi belə təyin olunur:  
  $$
  \frac{\partial f}{\partial y} = \lim_{\Delta y \to 0} \frac{f(x, y + \Delta y) - f(x, y)}{\Delta y}
  $$

Bu xüsusi törəmələr, çox dəyişənli funksiyaların lokal dəyişkənliyini qiymətləndirmək üçün istifadə olunur.

---

### Xüsusi Törəmələrin Riyazi Qaydaları
Xüsusi törəmələr bir dəyişən üzrə törəmə qaydalarına əsaslanır. Misal olaraq:

1. **Sabitlərin törəməsi:**  
   $$
   \frac{\partial c}{\partial x} = 0
   $$  
   burada $c$ sabitdir.

2. **Cəmin törəməsi:**  
   $$
   \frac{\partial}{\partial x} [f(x, y) + g(x, y)] = \frac{\partial f}{\partial x} + \frac{\partial g}{\partial x}
   $$

3. **Çoxluğun törəməsi:**  
   $$
   \frac{\partial}{\partial x} [f(x, y) \cdot g(x, y)] = f_x \cdot g + f \cdot g_x
   $$

4. **Kvotientin törəməsi:**  
   $$
   \frac{\partial}{\partial x} \left[ \frac{f(x, y)}{g(x, y)} \right] = \frac{f_x \cdot g - f \cdot g_x}{g^2}
   $$

---

### Misallar

#### Misal 1: Xüsusi törəmələr
Funksiyanı nəzərdən keçirək:  
$$
f(x, y) = x^2y + 3xy^2
$$

- **$x$ üzrə xüsusi törəmə:**  
  $$
  \frac{\partial f}{\partial x} = \frac{\partial}{\partial x} \left( x^2y + 3xy^2 \right) = 2xy + 3y^2
  $$

- **$y$ üzrə xüsusi törəmə:**  
  $$
  \frac{\partial f}{\partial y} = \frac{\partial}{\partial y} \left( x^2y + 3xy^2 \right) = x^2 + 6xy
  $$

---

#### Misal 2: Sabit saxlanılma
Verilən funksiyanı götürək:  
$$
f(x, y) = e^{x^2 + y^2}
$$

- **$x$ üzrə xüsusi törəmə:**  
  $$
  \frac{\partial f}{\partial x} = \frac{\partial}{\partial x} \left( e^{x^2 + y^2} \right) = 2xe^{x^2 + y^2}
  $$

- **$y$ üzrə xüsusi törəmə:**  
  $$
  \frac{\partial f}{\partial y} = \frac{\partial}{\partial y} \left( e^{x^2 + y^2} \right) = 2ye^{x^2 + y^2}
  $$

---

#### Misal 3: Daha mürəkkəb funksiya
$$
f(x, y) = \ln(x^2 + y^2)
$$

- **$x$ üzrə xüsusi törəmə:**  
  $$
  \frac{\partial f}{\partial x} = \frac{\partial}{\partial x} \ln(x^2 + y^2) = \frac{2x}{x^2 + y^2}
  $$

- **$y$ üzrə xüsusi törəmə:**  
  $$
  \frac{\partial f}{\partial y} = \frac{\partial}{\partial y} \ln(x^2 + y^2) = \frac{2y}{x^2 + y^2}
  $$

1. **Gradient:** Xüsusi törəmələrdən gradient hesablanmasında istifadə edilir:  
   $$
   \nabla f = \left( \frac{\partial f}{\partial x}, \frac{\partial f}{\partial y} \right)
   $$

2. **Tangent müstəvi:** Xüsusi törəmələr, çox dəyişənli funksiyanın tangent müstəvisini qurmaq üçün vacibdir.

## İkidəyişənli funksiyanın **tam diferensialı** (1)

>**Tərif:** İkidəyişənli funksiyanın tam diferensialı, funksiya daxilindəki kiçik dəyişiklikləri hər iki dəyişən üzrə qismən törəmələr əsasında ifadə edir.  

**Düstur:**  
$$
dz = \frac{\partial f}{\partial x} dx + \frac{\partial f}{\partial y} dy
$$  
Burada:  
- $\frac{\partial f}{\partial x}$ və $\frac{\partial f}{\partial y}$ funksiyanın $x$ və $y$-yə görə qismən törəmələridir,  
- $dx$ və $dy$ isə $x$ və $y$-dəki kiçik dəyişikliklərdir.




İkidəyişənli funksiyanın **tam diferensialı** çox dəyişənli funksiyanın təhlilində vacib bir anlayışdır. Tam diferensial, funksiyanın bir nöqtədəki kiçik dəyişikliklərini və bu dəyişikliklərin nəticələrini təyin etməyə imkan verir. Bu, analitik funksiyaların təkcə törəmələrini deyil, həm də həmin funksiyaların ümumi dəyişikliklərini anlamağa kömək edir.

---

### İkidəyişənli Funksiyanın Tam Diferensialı

Bir funksiyanı $ z = f(x, y) $ şəklində düşünək. Burada $ x $ və $ y $ müstəqil dəyişənlərdir.

Funksiyanın **tam diferensialı** aşağıdakı kimi təyin olunur:
$$
dz = \frac{\partial f}{\partial x} dx + \frac{\partial f}{\partial y} dy
$$

Burada:

- $ \frac{\partial f}{\partial x} $ — funksiyanın $ x $-ə görə xüsusi törəməsi,
- $ \frac{\partial f}{\partial y} $ — funksiyanın $ y $-yə görə xüsusi törəməsi,
- $ dx $ və $ dy $ — müvafiq olaraq $ x $-in və $ y $-nin kiçik dəyişiklikləridir.

Tam diferensial, $ x $ və $ y $ dəyişənləri üzrə funksiyanın kiçik dəyişikliklərinin cəmindən ibarətdir.

---

### Tam Diferensialın Anlamı

Tam diferensial bir çox dəyişənli funksiyanın **yerindəki** kiçik dəyişiklikləri təsvir edir. Məsələn, bir nöqtədə $ x $ və $ y $-nin kiçik dəyişikliklərinin funksiyadakı ümumi dəyişikliklə necə əlaqəli olduğunu göstərir.

---

### Misal 1: Funksiya $ f(x, y) = x^2 + y^2 $

Bu funksiyanın tam diferensialını hesablayaq.

1. Funksiyanın $ x $-ə görə törəməsi:
$$
\frac{\partial f}{\partial x} = 2x
$$
   
2. Funksiyanın $ y $-yə görə törəməsi:
$$
\frac{\partial f}{\partial y} = 2y
$$

Beləliklə, tam diferensial:
$$
dz = 2x \, dx + 2y \, dy
$$

---

### Misal 2: Funksiya $ f(x, y) = e^{x + y} $

Bu funksiyanın tam diferensialını hesablayaq.

1. Funksiyanın $ x $-ə görə törəməsi:
$$
\frac{\partial f}{\partial x} = e^{x + y}
$$
   
2. Funksiyanın $ y $-yə görə törəməsi:
$$
\frac{\partial f}{\partial y} = e^{x + y}
$$

Beləliklə, tam diferensial:
$$
dz = e^{x + y} \, dx + e^{x + y} \, dy
$$

---

### Misal 3: Funksiya $ f(x, y) = \ln(x^2 + y^2) $

Bu funksiyanın tam diferensialını hesablayaq.

1. Funksiyanın $ x $-ə görə törəməsi:
$$
\frac{\partial f}{\partial x} = \frac{2x}{x^2 + y^2}
$$

2. Funksiyanın $ y $-yə görə törəməsi:
$$
\frac{\partial f}{\partial y} = \frac{2y}{x^2 + y^2}
$$

Beləliklə, tam diferensial:
$$
dz = \frac{2x}{x^2 + y^2} \, dx + \frac{2y}{x^2 + y^2} \, dy
$$

---

### Tam Diferensialın Tətbiqləri

1. **Kiçik dəyişikliklərin hesablama:** Tam diferensial, funksiyanın kiçik dəyişikliklərinin təsirini qiymətləndirmək üçün istifadə olunur. Məsələn, bir məsələdə $ x $ və $ y $ müstəqil dəyişənlərinin kiçik dəyişiklikləri nəticəsində $ f(x, y) $-nin nə qədər dəyişdiyini hesablaya bilərik.

2. **Lineyer yaxınlaşma:** Tam diferensial, çox dəyişənli funksiyaların yaxınlaşma tətbiqlərində istifadə edilir. Məsələn, bir nöqtədəki kiçik dəyişikliklərə əsaslanaraq funksiyanın yaxınlaşmış qiymətini tapmaq üçün istifadə oluna bilər.

3. **Error analizi:** Mühəndislik və statistik analizlərdə tam diferensial, ölçü xətalarını qiymətləndirmək üçün geniş istifadə olunur.

4. **Optimallaşdırma:** Tam diferensial optimallaşdırma problemlərində funksiyanın lokal dəyişikliklərinin təhlilində istifadə edilir.

---

### Nəticə
