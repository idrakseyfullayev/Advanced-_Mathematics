## Qeyri-Aşkar Funksiyanın Törəməsi(1)

### Qeyri-Aşkar Funksiyanın Törəməsi

Qeyri-aşkar funksiya, $y$-in $x$-ə görə açıq formada ifadə edilmədiyi, lakin iki dəyişənin (məsələn, $x$ və $y$) arasında əlaqə verildiyi funksiyadır. Məsələn, $F(x, y) = 0$ şəklində verilən münasibətlər qeyri-aşkar funksiyalardır.

**Törəmə prosesi** qeyri-aşkar törəmə metodu ilə aparılır və adətən $y$-ni $x$-ə görə diferensiallamaq üçün zəncir qaydasından istifadə edilir. 

---

#### Tərif
Qeyri-aşkar funksiya $F(x, y) = 0$ şəklində verilmişdir. Bu münasibətdə $x$ və $y$ arasında əlaqə olduğu üçün hər iki tərəfi $x$-ə görə diferensiallayırıq:

$$
\frac{\partial F}{\partial x} + \frac{\partial F}{\partial y} \cdot \frac{dy}{dx} = 0
$$

Buradan $\frac{dy}{dx}$-i tapmaq üçün həll edirik:

$$
\frac{dy}{dx} = -\frac{\frac{\partial F}{\partial x}}{\frac{\partial F}{\partial y}}
$$

---

#### Misallar

1. **Misal 1: Sadə kvadrat əlaqə**
   $$
   x^2 + y^2 = 25
   $$

   **Həll:**
   Hər iki tərəfi $x$-ə görə diferensiallayırıq:
   $$
   2x + 2y \cdot \frac{dy}{dx} = 0
   $$

   Buradan:
   $$
   \frac{dy}{dx} = -\frac{x}{y}
   $$

---

2. **Misal 2: Daha mürəkkəb funksiya**
   $$
   e^x + xy + \sin(y) = 0
   $$

   **Həll:**
   Hər iki tərəfi $x$-ə görə diferensiallayırıq:
   $$
   \frac{d}{dx}(e^x) + \frac{d}{dx}(xy) + \frac{d}{dx}(\sin(y)) = 0
   $$

   Hesablayırıq:
   $$
   e^x + \left(1 \cdot y + x \cdot \frac{dy}{dx}\right) + \cos(y) \cdot \frac{dy}{dx} = 0
   $$

   Törəməni $\frac{dy}{dx}$-yə görə ayırırıq:
   $$
   e^x + y + x \cdot \frac{dy}{dx} + \cos(y) \cdot \frac{dy}{dx} = 0
   $$

   $$
   \frac{dy}{dx} \cdot (x + \cos(y)) = - (e^x + y)
   $$

   $$
   \frac{dy}{dx} = \frac{-(e^x + y)}{x + \cos(y)}
   $$

---

#### Tətbiqi və Analizi
- **Tətbiq sahələri:** Fizikada (məsələn, orbitlərin hesablanması), iqtisadiyyatda (məsələn, optimal qiymətlərin tapılması), mühəndislikdə və digər sahələrdə geniş istifadə olunur.
- **Üstünlük:** Funksiyanın açıq formada ifadə olunmasına ehtiyac olmadan törəmə hesablanır.

## İstiqamət üzrə Törəmə(1)



**İstiqamət üzrə törəmə** (directional derivative), çox dəyişənli funksiyalarla işləyərkən, funksiyanın müəyyən bir istiqamətdə necə dəyişdiyini təyin edən bir anlayışdır. Bu, funksiyanın bir nöqtəsindəki dəyişməsinin yalnız bir istiqamətdəki sürətini göstərir.

### İstiqamət üzrə Törəmə Tərifi:
Bir funksiyanın $f(x, y, z, \dots)$ çox dəyişənli funksiyası və bir vektor $\mathbf{v} = (v_1, v_2, \dots)$ veriləndə, istiqamət üzrə törəmə funksiyanın həmin nöqtədə müəyyən istiqamətdə necə dəyişdiyini göstərir. Bu törəmə, **gradian vektoru** ilə istiqamət vektorunun daxili çarpmasının nəticəsidir.

Formul olaraq, $\mathbf{v}$ istiqamətindəki törəmə belə verilir:

$$
D_{\mathbf{v}} f(\mathbf{x}) = \nabla f(\mathbf{x}) \cdot \hat{\mathbf{v}}
$$

Burada:
- $\nabla f(\mathbf{x})$ - funksiyanın gradianı (hər bir dəyişənə görə qismən törəmələrin vektoru),
- $\hat{\mathbf{v}}$ - istiqamət vektorunun normallaşdırılmış (unit) versiyası,
- $\cdot$ - daxili çarpma (dot product).

### İstiqamət üzrə Törəmənin Anlayışı:
- **Gradian** ($\nabla f$) funksiyanın bütün dəyişənlərinə görə qismən törəmələrin vektorudur və funksiyanın ən sürətli artım istiqamətini göstərir.
- **İstiqamət üzrə törəmə** bu gradianın verilən bir istiqamətdəki komponentini ölçür.

### Misal 1:
Təsəvvür edin ki, $f(x, y) = x^2 + y^2$ funksiyasını veririk və $\mathbf{x} = (1, 1)$ nöqtəsindəki istiqamət üzrə törəməni tapmaq istəyirik. İstiqamət vektoru $\mathbf{v} = (1, 1)$ olsun.

1. **Gradianı hesablayırıq:**

$$
\nabla f(x, y) = \left( \frac{\partial f}{\partial x}, \frac{\partial f}{\partial y} \right) = (2x, 2y)
$$

Bu, $(1, 1)$ nöqtəsində:

$$
\nabla f(1, 1) = (2 \times 1, 2 \times 1) = (2, 2)
$$

2. **İstiqamət vektorunun normallaşdırılması:**
$$
\hat{\mathbf{v}} = \frac{(1, 1)}{\sqrt{1^2 + 1^2}} = \frac{(1, 1)}{\sqrt{2}} = \left( \frac{1}{\sqrt{2}}, \frac{1}{\sqrt{2}} \right)
$$

3. **İstiqamət üzrə törəməni tapırıq:**

$$
D_{\mathbf{v}} f(1, 1) = (2, 2) \cdot \left( \frac{1}{\sqrt{2}}, \frac{1}{\sqrt{2}} \right) = 2 \times \frac{1}{\sqrt{2}} + 2 \times \frac{1}{\sqrt{2}} = \frac{4}{\sqrt{2}} = 2\sqrt{2}
$$

### Misal 2:
Funksiya $f(x, y) = x^3 - 3xy^2$ və istiqamət vektoru $\mathbf{v} = (1, 0)$.

1. **Gradianı hesablayırıq:**

$$
\nabla f(x, y) = \left( \frac{\partial f}{\partial x}, \frac{\partial f}{\partial y} \right) = (3x^2 - 3y^2, -6xy)
$$

2. **Verilən nöqtə $(1, 1)$ üçün gradianı tapırıq:**

$$
\nabla f(1, 1) = (3 \times 1^2 - 3 \times 1^2, -6 \times 1 \times 1) = (0, -6)
$$

3. **İstiqamət üzrə törəməni hesablayırıq:**

İstiqamət vektoru $\mathbf{v} = (1, 0)$ olduğuna görə, normallaşdırmaya ehtiyac yoxdur.

$$
D_{\mathbf{v}} f(1, 1) = (0, -6) \cdot (1, 0) = 0 \times 1 + (-6) \times 0 = 0
$$

### Nəticə:
İstiqamət üzrə törəmə funksiyanın verilən nöqtədə, müəyyən bir istiqamətdə necə dəyişdiyini ölçür. Bu, yalnız o istiqamətə görə funksiyanın sürətini göstərir və gradianın o istiqamətdəki komponentini göstərir.

## İstiqamət Vektorunun Normallaşdırılması


İstiqamət vektorunun **normallaşdırılması** (və ya **unit vektora çevrilməsi**), vektorun uzunluğunu (normasını) 1-ə endirmək məqsədini güdür. Bu proses, vektoru eyni istiqamətdə saxlayır, amma onun uzunluğunu 1-ə bərabər edir. İstiqamət vektorunun normallaşdırılması, adətən, **unit vektor** tapmaq adlanır.


### İstiqamət Vektorunun Normallaşdırılması:
Verilən vektor $ \mathbf{v} = (v_1, v_2, \dots, v_n) $ olsun. Bu vektorun **normallaşdırılması** üçün əvvəlcə onun uzunluğunu (yəni normasını) hesablamaq lazımdır.

#### Vektorun uzunluğunu tapmaq:
Bir vektor $ \mathbf{v} = (v_1, v_2, \dots, v_n) $ üçün **uzunluq** (norma) belə hesablanır:

$$
\|\mathbf{v}\| = \sqrt{v_1^2 + v_2^2 + \dots + v_n^2}
$$

Bu uzunluğu tapdıqdan sonra, vektorun **normallaşdırılmış** (unit) versiyasını almaq üçün vektoru onun uzunluğuna bölerik:

$$
\hat{\mathbf{v}} = \frac{\mathbf{v}}{\|\mathbf{v}\|}
$$

Burada:
- $ \mathbf{v} = (v_1, v_2, \dots, v_n) $ - verilən istiqamət vektoru,
- $ \|\mathbf{v}\| $ - vektorun uzunluğu (norması),
- $ \hat{\mathbf{v}} $ - normallaşdırılmış vektor (unit vektor).

### Misal:
Təsəvvür edin ki, $ \mathbf{v} = (3, 4) $ vektorunu normallaşdırmaq istəyirik.

1. **Vektorun uzunluğunu tapırıq:**

$$
\|\mathbf{v}\| = \sqrt{3^2 + 4^2} = \sqrt{9 + 16} = \sqrt{25} = 5
$$

2. **Normallaşdırılmış vektoru tapırıq:**

$$
\hat{\mathbf{v}} = \frac{\mathbf{v}}{\|\mathbf{v}\|} = \frac{(3, 4)}{5} = \left( \frac{3}{5}, \frac{4}{5} \right)
$$

Beləliklə, normallaşdırılmış vektor $ \hat{\mathbf{v}} = \left( \frac{3}{5}, \frac{4}{5} \right) $ olacaq.

### Nəticə:
İstiqamət vektorunun normallaşdırılması vektorun uzunluğunu 1-ə gətirən prosesdir. Bu, vektorun yalnız istiqamətini qoruyur, amma onun uzunluğunu dəyişdirir. Normallaşdırılmış vektor, **unit vektor** adlanır.

## Funksiyanın Gradienti(1)

### Funksiyanın Gradientinin Tərifi

>**Tərif**: Gradient, çoxdəyişənli bir funksiyanın hər bir dəyişənə görə qismən törəmələrinin vektor şəklində birləşməsidir. Bir funksiyanın gradienti, həmin funksiyanın dəyişmə sürətinin və istiqamətinin təsviridir. Gradient həm də funksiyanın ən böyük artım istiqamətini göstərir.

Əgər $f(x_1, x_2, ..., x_n)$ çoxdəyişənli bir skalyar funksiyadırsa, onun gradienti belə tərif olunur:

$$
\nabla f = \left[ \frac{\partial f}{\partial x_1}, \frac{\partial f}{\partial x_2}, ..., \frac{\partial f}{\partial x_n} \right]
$$

Burada:
- $\nabla f$: Funksiyanın gradienti.
- $\frac{\partial f}{\partial x_i}$: Funksiyanın $x_i$ dəyişəninə görə qismən törəməsi.

---

### Gradientin Anlayışı

1. **Geometrik mənası**: Gradient bir vektordur və funksiyanın ən böyük artım istiqamətini göstərir. Gradientin uzunluğu isə həmin istiqamətdəki artımın sürətini ifadə edir.
2. **Tətbiqləri**:
   - **Optimallaşdırma**: Gradientlər optimallaşdırma prosesində, xüsusilə Gradient Descent alqoritmində istifadə olunur.
   - **Fizika və mühəndislik**: Temperatur, potensial və digər sahələrin dəyişmə sürətini analiz etmək üçün.
   - **Maşın öyrənməsi**: Neyron şəbəkələrdə itki funksiyasının minimumunu tapmaq üçün istifadə olunur.

---

### Gradientin Formal Misalları

#### Misal 1: İki dəyişənli bir funksiya
Verilən funksiya:

$$
f(x, y) = x^2 + y^2
$$

Gradienti tapmaq üçün dəyişənlər üzrə qismən törəmələr alınır:

$$
\frac{\partial f}{\partial x} = 2x, \quad \frac{\partial f}{\partial y} = 2y
$$

Beləliklə:

$$
\nabla f(x, y) = [2x, \ 2y]
$$

#### Misal 2: Kompleks bir funksiya
Verilən funksiya:

$$
f(x, y) = x^2y + 3xy^2
$$

Gradienti tapmaq üçün dəyişənlər üzrə qismən törəmələr alınır:

$$
\frac{\partial f}{\partial x} = 2xy + 3y^2, \quad \frac{\partial f}{\partial y} = x^2 + 6xy
$$

Beləliklə:

$$
\nabla f(x, y) = [2xy + 3y^2, \ x^2 + 6xy]
$$

---

### Gradientin Xüsusiyyətləri
1. **Sıfır Gradient**: Gradient sıfır olan nöqtələrdə ($\nabla f = 0$) funksiya maksimum, minimum və ya saddle nöqtədə ola bilər.
2. **Uzunluq və İstiqamət**: Gradientin norması ($|\nabla f|$) dəyişmə sürətini, vektorun istiqaməti isə funksiyanın ən böyük artım istiqamətini göstərir.

---

### Fiziki Misal

Bir potensial sahəni təsvir edən funksiya:

$$
f(x, y, z) = x^2 + y^2 + z^2
$$

Gradient:

$$
\nabla f(x, y, z) = [2x, 2y, 2z]
$$

Bu, həmin sahənin hər nöqtəsində dəyişmə sürətinin və istiqamətinin vektor şəklində təsviridir. Bu misalda gradient, funksiyanın həndəsi mənada bir kürə səthində necə dəyişdiyini göstərir.

Əlavə suallarınız varsa və ya nümunələri artırmaq istəyirsinizsə, mənə bildirin! 😊

## Gradientin $ \mathbf{i} $, $ \mathbf{j} $, və digər vektorlarla ifadəsi


### Gradientin $ \mathbf{i} $, $ \mathbf{j} $, və digər vektorlarla ifadəsi

Gradient vektorunu standart bazis vektorları (\$ \mathbf{i}, \mathbf{j}, \mathbf{k} \$) ilə ifadə etmək mümkündür. Bu formada, gradient vektorunun hər bir komponenti funksiyanın müvafiq dəyişənə görə qismən törəməsi ilə çarpılır.

---

### İki Dəyişənli Funksiya Üçün

Verilmiş funksiya:

$$
f(x, y)
$$

Gradientin ifadəsi:

$$
\nabla f(x, y) = \frac{\partial f}{\partial x} \mathbf{i} + \frac{\partial f}{\partial y} \mathbf{j}
$$

Burada:
- \$ \mathbf{i} \$: \$ x \$-oxu boyunca vahid vektor.
- \$ \mathbf{j} \$: \$ y \$-oxu boyunca vahid vektor.

---

#### Misal:

Funksiya:

$$
f(x, y) = x^2 + y^2
$$

Qismən törəmələr:

$$
\frac{\partial f}{\partial x} = 2x, \quad \frac{\partial f}{\partial y} = 2y
$$

Gradient:

$$
\nabla f(x, y) = 2x \mathbf{i} + 2y \mathbf{j}
$$

Məsələn, nöqtədə \$ (1, 2) \$ gradient:

$$
\nabla f(1, 2) = 2 \cdot 1 \mathbf{i} + 2 \cdot 2 \mathbf{j} = 2\mathbf{i} + 4\mathbf{j}
$$

---

### Üç Dəyişənli Funksiya Üçün

Verilmiş funksiya:

$$
f(x, y, z)
$$

Gradientin ifadəsi:

$$
\nabla f(x, y, z) = \frac{\partial f}{\partial x} \mathbf{i} + \frac{\partial f}{\partial y} \mathbf{j} + \frac{\partial f}{\partial z} \mathbf{k}
$$

Burada:
- \$ \mathbf{k} \$: \$ z \$-oxu boyunca vahid vektor.

---

#### Misal:

Funksiya:

$$
f(x, y, z) = x^2 + y^2 + z^2
$$

Qismən törəmələr:

$$
\frac{\partial f}{\partial x} = 2x, \quad \frac{\partial f}{\partial y} = 2y, \quad \frac{\partial f}{\partial z} = 2z
$$

Gradient:

$$
\nabla f(x, y, z) = 2x \mathbf{i} + 2y \mathbf{j} + 2z \mathbf{k}
$$

Məsələn, nöqtədə \$ (1, 1, 1) \$ gradient:

$$
\nabla f(1, 1, 1) = 2 \cdot 1 \mathbf{i} + 2 \cdot 1 \mathbf{j} + 2 \cdot 1 \mathbf{k} = 2\mathbf{i} + 2\mathbf{j} + 2\mathbf{k}
$$

---

### Ümumi Yekun

Gradient vektoru standart bazis vektorları ilə yazıldıqda belə təsvir olunur:

$$
\nabla f = \sum_{i=1}^n \frac{\partial f}{\partial x_i} \mathbf{e}_i
$$

Burada:
- \$ \mathbf{e}_i \$: \$ x_i \$-dəyişəninə uyğun vahid vektor.
- \$ \frac{\partial f}{\partial x_i} \$: Funksiyanın \$ x_i \$-yə görə qismən törəməsi.

---

## **İstiqamət üzrə törəmə** və **Gradient** Fərqi 

**İstiqamət üzrə törəmə** və **gradient** əlaqəli olsa da, fərqli anlayışlardır. Bu iki anlayış arasında əsas fərqlər və əlaqələri izah edək:

---

### 1. **Gradient:**
Gradient çoxdəyişənli bir skalyar funksiyanın hər dəyişənə görə qismən törəmələrindən ibarət olan **vektordur**. Funksiyanın artma sürətini və istiqamətini göstərir.

$$
\nabla f(x_1, x_2, ..., x_n) = \begin{bmatrix} 
\frac{\partial f}{\partial x_1} \\
\frac{\partial f}{\partial x_2} \\
\vdots \\
\frac{\partial f}{\partial x_n}
\end{bmatrix}
$$

**Xüsusiyyətlər:**
- Gradient funksiyanın ən sürətli artdığı istiqaməti göstərir.
- Gradientin uzunluğu ($ |\nabla f| $) funksiyanın həmin istiqamətdəki dəyişmə sürətini təsvir edir.

#### Misal:
Funksiya:

$$
f(x, y) = x^2 + y^2
$$

Qismən törəmələr:

$$
\frac{\partial f}{\partial x} = 2x, \quad \frac{\partial f}{\partial y} = 2y
$$

Gradient:

$$
\nabla f(x, y) = 2x \mathbf{i} + 2y \mathbf{j}
$$

Məsələn, $ (x, y) = (1, 2) $ nöqtəsində:

$$
\nabla f(1, 2) = 2 \cdot 1 \mathbf{i} + 2 \cdot 2 \mathbf{j} = 2\mathbf{i} + 4\mathbf{j}
$$

---

### 2. **İstiqamət üzrə törəmə:**

İstiqamət üzrə törəmə ($ D_{\mathbf{u}} f $) bir funksiyanın müəyyən bir **istiqamət** boyunca necə dəyişdiyini göstərir. Burada istiqamət verilmiş bir vahid vektor ($ \mathbf{u} $) ilə təyin edilir.

**Tərifi:**

$$
D_{\mathbf{u}} f = \nabla f \cdot \mathbf{u} = |\nabla f| |\mathbf{u}| \cos \theta
$$

Burada:
- $ \nabla f $: Gradient vektoru,
- $ \mathbf{u} $: Verilmiş istiqamət üzrə vahid vektor ($ |\mathbf{u}| = 1 $),
- $ \cos \theta $: Gradient ilə istiqamət vektoru arasındakı bucaq.

#### Misal:
Əgər funksiyamız $ f(x, y) = x^2 + y^2 $ və istiqamət vektorumuz $ \mathbf{u} = \frac{1}{\sqrt{2}} \mathbf{i} + \frac{1}{\sqrt{2}} \mathbf{j} $ (45° bucaqla olan istiqamət) olsun.

Hesablama:

- Gradient: $ \nabla f(x, y) = 2x \mathbf{i} + 2y \mathbf{j} $
- Nöqtədə $ (x, y) = (1, 2) $:

$$
\nabla f(1, 2) = 2\mathbf{i} + 4\mathbf{j}
$$

İstiqamət üzrə törəmə:

$$
D_{\mathbf{u}} f = (2\mathbf{i} + 4\mathbf{j}) \cdot \left( \frac{1}{\sqrt{2}} \mathbf{i} + \frac{1}{\sqrt{2}} \mathbf{j} \right)
$$

Hesablama:

$$
D_{\mathbf{u}} f = 2 \cdot \frac{1}{\sqrt{2}} + 4 \cdot \frac{1}{\sqrt{2}} = \frac{2}{\sqrt{2}} + \frac{4}{\sqrt{2}} = \frac{6}{\sqrt{2}} = 3\sqrt{2}
$$

---

### Əlaqə: Gradient və İstiqamət üzrə Törəmə

1. **İstiqamət üzrə törəmə gradientdən asılıdır:**
   İstiqamət üzrə törəmə gradient ilə istiqamət vektorunun skalyar hasilidir.

   $$
   D_{\mathbf{u}} f = \nabla f \cdot \mathbf{u}
   $$

2. **Gradient ən böyük istiqamət üzrə törəməni verir:**
   Gradient vektorunun istiqamətindəki törəmə maksimum olur. Bu, gradient ilə istiqamət vektorunun paralel olduğu ($ \theta = 0^\circ $) haldır. Beləliklə:

   $$
   D_{\mathbf{u}} f \text{ maksimum olduqda } D_{\mathbf{u}} f = |\nabla f|
   $$

   **Misal:**
   Əgər \( \mathbf{u} \) gradientin istiqamətinə paralel olsa, onda:

   $$
   D_{\mathbf{u}} f = |\nabla f|
   $$

   Nöqtə \( (1, 2) \) üçün:

   $$
   D_{\mathbf{u}} f = |\nabla f(1, 2)| = \sqrt{2^2 + 4^2} = \sqrt{4 + 16} = \sqrt{20} = 2\sqrt{5}
   $$

3. **İstiqamət üzrə törəmə müxtəlif istiqamətləri nəzərə alır:**
   Fərqli istiqamətlər üçün $ \mathbf{u} $-ya görə dəyişir, amma gradient funksiyanın lokal xüsusiyyətini tam təsvir edir.

---

### Nəticə:

- Gradient **vektordur** və funksiyanın ən sürətli dəyişmə istiqamətini göstərir.
- İstiqamət üzrə törəmə isə gradientin müəyyən istiqamət boyunca "projeksiyası" kimi qəbul edilə bilər.

Əgər bir istiqamət üzrə maksimum dəyişmə sürətini hesablamaq istəyirsinizsə, gradient vektorunun uzunluğundan istifadə edirsiniz. Əks halda, spesifik bir istiqamət üzrə törəmə hesablamaq üçün $ D_{\mathbf{u}} f $ düsturundan istifadə olunur.
