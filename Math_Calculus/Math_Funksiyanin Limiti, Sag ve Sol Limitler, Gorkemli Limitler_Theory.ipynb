## Limit 

### **Limit haqqında geniş məlumat**

Riyaziyyatda **limit** anlayışı bir funksiya və ya ardıcıllığın müəyyən bir nöqtədə və ya sonsuzluqda yaxınlaşdığı dəyəri təsvir edir. Limit anlayışı **diferensial və inteqral hesabın** təməlini təşkil edir və bir çox riyazi, fiziki və mühəndislik tətbiqlərində istifadə olunur.

> **Limit**: Bir funksiya və ya ardıcıllığın dəyişəninin müəyyən bir nöqtəyə və ya sonsuzluğa yaxınlaşdığı zaman yaxınlaşdığı dəyəri təsvir edən anlayışdır.


## **1. Limitin Tərifi**

### a) **Funksiyanın limiti**  
Funksiyanın $\lim_{x \to a} f(x) = L$ olması bu deməkdir:  
- $\forall \epsilon > 0$ (kiçik bir müsbət ədəd),  
- $\exists \delta > 0$ (başqa bir müsbət ədəd) belə ki,  
- Əgər $0 < |x - a| < \delta$, o zaman $|f(x) - L| < \epsilon$.  

Bu o deməkdir ki, $x$ dəyəri $a$-ya yaxınlaşdıqca $f(x)$ funksiyası $L$-ə yaxınlaşır.

---

### b) **Ardıcıllığın limiti**  
Bir ardıcıllığın limiti belə təsvir olunur:  
$$ 
\lim_{n \to \infty} a_n = L \quad \Leftrightarrow \quad \forall \epsilon > 0, \, \exists N \in \mathbb{N} \text{ belə ki, } n > N \Rightarrow |a_n - L| < \epsilon. 
$$  

Burada ardıcıllıq ($a_n$) nə qədər irəli getsə, $L$-ə o qədər yaxınlaşır.

---

## **2. Limitin Təsviri**  
Limit konsepti vasitəsilə dəyişənlərin bir-birinə yaxınlaşması araşdırılır. Bu yaxınlaşma həm sonsuz ($x \to \infty$) zamanlarda, həm də kiçik dəyişikliklər zamanı ($x \to a$) tədqiq edilir.

Məsələn:  
- Funksiyanın davranışını müəyyən etmək üçün ($\lim_{x \to 0} \frac{\sin x}{x} = 1$),  
- Sonsuz kiçik və ya böyük dəyərlərə yaxınlaşmada ($\lim_{x \to \infty} \frac{1}{x} = 0$) istifadə edilir.

---

## **3. Limitin Növləri**

### a) **Sonsuzluqda limit ($x \to \infty$):**  
Əgər $x$ sonsuzluğa yaxınlaşdıqca $f(x)$ müəyyən bir dəyərə yaxınlaşırsa:  
$$ 
\lim_{x \to \infty} f(x) = L 
$$  
Məsələn:  
$$ 
\lim_{x \to \infty} \frac{1}{x} = 0 
$$  

---

### b) **Sonsuzluğa yaxınlaşma ($\lim_{x \to \infty} f(x) = \infty$):**  
Əgər $f(x)$ qeyri-məhdud şəkildə artırsa:  
$$ 
\lim_{x \to \infty} f(x) = \infty 
$$  
Məsələn:  
$$ 
\lim_{x \to \infty} x^2 = \infty 
$$  

---

### c) **Sol və sağ limitlər:**  
Əgər limit yalnız bir tərəfdən araşdırılırsa:  
- **Sol limit:** $\lim_{x \to a^-} f(x)$,  
- **Sağ limit:** $\lim_{x \to a^+} f(x)$.  

Əgər:  
$$ 
\lim_{x \to a^-} f(x) = \lim_{x \to a^+} f(x) = L, 
$$  
o zaman limit mövcuddur və $\lim_{x \to a} f(x) = L$.

---

## **4. Əsas Limit Xassələri**

1. **Cəmin limiti:**  
   $$ 
   \lim_{x \to a} [f(x) + g(x)] = \lim_{x \to a} f(x) + \lim_{x \to a} g(x) 
   $$  

2. **Hasilin limiti:**  
   $$ 
   \lim_{x \to a} [f(x) \cdot g(x)] = \lim_{x \to a} f(x) \cdot \lim_{x \to a} g(x) 
   $$  

3. **Nisbin limiti:**  
   $$ 
   \lim_{x \to a} \frac{f(x)}{g(x)} = \frac{\lim_{x \to a} f(x)}{\lim_{x \to a} g(x)}, \quad \text{əgər } \lim_{x \to a} g(x) \neq 0. 
   $$  

---

$$  

---

2. $ \lim_{x \to 0^+} \ln(x) = -\infty. $

## Limitin Praktiki mənası

**Funksiya: $ \frac{1}{x} $**

Bu funksiya üçün limitə baxırıq: 

$$
\lim_{x \to 0} \frac{1}{x}
$$

#### Nəticə:
- $ x $ müsbət tərəfdən 0-a yaxınlaşanda, $\frac{1}{x}$ müsbət sonsuzluğa yaxınlaşır.
- $ x $ mənfi tərəfdən 0-a yaxınlaşanda, $\frac{1}{x}$ mənfi sonsuzluğa yaxınlaşır.

---

### Praktik Mənası
Bu limitin praktik mənası ondan ibarətdir ki, $ x $-in 0-a yaxınlaşdığı yerlərdə, $ \frac{1}{x} $ funksiyası çox sürətlə böyüyür və idarəolunmaz dəyərlərə çatır.

#### Misal:
Tutaq ki, $ x $-i kiçik dəyişikliklərlə yaxınlaşdırırıq:
- $ x = 0.1 $ olduqda, $\frac{1}{x} = 10$.
- $ x = 0.01 $ olduqda, $\frac{1}{x} = 100$.
- $ x = 0.001 $ olduqda, $\frac{1}{x} = 1000$.

Hər dəfə $ x $-dəki kiçik azalma, $ \frac{1}{x} $-də çox böyük artıma səbəb olur.

---

### Tətbiqi Nümunə:
Bu tip davranışı real həyatda da görə bilərik. Misal üçün:

- **Təzyiq və sahə münasibəti:** Təsəvvür edin ki, bir ağırlıq çox kiçik bir sahəyə tətbiq olunur. Sahə azaldıqca təzyiq sürətlə artır və sonsuz dəyərlərə çatır.
  
  Məsələn, bir oxun ucunu düşünün. Sahə çox kiçik olduğu üçün təzyiq çox böyükdür və bu da oxun nüfuz etmə qabiliyyətini izah edir.

---

### Nəticə:
Limit $ \frac{1}{x} $-də $ x \to 0 $ funksiyanın çox sürətlə böyüməsini göstərir. Bu, sistemin sabit olmadığını və kiçik dəyişikliklərin çox böyük təsirlərə səbəb olduğunu izah edir.

## Limitin hesablanma qaydaları və hər qaydaya uyğun bir real nümunə misalı



### 1. **Cəmin limiti**

**Qayda:**  
Əgər $ \lim_{x \to a} f(x) $ və $ \lim_{x \to a} g(x) $ mövcuddursa, o zaman:  
$$
\lim_{x \to a} [f(x) + g(x)] = \lim_{x \to a} f(x) + \lim_{x \to a} g(x)
$$

**Misal:**  
Funkisiyalar: $ f(x) = 2x $ və $ g(x) = x^2 $, $ a = 1 $.
$$
\lim_{x \to 1} \left( 2x + x^2 \right) = \lim_{x \to 1} 2x + \lim_{x \to 1} x^2 = 2(1) + 1^2 = 2 + 1 = 3
$$

---

### 2. **Hasilin limiti**

**Qayda:**  
Əgər $ \lim_{x \to a} f(x) $ və $ \lim_{x \to a} g(x) $ mövcuddursa, o zaman:  
$$
\lim_{x \to a} [f(x) \cdot g(x)] = \lim_{x \to a} f(x) \cdot \lim_{x \to a} g(x)
$$

**Misal:**  
Funkisiyalar: $ f(x) = x $ və $ g(x) = x + 1 $, $ a = 3 $.  
$$
\lim_{x \to 3} (x \cdot (x + 1)) = \lim_{x \to 3} x \cdot \lim_{x \to 3} (x + 1) = 3 \cdot (3 + 1) = 3 \cdot 4 = 12
$$

---

### 3. **Nisbinin limiti**

**Qayda:**  
Əgər $ \lim_{x \to a} f(x) $ və $ \lim_{x \to a} g(x) $ mövcuddursa və $ \lim_{x \to a} g(x) \neq 0 $, o zaman:  
$$
\lim_{x \to a} \frac{f(x)}{g(x)} = \frac{\lim_{x \to a} f(x)}{\lim_{x \to a} g(x)}
$$

**Misal:**  
Funkisiyalar: $ f(x) = x^2 + 2x $ və $ g(x) = x - 1 $, $ a = 2 $.  
$$
\lim_{x \to 2} \frac{x^2 + 2x}{x - 1} = \frac{\lim_{x \to 2} (x^2 + 2x)}{\lim_{x \to 2} (x - 1)} = \frac{2^2 + 2(2)}{2 - 1} = \frac{4 + 4}{1} = 8
$$

---

### 4. **L'Hopital Qaydası**

**Qayda:**  
Əgər limit $ 0/0 $ və ya $ \infty/\infty $ formasında olarsa, L'Hopital qaydasını tətbiq edirik:  
$$
\lim_{x \to a} \frac{f(x)}{g(x)} = \lim_{x \to a} \frac{f'(x)}{g'(x)}
$$

**Misal:**  
Funkisiyalar: $ f(x) = \sin x $ və $ g(x) = x $, $ a = 0 $.  

Bu funksiya $ \frac{0}{0} $ formasında olduğu üçün:  
$$
\lim_{x \to 0} \frac{\sin x}{x} = \lim_{x \to 0} \frac{\cos x}{1} = 1
$$

---

### 5. **Trigonometrik limitlər**

**Qayda:**  
Əsas trigonometrik limitlərdən biri:  
$$
\lim_{x \to 0} \frac{\sin x}{x} = 1
$$

**Misal:**  
Funksiya: $ f(x) = \frac{\sin x}{x} $, $ a = 0 $.  
$$
\lim_{x \to 0} \frac{\sin x}{x} = 1
$$

---

### 6. **Eksponensial və logaritmik limitlər**

**Qayda:**  
Eksponensial limitlərdən biri:  
$$
\lim_{x \to \infty} \left(1 + \frac{1}{x}\right)^x = e
$$

**Misal:**  
Funksiya: $ f(x) = \left( 1 + \frac{1}{x} \right)^x $, $ a = \infty $.  
$$
\lim_{x \to \infty} \left( 1 + \frac{1}{x} \right)^x = e
$$

---

### 7. **Sol və sağ limitlər**

**Qayda:**  
Sol və sağ limitlər:  
Əgər $ \lim_{x \to a^-} f(x) = \lim_{x \to a^+} f(x) = L $, o zaman limit mövcuddur.  

**Misal:**  
Funksiya:  
$$
f(x) = 
\begin{cases} 
x + 2, & x < 2 \\
3x, & x \geq 2
\end{cases}
$$

Sol limit ($ x \to 2^- $):  
$$
\lim_{x \to 2^-} (x + 2) = 4
$$

Sağ limit ($ x \to 2^+ $):  
$$
\lim_{x \to 2^+} 3x = 6
$$

Sol və sağ limitlər fərqli olduğu üçün ümumi limit mövcud deyil.

---

### 8. **Sonsuz limitlər**

**Qayda:**  
$$
\lim_{x \to \infty} \frac{1}{x} = 0
$$

**Misal:**  
Funksiya: $ f(x) = \frac{1}{x} $, $ a = \infty $.  
$$
\lim_{x \to \infty} \frac{1}{x} = 0
$$

## Müxtəlif növ Limitlər üçün Qaydalar və Nümunələr

Əlbəttə, aşağıda hər bir qayda üçün konkret misalların necə həll olunduğunu göstərdim:

---

### 1. **Ən sadə limit:**
$$ \lim_{x \to 3} 5 = 5 $$

Burada sabit bir ədəd olduğu üçün nəticə birbaşa 5-dir.

---

### 2. **Direkt əvəz etmə (Kontinuitet):**
$$ \lim_{x \to 2} (3x + 1) $$

Burada, birbaşa $x = 2$ ilə əvəz edirik:
$$ 3(2) + 1 = 6 + 1 = 7 $$

Beləliklə, limitin nəticəsi $7$-dir.

---

### 3. **Sonsuzluğa yaxınlaşan limitlər:**
$$ \lim_{x \to \infty} \frac{2x + 5}{x + 3} $$

Bu limitin həlli üçün ən böyük $x$-in dərəcəsini nəzərə alırıq. sürət və məxrəcin başlıca $x$-ləri var, ona görə hər iki hissədə $x$-i çıxarırıq:
$$ \lim_{x \to \infty} \frac{x(2 + \frac{5}{x})}{x(1 + \frac{3}{x})} = \frac{2 + 0}{1 + 0} = 2 $$

Beləliklə, nəticə $2$-dir.

---

### 4. **Sonsuzluğa yaxınlaşan limitlərdə sürət və məxrəci eyni dərəcəsi:**
$$ \lim_{x \to \infty} \frac{3x + 2}{4x + 5} $$

Burada həm sürət, həm də məxrəci n dərəcəsi 1-dir, yəni $x$-lər var. Ən böyük $x$-i çıxarırıq:
$$ \lim_{x \to \infty} \frac{3 + \frac{2}{x}}{4 + \frac{5}{x}} = \frac{3}{4} $$

Beləliklə, nəticə $\frac{3}{4}$-dür.

---
### 4.1 **Sıfıra yaxınlaşan limitlərdə sürət və məxrəci eyni dərəcəsi:**
$$ \lim_{x \to 0} \frac{3x + 2}{4x + 5} $$

Bu ifadəni $x \to 0$ üçün hesablayaq. Burada $x$-i birbaşa $0$ ilə əvəz edə bilərik, çünki ifadə müəyyənlikdə qalır:

$$ \frac{3(0) + 2}{4(0) + 5} = \frac{0 + 2}{0 + 5} = \frac{2}{5}. $$

Beləliklə, limitin nəticəsi:  
$$ \boxed{\frac{2}{5}}. $$
---

### 5. **Sonsuzluğa yaxınlaşan limitlərdə sürətin daha yüksək dərəcəsi:**
$$ \lim_{x \to \infty} \frac{x^2 + 3x + 5}{x + 2} $$

Burada sürətin dərəcəsi $2$, məxrəcin dərəcəsi isə $1$-dir. Buna görə də, limitin nəticəsi sonsuz olacaq:
$$ \lim_{x \to \infty} \frac{x^2}{x} = \infty $$

Beləliklə, nəticə sonsuzdur.

---

### 6. **Sonsuzluğa yaxınlaşan limitlərdə məxrəcin daha yüksək dərəcəsi:**
$$ \lim_{x \to \infty} \frac{x}{x^2 + 1} $$

Burada sürətin dərəcəsi $1$, məxrəcin dərəcəsi isə $2$-dir. Ona görə də, limit sıfıra yaxınlaşacaq:
$$ \lim_{x \to \infty} \frac{x}{x^2} = 0 $$

Beləliklə, nəticə $0$-dır.

---

### 7. **Sonsuzluğa yaxınlaşan limitlərdə  sürət və məxrəcin sıfır olması:**
$$ \lim_{x \to 0} \frac{x}{x^2 + 1} $$

Burada həm sürət, həm də məxrəcin qiyməti sıfır olduğu üçün birbaşa əvəz etmə ilə hesablaya bilərik:
$$ \frac{0}{0^2 + 1} = \frac{0}{1} = 0 $$

Beləliklə, nəticə $0$-dır.

---

### 8. **Modullu ifadənin limitləri:**
$$ \lim_{x \to 0^+} |x| = 0 \quad \text{və} \quad \lim_{x \to 0^-} |x| = 0 $$

Burada, $x = 0$ nöqtəsindən modullu ifadə sıfıra yaxınlaşır, buna görə də nəticə $0$-dır.

---

### 9. **İnteqral limitlər (İnteqralın yaxınlaşması):**
$$ \lim_{x \to \infty} \int_0^x e^{-t^2} \, dt $$

Bu inteqralın həlli birbaşa mümkün olmasa da, bilirik ki:
$$ \int_0^\infty e^{-t^2} \, dt = \frac{\sqrt{\pi}}{2} $$

Bu səbəbdən limitin nəticəsi $\frac{\sqrt{\pi}}{2}$-dir.

---

### 10. **L'Hopital qaydası:**
$$ \lim_{x \to 0} \frac{\sin(x)}{x} $$

Bu limitin nəticəsi $0/0$ qeyri-müəyyən formasında olduğu üçün L'Hopital qaydasını tətbiq edirik:
$$ \lim_{x \to 0} \frac{\sin(x)}{x} = \lim_{x \to 0} \frac{\cos(x)}{1} = 1 $$

Beləliklə, nəticə $1$-dir.

---

### 11. **Exponent və logaritm limitləri:**
$$ \lim_{x \to \infty} \frac{e^x}{x^2} $$

Burada $e^x$ daha sürətli böyüyür, ona görə də limitin nəticəsi sonsuz olacaq:
$$ \lim_{x \to \infty} \frac{e^x}{x^2} = \infty $$

---

### 12. **Trigonometric limitlər:**
$$ \lim_{x \to 0} \frac{\sin(2x)}{x} $$

Burada $\sin(2x)$ funksiyasının limitini tapmaq üçün $2x$-i dəyişdiririk:
$$ \lim_{x \to 0} \frac{\sin(2x)}{x} = 2 \cdot \lim_{x \to 0} \frac{\sin(2x)}{2x} = 2 \cdot 1 = 2 $$

Beləliklə, nəticə $2$-dir.

---

Bu misallar limitlər mövzusunda müxtəlif növ funksiyalar üçün tətbiq oluna biləcək ümumi qaydaları əhatə edir. Hər bir misal göstərir ki, müəyyən qaydalara əməl etməklə bu növ limitləri hesablamaq mümkündür.

## **L'Hôpital qaydası**

**L'Hôpital qaydası** qeyri-müəyyən limitləri həll etmək üçün istifadə olunan çox faydalı bir qaydadır. Bu qayda, aşağıdakı iki qeyri-müəyyən formalar üçün tətbiq edilir:

$$ 1) \ \frac{0}{0} $$
$$ 2) \ \frac{\infty}{\infty} $$

### Qayda:
Əgər bir limit $ \lim_{x \to a} \frac{f(x)}{g(x)} $ formasındadırsa və bu limit $\frac{0}{0} $ və ya $ \frac{\infty}{\infty} $ qeyri-müəyyənliyinə malikdirsə, o zaman belə həll olunur:

$$
\lim_{x \to a} \frac{f(x)}{g(x)} = \lim_{x \to a} \frac{f'(x)}{g'(x)}
$$

Burada $ f'(x) $ və $ g'(x) $ müvafiq olaraq $ f(x) $ və $ g(x) $ funksiyalarının törəmələridir.

---

### Şərtlər:
1. $ f(x) $ və $ g(x) $ funksiyaları verilmiş nöqtədə törəməli olmalıdır.
2. L'Hôpital qaydası bir neçə dəfə təkrar istifadə edilə bilər, əgər ilk tətbiqdən sonra yenə $ \frac{0}{0} $ və ya $ \frac{\infty}{\infty} $ qeyri-müəyyənliyi yaranarsa.

---

### Misallar:

#### Misal 1: $ \lim_{x \to 0} \frac{\sin(x)}{x} $
Bu ifadəni birbaşa əvəz etsək:
$$
\frac{\sin(0)}{0} = \frac{0}{0}.
$$
Bu, $\frac{0}{0} $ qeyri-müəyyənliyi olduğu üçün L'Hôpital qaydasını tətbiq edirik:
$$
f(x) = \sin(x), \quad f'(x) = \cos(x),
$$
$$
g(x) = x, \quad g'(x) = 1.
$$
İndi L'Hôpital tətbiq edirik:
$$
\lim_{x \to 0} \frac{\sin(x)}{x} = \lim_{x \to 0} \frac{\cos(x)}{1} = \cos(0) = 1.
$$
Beləliklə, nəticə $ 1 $-dir.

---

#### Misal 2: $ \lim_{x \to \infty} \frac{x^2}{e^x} $
Birbaşa əvəz etdikdə:
$$
\frac{\infty}{\infty}.
$$
Bu qeyri-müəyyənlikdir, ona görə L'Hôpital tətbiq edirik:
$$
f(x) = x^2, \quad f'(x) = 2x,
$$
$$
g(x) = e^x, \quad g'(x) = e^x.
$$
Tətbiq edək:
$$
\lim_{x \to \infty} \frac{x^2}{e^x} = \lim_{x \to \infty} \frac{2x}{e^x}.
$$
Bu yenə $ \frac{\infty}{\infty} $ formasında olduğuna görə bir dəfə də L'Hôpital tətbiq edirik:
$$
f(x) = 2x, \quad f'(x) = 2,
$$
$$
g(x) = e^x, \quad g'(x) = e^x.
$$
$$
\lim_{x \to \infty} \frac{2x}{e^x} = \lim_{x \to \infty} \frac{2}{e^x} = 0.
$$
Beləliklə, limit \$ 0 $-dır.

---

#### Misal 3:$ \lim_{x \to 0^+} \frac{\ln(x)}{\frac{1}{x}} $
Birbaşa əvəz etdikdə:
$$
\frac{-\infty}{\infty}.
$$
Bu da $ \frac{\infty}{\infty} $ formasındadır. L'Hôpital tətbiq edirik:
$$
f(x) = \ln(x), \quad f'(x) = \frac{1}{x},
$$
$$
g(x) = \frac{1}{x}, \quad g'(x) = -\frac{1}{x^2}.
$$
Tətbiq edirik:
$$
\lim_{x \to 0^+} \frac{\ln(x)}{\frac{1}{x}} = \lim_{x \to 0^+} \frac{\frac{1}{x}}{-\frac{1}{x^2}} = \lim_{x \to 0^+} -x = 0.
$$
Beləliklə, nəticə $ 0 $-dır.

---

Bu qayda ilə bağlı başqa sualınız varsa, soruşa bilərsiniz!

## Görkəmli Limitlər(1)

Şəkildəki "gözəl limitlər" haqqında izahlı məlumat:

---

### **I.Görkəmli Limitlər (Standart Limitlər)**:
1. $$
   \lim_{x \to 0} \frac{\sin x}{x} = 1
   $$  
   **İzah**: Bu, trigonometriyada əsas limitdir və kiçik bucaqlar üçün $\sin x \approx x$ ilə bağlıdır.
---
2.$$
   \lim_{x \to 0} \frac{\arcsin x}{x} = 1
   $$  
   **İzah**: $\arcsin x$ kiçik $|x|$-lər üçün $\sin x$-ə bənzəyir.

---

3. $$
   \lim_{x \to 0} \frac{\tan x}{x} = 1
   $$  
   **İzah**: $\tan x \approx x$, kiçik bucaqlar üçün doğrudur.

---

4. $$
   \lim_{x \to 0} \frac{\arctan x}{x} = 1
   $$  
   **İzah**: $\arctan x$ də kiçik $|x|$-lərdə xətti davranış göstərir.

---

5. $$
   \lim_{x \to 0} \frac{\sin\alpha x}{\beta x} = \frac{\alpha}{\beta}
   $$  
   **İzah**: $\sin(ax) \approx ax$ yaxınlaşmasını istifadə edirik.

---

6. $$
   \lim_{x \to 0} \frac{\sin\alpha x}{tan\beta x} = \frac{\alpha}{\beta}
   $$  
   **İzah**: Kiçik bucaqlar üçün $\tan(ax) \approx ax$ qəbul edilir.

---

### **II. Görkəmli Limitlər (Ekponential və Logaritmik)**:
1. $$
   \lim_{x \to \infty} \left(1 + \frac{1}{x}\right)^x = e
   $$
   və ya
   $$
   \lim_{x \to 0} \left(1 + x\right)^{\frac{1}{x}} = e
   $$
   **İzah**: Bu, $e$-nin tərifi ilə əlaqəlidir.

---

2. $$
   \lim_{x \to 0} \frac{\ln(1 + x)}{x} = 1
   $$  
   **İzah**: $\ln(1 + x)$ funksiyasının yaxınlaşmasını istifadə edirik ($\ln(1 + x) \approx x$).

---

3. $$
   \lim_{x \to 0} \frac{e^x - 1}{x}= 1
   $$  
   **İzah**: Burada $e^x$ funksiyasının Taylor açılımından istifadə edirik:  
   $e^x \approx 1 + x + \frac{x^2}{2!} + \dots$.
   $x \to 0$ zamanı yüksək dərəcəli $x$ ifadələri kiçik olduğu üçün $\frac{e^x - 1}{x} \approx \frac{(1 + x) - 1}{x} = 1$ olur.

---

4. $$
   \lim_{x \to 0} \frac{a^x - 1}{x} = \ln(a)
   $$
   **İzah**: Burada $a^x$ funksiyasının Taylor açılımından istifadə edilir:  
   $a^x \approx 1 + x \ln(a) + \frac{(x \ln(a))^2}{2!} + \dots$.  
   $x \to 0$ zamanı yüksək dərəcəli $x$ ifadələri təsirsiz qalır və $\frac{a^x - 1}{x} \approx \ln(a)$ olur.

---

5. $$
   \lim_{x \to 0} \frac{(1 + x)^\alpha - 1}{x} = \alpha
   $$  
   **İzah**: Binom açılımı ilə əldə edilir ($(1 + x)^d \approx 1 + dx$).

---

5. $$
   \lim_{x \to \infty} \left(1 + \frac{k}{x}\right)^x = e^k
   $$  
   **İzah**: Ekponential artımı göstərir.

---

6. $$
   \lim_{x \to 0} \frac{\sqrt[n]{1 + x} - 1}{x} = \frac{1}{n}
   $$  
   **İzah**: $\sqrt[n]{1+x}$ funksiyasının binomial yaxınlaşması $\sqrt[n]{1 + x} \approx 1 + \frac{x}{n}$istifadə edilərək, limit $\frac{1}{n}$ olaraq tapılır..

---

7. $$
   \lim_{x \to \infty} \left(1 + \frac{k}{x}\right)^{\alpha x} = e^{\alpha k}
   $$  
   **İzah**: Genişlənmiş ekponential limit formulu.

## $\lim_{x \to \infty} \frac{\sin x}{x} = 0$

Bu limit ifadəsini yazıram və qısa izahını təqdim edirəm:

**Mətn:**

$$
\lim_{x \to \infty} \frac{\sin x}{x} = 0, \quad \text{çünki } |\sin x| \leq 1.
$$

**İzah 1:**

- $\sin x$ məhdud  funksiyasının qiymətləri həmişə $[-1, 1]$) aralığında olur.
- Bölən $x$ isə sonsuz böyük olduğu üçün $\frac{\sin x}{x}$ ifadəsi getdikcə sıfıra yaxınlaşır.

Bu səbəbdən nəticə $0$-dır.


---

$$
\lim_{x \to \infty} \frac{\sin x}{x} = 0, \quad \text{çünki } |\sin x| \leq 1.
$$

**İzah 2:**

1. $\sin x$ məhdud funksiyadır, yəni onun qiymətləri həmişə $[-1, 1]\$ aralığında dəyişir.

2. $x \to \infty$ olduqda, $\frac{1}{x}$ sonsuz kiçilən funksiyadır, yəni getdikcə sıfıra yaxınlaşır.

3. Məhdud bir funksiyanın ($\sin x$) sonsuz kiçilən bir funksiyaya ($\frac{1}{x}$) hasili həmişə $0$-a yaxınlaşır. Buna görə də:

$$
\lim_{x \to \infty} \frac{\sin x}{x} = 0.
$$ 

Bu, limitin nəticəsini izah edir.

## Ekvivalentlik ilə Limit hesablanması(1)

Aşağıda hər bir ekvivalentlik üçün qısa izah təqdim edirəm:

---

$$
x \to 0 \text{ üçün:}
$$

1. **$\sin x \sim x$:**  
   Kiçik bucaqlar üçün sinus funksiyası ilə bucağın özünün dəyəri eynidir, çünki $\sin x \approx x$.
---

2. **\$\arcsin x \sim x$:**  
   Kiçik $x$-lər üçün $\arcsin x$ funksiyası təxminən $x$-ə bərabər olur, çünki tərs sinus da kiçik dəyərlərdə xətti davranır.
---

3. **$\tan x \sim x$:**  
   Kiçik bucaqlarda $\tan x$ təxminən $\sin x$ və $\sin x \sim x$ olduğundan, $\tan x \sim x$.
---

4. **$\ln(1 + x) \sim x$:**  
   Taylor açılımı ilə $\ln(1 + x) = x - \frac{x^2}{2} + \cdots$. Kiçik $x$-lər üçün yalnız $x$ dominant olur.
---

5. **$e^x - 1 \sim x$:**  
   Taylor açılımına görə, $e^x = 1 + x + \frac{x^2}{2} + \cdots$. Kiçik $x$-lər üçün $e^x - 1 \approx x$.
---

6. **$a^x - 1 \sim x \ln a \quad (a > 0, a \neq 1)$:**  
   Ümumi eksponent funksiyası üçün $a^x = 1 + x \ln a + \cdots$. Kiçik $x$-lərdə yalnız $x \ln a$ dominantdır.
---

7. **$(1 + x)^\alpha - 1 \sim \alpha x \quad (\alpha \in \mathbb{R})$:**  
   Binom açılımına görə, $(1 + x)^\alpha \approx 1 + \alpha x$. Kiçik $x$-lərdə yalnız $\alpha x$ dominantdır.
---

8. **$\sqrt[n]{1 + x} - 1 \sim \frac{x}{n}$:**  
   $n$-ci kökün Taylor açılımına görə, $\sqrt[n]{1 + x} \approx 1 + \frac{x}{n}$ (kiçik $x$-lər üçün).  
   Bu səbəbdən,$\sqrt[n]{1 + x} - 1 \approx \frac{x}{n}$.

---

Bu ekvivalentliklər, kiçik $x$-lərdə funksiyaların davranışını anlamaq üçün vacibdir və limit hesablama, təxmini analiz kimi sahələrdə tətbiq olunur.

---

**Misallar**

Sizin verdiyiniz tənlikləri addım-addım təhlil edək:

1. **Limitin başlanğıcı:**
   $$
   \lim_{x \to 0} \frac{\ln(\cos x)}{\tan^2 x} = \frac{2 \ln(\cos x)}{2 \tan^2 x} = \frac{\ln(\cos^2 x)}{2 \tan^2 x} =
        \frac{-x^2}{2 x^2} = -\frac{1}{2}
   $$



2. **İzah:**
   $$\ln(\cos^2 x) = \ln(1 - \sin^2 x) \approx \ln(- \sin^2 x + 1) \approx -\sin^2 x \approx -x^2$$

   $$
   2 \tan^2 x \approx 2 x^2
   $$

3. **Son nəticə:**
   Nəticədə:
   $$
   \frac{-x^2}{2 x^2} = -\frac{1}{2}
   $$

Beləliklə, tənlik $-\frac{1}{2}$ nəticəsini verir.
