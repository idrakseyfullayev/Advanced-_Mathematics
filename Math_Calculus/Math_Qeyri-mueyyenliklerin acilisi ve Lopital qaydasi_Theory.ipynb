## Qeyri-müəyyənliklərin açılışı və Lopital Qaydası(1)

### Qeyri-müəyyənliklərin açılışı

**Qeyri-müəyyənlik** anlayışı, riyaziyyatda müəyyən limit ifadələrində hesablama zamanı yaranan nəticələrin "naməlum" və ya "açıq şəkildə həll edilə bilməz" olduğu halları təsvir edir. Əsas qeyri-müəyyənlik formaları bunlardır:

- $ \frac{0}{0} $: Bölmə, həm payın, həm də məxrəcinin 0-a yaxınlaşdığı halda.
- $ \frac{\infty}{\infty} $: Sonsuzluq bölü sonsuzluq.
- $ \frac{\infty}{0} $: Sonsuzluq bölü 0.
- $ 0 \cdot \infty $: Sonsuzluğun sıfırla çarpımı.
- $ \infty - \infty $: Sonsuzluğun sonsuzluqdan çıxılması.
- $ 1^\infty $, $ 0^0 $, və $ \infty^0 $: Güc funksiyası ilə bağlı qeyri-müəyyənliklər.

Bu qeyri-müəyyənlikləri aradan qaldırmaq üçün müxtəlif üsullar istifadə olunur və ən çox istifadə olunanlardan biri **Lopital qaydasıdır**.

---

## Lopital Qaydası

**Lopital qaydası**, $ \frac{0}{0} $ və $ \frac{\infty}{\infty} $ tipli limitlərdə tətbiq olunur. Bu qayda aşağıdakı şəkildə ifadə olunur:

#### Qayda:
Əgər $ \lim_{x \to c} f(x) = 0 $ və $ \lim_{x \to c} g(x) = 0 $ (və ya $ \pm\infty $), onda
$$
\lim_{x \to c} \frac{f(x)}{g(x)} = \lim_{x \to c} \frac{f'(x)}{g'(x)},
$$
əgər sağ tərəfdəki limit mövcuddursa.

---

### Məsələnin Həlli

#### Misal 1: $ \lim_{x \to 0} \frac{\sin x}{x} $
Bu halda $ \frac{\sin x}{x} $ $ \frac{0}{0} $ qeyri-müəyyənlik forması yaradır. Lopital qaydasına görə:
$$
\lim_{x \to 0} \frac{\sin x}{x} = \lim_{x \to 0} \frac{\cos x}{1} = \cos(0) = 1.
$$

#### Misal 2: $ \lim_{x \to \infty} \frac{x}{e^x} $
Bu halda $ \frac{x}{e^x} $ $ \frac{\infty}{\infty} $ forması yaradır. Lopital qaydasını tətbiq edək:
$$
\lim_{x \to \infty} \frac{x}{e^x} = \lim_{x \to \infty} \frac{1}{e^x} = 0.
$$

#### Misal 3: $ \lim_{x \to 0^+} x \ln x $
Bu ifadəni $ \frac{\ln x}{1/x} $ şəklinə salırıq. İndi $ \frac{\ln x}{1/x} $ ifadəsi $ \frac{-\infty}{\infty} $ formasını alır. Lopital qaydasını tətbiq edək:
$$
\lim_{x \to 0^+} x \ln x = \lim_{x \to 0^+} \frac{\ln x}{1/x} = \lim_{x \to 0^+} \frac{\frac{1}{x}}{-\frac{1}{x^2}} = \lim_{x \to 0^+} -x = 0.
$$

---

### Qeyd:
- **Lopital qaydası yalnız** $ \frac{0}{0} $ və $ \frac{\infty}{\infty} $ tipli qeyri-müəyyənliklərdə istifadə edilə bilər.
- Digər qeyri-müəyyənlik formalarını həll etmək üçün riyazi çevirmə, Taylor sırası və ya logaritmik xassələr kimi metodlar tətbiq olunur.

## Limitdə $ \frac{0}{\infty} $ formasını əldə edirsinizsə, bu nəticə **həmişə 0-a bərabərdir**

Bəli, əgər bir limitdə $ \frac{0}{\infty} $ formasını əldə edirsinizsə, bu nəticə **həmişə 0-a bərabərdir**. Bunun səbəbi aşağıdakı riyazi məntiqlə izah olunur:

---

### Riyazi İzah:

1. **Pay 0-dır:**
   - $ 0 $-ı hər hansı bir müsbət və ya mənfi sonlu və ya sonsuz böyük ədədlə bölsək, nəticə **həmişə 0** olur:
   $$
   \frac{0}{x} = 0 \quad \text{(burada $ x \neq 0 $)}.
   $$

2. **Məxrəc sonsuz böyükdür:**
   - Məxrəc sonsuz böyük olduqda (yəni $ x \to \infty $), pay 0 olduğu üçün nəticə $ \frac{0}{\infty} = 0 $ olur. Bu, o deməkdir ki, sonsuz böyük bir ədəddəki "sıfır" çox kiçikdir.

---

### Məsələn:
#### Misal 1:
$$
\lim_{x \to \infty} \frac{0}{x} = 0.
$$

#### Misal 2:
$$
\lim_{x \to \infty} \frac{5 \cdot 0}{e^x} = \frac{0}{\infty} = 0.
$$

---

### Nəticə:
Əgər limitdə $ \frac{0}{\infty} $ kimi forma yaranırsa, bu nəticə **0**-a bərabərdir. Bu forma qeyri-müəyyənlik deyil, çünki nəticə aydın şəkildə müəyyən olunub.
