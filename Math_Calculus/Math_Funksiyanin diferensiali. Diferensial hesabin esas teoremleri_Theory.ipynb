## Diferensial 

### **Diferensialın Tərifi**  

> **Diferensial**, bir funksiyanın müstəqil dəyişənindəki kiçik bir dəyişməyə görə asılı dəyişəndəki dəyişmənin təqribi qiymətləndirilməsini ifadə edir. Riyazi olaraq, diferensial aşağıdakı kimi təyin olunur:  

$$
dy = f'(x) \cdot dx
$$  

Burada:  
- $dy$: Funksiyanın asılı dəyişənindəki təqribi dəyişmə (diferensial).  
- $f'(x)$: Funksiyanın törəməsi (bu, $x$ nöqtəsindəki yamacı göstərir).  
- $dx$: Müstəqil dəyişənin kiçik bir dəyişməsi.

---

### **Diferensialın İzahı**  
Diferensial, bir funksiyanın $x$-dəki dəyişməsinə ($dx$) əsaslanaraq, $y$-dəki ($dy$) dəyişməni təxmin etmək üçün istifadə edilir. Funksiyanın qrafiki üzərində bir nöqtədən yaxın bir nöqtəyə qədərki kiçik bir dəyişməni təsvir edir.

---

### **Geometrik Baxış**  
Diferensial, bir funksiyanın toxunan xəttinin (tangent line) $\Delta x = dx$ qədər bir dəyişmə üçün $y$ oxundakı dəyişməni ($dy$) təxmin etmək üçün istifadə olunur. 

Funksiyanın real dəyişməsi ilə ($\Delta y$) diferensial arasında kiçik fərq ola bilər, lakin $dx \to 0$ olduqda diferensial bu real dəyişməyə yaxınlaşır:

$$
\Delta y \approx dy
$$

---

### **Misal:**
Verilmiş $y = x^2$ funksiyası üçün:  
- Törəmə: $f'(x) = 2x$  
- Diferensial: $dy = 2x \cdot dx$  

Əgər $x = 3$ və $dx = 0.1$ olsa:  
- $dy = 2 \cdot 3 \cdot 0.1 = 0.6$  

Bu halda $x$-dəki $0.1$ qədər dəyişiklik üçün $y$-də təqribən $0.6$ qədər dəyişiklik olur.  

Diferensial, funksiyanın xətti yaxınlaşması ilə dəyişməni təsvir edir.

---

### **Törəmə nədir?**  
Törəmə bir funksiyanın müəyyən bir nöqtədəki dəyişmə sürətini və ya meylini ifadə edir. Yəni, bir funksiyanın hər bir nöqtədəki "təpə və ya yamac" meyilliyini hesablayır.

Məsələn:  
- $f'(x)$ bir funksiyanın $f(x)$-nin törəməsidir.  
- Riyazi olaraq:  
  $$ 
  f'(x) = \lim_{h \to 0} \frac{f(x+h) - f(x)}{h} 
  $$

---

### **Diferensial nədir?**  
Diferensial törəmənin **praktiki tətbiqi** və ya **təqribi dəyişimi** ifadə edir. Törəmə daha nəzəri bir anlayışdır, diferensial isə onun konkret bir dəyişiklikdə istifadəsidir.

- $dy$, diferensial olaraq adlandırılır və aşağıdakı kimi təyin edilir:  
  $$ 
  dy = f'(x) \cdot dx 
  $$  
Burada:  
  - $dx$: Müstəqil dəyişənin kiçik bir dəyişməsi.  
  - $dy$: Funksiyanın kiçik dəyişməsi.  

---

## Törəmə və Diferensialın Fərqi

### **Fərqləri necə başa düşmək olar?**  
1. **Törəmə:** Meylin və ya dəyişmə sürətinin hesablanmasıdır. (Teorik tərəf)  
   - Məsələn, bir qrafikdə müəyyən nöqtədəki yamac.  
   
2. **Diferensial:** Bu yamacdan istifadə edərək, kiçik dəyişikliklər üçün funksiyadakı dəyişməni tapmaqdır. (Tətbiqi tərəf)  
   - Məsələn, $dx$ qədər bir dəyişmə olarsa, $dy$-ni təqribən hesablamaq.  

---

### Məsələn:  
Verilmiş $y = x^2$ funksiyasını götürək:  
- Törəmə: $f'(x) = 2x$.  
- Diferensial: $dy = 2x \cdot dx$.  

Əgər $x = 3$ və $dx = 0.1$ olsa:  
- $dy = 2 \cdot 3 \cdot 0.1 = 0.6$.  

Bu diferensial $x$-dəki kiçik dəyişiklik nəticəsində $y$-dəki təqribi dəyişməni ifadə edir.

---

### Nəticə:  
- **Törəmə** - bir funksiyanın dəyişmə sürətinin riyazi təyinidir.  
- **Diferensial** - bu sürətə əsasən funksiyanın dəyişməsinin təqribi qiymətləndirilməsidir.  

Hər ikisi riyaziyyatın analitik bölməsində çox əhəmiyyətli anlayışlardır.  

## $x^2$ Funksiyasının  x=-2  noqtesində  Törəməsi və Diferensialı

Gəlin verilənləri istifadə edərək **diferensialı** hesablayaq:  

### Verilənlər:  
- Funksiya: $y = x^2$  
- Törəmə: $y' = 2x$  
- Diferensial: $dy = 2x \cdot dx$  
- $x = -2$ və $dx$ — kiçik bir dəyişiklik.  

### Törəmə nöqtədə:  
Törəmə sadəcə $x$-dəki dəyişmə sürətini göstərir:  
$$
y' = 2x = 2 \cdot (-2) = -4
$$  

### Diferensial:  
Diferensial üçün $dx$-in konkret dəyərini bilməliyik. Məsələn, əgər $dx = 0.1$ olarsa:  
$$
dy = 2x \cdot dx = 2 \cdot (-2) \cdot 0.1 = -0.4
$$  

### Yekun:  
- **Törəmə:** $-4$  
- **Diferensial:** $-0.4$, əgər $dx = 0.1$.  

**Qeyd:** Diferensialın dəqiq dəyəri üçün $dx$-i siz seçməlisiniz. $dx$-in kiçik dəyişiklik olduğunu unutmayın!

## Funksiyanın diferensialının tərifi (1)

Funksiyanın diferensialı anlayışı, funksiya ilə onun yaxınlaşmasının xətti bir təxmini arasındakı əlaqəni ifadə edir. **Funksiyanın diferensialı** bir nöqtədə funksiyanın kiçik bir dəyişməsini xətti təxmini ilə hesablamağa imkan verir.

### Funksiyanın diferensialının tərifi:

> **Tərif** : Əgər $ y = f(x) $ funksiyası diferensiallanan bir funksiya olarsa, onda onun diferensialı $ dx $ (müstəqil dəyişən $ x $-in kiçik dəyişməsi) ilə $ dy $-yə (asılı dəyişənin kiçik dəyişməsi) belə ifadə olunur:

$$
dy = f'(x) \, dx
$$

Burada:
- $ f'(x) $: $ f(x) $-in törəməsidir,
- $dx $: $ x $-dəki kiçik dəyişmədir,
- $ dy $: funksiyanın bu dəyişmə nəticəsindəki yaxınlaşan dəyişməsidir.

---

### Əsas məqamlar:

1. **Diferensial xətti təxmindir**:
   Diferensial, $ f(x) $-in kiçik dəyişmələri üçün xətti təxmini verir:
   $$
   \Delta y \approx dy = f'(x) \, dx
   $$

2. **Kiçik dəyişmələr**:
   $ dx $ sərbəst şəkildə seçilə bilən kiçik dəyişməni ifadə edir, $ dy $ isə bu dəyişmə nəticəsində yaranan xətti təxmini ifadə edir.

3. **Qrafik mənası**:
   Diferensial, funksiyanın müəyyən nöqtədəki toxunan xəttinin təxmini dəyişməsini təsvir edir.

---

### Nümunə:

$ y = x^2 $ funksiyasını nəzərdən keçirək. 

1. **Törəməni hesablayırıq**:
   $$
   f'(x) = 2x
   $$

2. **Diferensial**:
   $$
   dy = f'(x) \, dx = 2x \, dx
   $$

Əgər $ x = 3 $-də kiçik bir dəyişmə $dx = 0.1 $ olarsa, onda:
$$
dy = 2(3)(0.1) = 0.6
$$

Bu, $ x $-dəki kiçik dəyişməyə əsasən $ y $-dəki xətti dəyişməni göstərir.

---

### Nəticə:
Funksiyanın diferensialı, dəyişənlərin kiçik dəyişmələri zamanı funksiyanın yaxınlaşan xətti dəyişməsini hesablamaq üçün əsas alətdir. Diferensial, xüsusilə, hesablama və analitik tətbiqlərdə geniş istifadə olunur.

## Diferensialın Həndəsi Mənası(1)

**Diferensialın** həndəsi mənası, funksiyanın **kiçik dəyişikliklərə** qarşı reaksiyasını və **təqribi dəyişiklikləri** görselleştirməyə imkan verir. Diferensialın həndəsi mənası daha çox **tangent xətti** ilə əlaqələndirilir. Gəlin, bu anlayışı daha detallı şəkildə izah edək:

---

### **Diferensialın Həndəsi Mənası:**

1. **Tangent Xətti və Yaxınlaşma:**
   Diferensial $dy$, bir funksiyanın nöqtə üzərindəki **tangent xətti** ilə əlaqəlidir. Tangent xətti funksiyanın həmin nöqtədəki **meylini** (törəməsini) təmsil edir.

   Məsələn, $y = f(x)$ funksiyasının bir nöqtədəki **tangent xətti** verilən nöqtədə $(x_0, y_0)$ nöqtəsi üzərində $f'(x_0)$ meyli ilə belə ifadə olunur:
   $$
   y - y_0 = f'(x_0) (x - x_0)
   $$
   Bu ifadə tangent xəttinin tənliyidir. Diferensial $dy$, bu tangent xəttinin **yüksəlişini** (yəni $y$-dəki kiçik dəyişməni) göstərir.

2. **Kiçik Dəyişikliklər:**
   Diferensial $dy$, $x$-dəki kiçik bir dəyişiklik $dx$-in **funksiyadakı** təqribi təsirini göstərir. Yəni, bir funksiyanın $x$-dəki kiçik dəyişməsi, $y$-də necə bir dəyişiklik yaradacaqsa, bu dəyişiklik $dy$-nin qiyməti ilə təxmin edilir.

   $$
   dy = f'(x) \cdot dx
   $$

   Həndəsi olaraq bu, funksiyanın **qrafikində** müəyyən bir nöqtənin ətrafında **kiçik bir xətti yaxınlaşma** yaradır. Əgər $dx$-i çox kiçik seçsək, bu xətti yaxınlaşma **funksiyanın real dəyişməsi** ilə çox yaxınlaşacaqdır.

3. **Diferensialın Vizual Təsviri:**
   Təsəvvür edin ki, $y = x^2$ funksiyasının qrafiki üzərində $x = 2$ nöqtəsində bir dəyişiklik ediriksə:
   - Bu nöqtədə $y' = 4$ (törəmə) və $dx = 0.1$ (kiçik dəyişiklik).
   - Diferensial $dy = 4 \cdot 0.1 = 0.4$ olacaq.
   
   Bu, deməkdir ki, $x = 2$-də $0.1$ kiçik dəyişikliklə $y$-də təxminən $0.4$ dəyişiklik baş verəcəkdir. Bu, qrafikdəki tangent xəttinin **yüksəlişini** təmsil edir.

---

### **Qısa Həndəsi İzah:**
- **Diferensial**, **tangent xəttinin** yuxarı və ya aşağıya necə hərəkət etdiyini göstərir.
- Funksiyanın nöqtə üzərindəki **meylini** və **təqribi dəyişikliklərini** ehtiva edir.
- **Kiçik dəyişikliklər** ilə daha böyük dəyişikliklər arasında **yaxınlaşma** təmin edir.

Bu həndəsi mənada, diferensial bir nöqtədəki kiçik dəyişiklikləri hesablamaq və qrafikin xüsusiyyətlərini anlamaq üçün istifadə olunur.

## $dy = f'(x) \cdot dx$  necə alınır?

Diferensialın tənliyi $dy = f'(x) \cdot dx$-ni daha riyazi şəkildə izah etmək üçün, bunun necə alındığını addım-addım göstərək:

---

### **1. Funksiya və Törəmə:**
Əvvəlcə, bir funksiyanı götürək:
$$
y = f(x)
$$
Bu funksiya $x$-ə bağlı olaraq dəyişir. Bizim məqsədimiz $x$-in **kiçik dəyişikliklər** nəticəsində $y$-dəki dəyişməni tapmaqdır.

Funksiyanın törəməsi (yəni dəyişmə sürəti) $f'(x)$ ilə veriləcək. Törəmə funksiyanı $x$-in dəyişməsinə görə necə dəyişdiyini göstərir.

Törəmə tərifi belədir:
$$
f'(x) = \lim_{h \to 0} \frac{f(x+h) - f(x)}{h}
$$

---

### **2. Kiçik Dəyişiklik:**
Burada $h$ kiçik bir dəyişiklikdir. Biz $h$-i $dx$ ilə əvəzləyirik, çünki **kiçik dəyişiklik** olaraq $dx$-i götürürük. Buna görə də, fərqli bir şəkildə yazırıq:

$$
f'(x) = \lim_{dx \to 0} \frac{f(x + dx) - f(x)}{dx}
$$

Bu tənlik, $f(x)$-in **kiçik dəyişikliklərə qarşı reaksiyasını** göstərir.

---

### **3. Kiçik Dəyişməni Tapmaq:**
Daha sonra, biz **kiçik bir dəyişiklik** olan $dx$-i tətbiq edirik və $y$-dəki **kiçik dəyişməni** (yəni $dy$) tapırıq. Bu zaman:

$$
f(x + dx) - f(x) = f'(x) \cdot dx
$$

**Burda $f(x + dx) - f(x) = f'(x) = dy$**

$$
dy = f'(x) \cdot dx
$$

Bu tənlik, funksiyanın $x$-dəki kiçik dəyişiklik $dx$-ə qarşı olan **kiçik dəyişməni** göstərir.

---

### **4. Riyazi İzah:**
Əgər biz konkret bir funksiyanı götürsək, məsələn $y = f(x) = x^2$, onda:
- Törəmə $f'(x) = 2x$.
- Diferensial $dy = 2x \cdot dx$.

Bu, $x$-dəki kiçik dəyişikliklərlə əlaqəli olaraq, $y$-dəki dəyişməni tapmağın bir üsuludur.

---

### **Nəticə:**
Diferensialın tənliyi $dy = f'(x) \cdot dx$ **törəmənin** və **kiçik dəyişikliklərin** çarpanıdır və bu tənlik **kiçik dəyişikliklər nəticəsində funksiyada baş verən dəyişiklikləri** təqribən hesablamağa kömək edir.

## Yüksək Tərtibli Diferensiallar(1)

### **Yüksək Tərtibli Diferensiallar Nədir?**  

Yüksək tərtibli diferensiallar, bir funksiyanın birinci tərtib diferensialından sonra daha yüksək tərtibdəki diferensialları ifadə edir. Bu anlayış, funksiyanın yalnız dəyişmə sürətini deyil, həm də bu sürətin necə dəyişdiyini təhlil etmək üçün istifadə olunur.  

---

### **Təriflər və İfadələr**  

Bir funksiyanın yüksək tərtibli diferensialları aşağıdakı kimi göstərilir:  

1. **Birinci tərtib diferensial:**  
   Bu, funksiyanın dəyişmə sürətini ifadə edir:  
   $$
   dy = f'(x)dx
   $$  

2. **İkinci tərtib diferensial:**  
   Funksiyanın dəyişmə sürətinin necə dəyişdiyini göstərir:  
   $$
   d^2y = f''(x)dx^2
   $$  

3. **Üçüncü tərtib diferensial:**  
   Dəyişmənin üçüncü səviyyədə necə baş verdiyini ifadə edir:  
   $$
   d^3y = f'''(x)dx^3
   $$  

4. **n-ci tərtib diferensial:**  
   Ümumi halda, yüksək tərtib diferensial:  
   $$
   d^n y = f^{(n)}(x)dx^n
   $$  

---

### **Misallar:**  

1. **Funksiya: \( f(x) = x^3 \)**  
   - Birinci tərtib diferensial: $$ dy = 3x^2 dx $$  
   - İkinci tərtib diferensial: $$ d^2y = 6x dx^2 $$  
   - Üçüncü tərtib diferensial: $$ d^3y = 6 dx^3 $$  
   - Dördüncü tərtib diferensial: $$ d^4y = 0 $$  

2. **Funksiya: \( f(x) = \sin(x) \)**  
   - Birinci tərtib diferensial: $$ dy = \cos(x) dx $$  
   - İkinci tərtib diferensial: $$ d^2y = -\sin(x) dx^2 $$  
   - Üçüncü tərtib diferensial: $$ d^3y = -\cos(x) dx^3 $$  
   - Dördüncü tərtib diferensial: $$ d^4y = \sin(x) dx^4 $$  

---

### **Tətbiqlər:**  

1. **Mexanika:**  
   - Birinci tərtib diferensial: Sürət ($ v $).  
   - İkinci tərtib diferensial: Sürətlənmə ($ a $).  
   - Üçüncü tərtib diferensial: Zərbə ($ j $).  

2. **Optimallaşdırma:**  
   Yüksək tərtibli törəmələrdən istifadə edərək funksiyanın kritik nöqtələrinin təbiəti təyin edilir (minimum, maksimum və ya infleksiya nöqtəsi).  

3. **Diferensial Tənliklər:**  
   Mürəkkəb sistemlərin modelləşdirilməsində yüksək tərtibli diferensial tənliklər geniş istifadə olunur.

## Diferensial hesabının əsas teoremləri(1)

Başqa şərhə ehtiyac olmadan düzəldirəm:  

### **1. Rolle Teoremi**  
**Təlimat:**  
Əgər bir $f(x)$ funksiyası aşağıdakı şərtləri təmin edirsə:  
- Funksiya $[a, b]$-də kəsilməzdir.  
- Funksiya $(a, b)$-də törəməlidir.  
- $f(a) = f(b)$.  

O zaman, ən azı bir $c \in (a, b)$ nöqtəsi mövcuddur ki:  
$$
f'(c) = 0
$$  

**Təsviri:**  
Bu, göstərir ki, belə bir funksiya $[a, b]$-də ən azı bir sabit tangent xətti (üfüqi xətt) var.  

---

### **2. Orta Qiymət Teoremi (Mean Value Theorem - MVT)**  
**Təlimat:**  
Əgər $f(x)$:  
- $[a, b]$-də kəsilməzdir,  
- $(a, b)$-də törəməlidir.  

O zaman ən azı bir $c \in (a, b)$ nöqtəsi mövcuddur ki:  
$$
f'(c) = \frac{f(b) - f(a)}{b - a}
$$  

**Təsviri:**  
Funksiyanın müəyyən bir intervallarda orta dəyişmə sürəti ən azı bir nöqtədəki anlıq dəyişmə sürətinə bərabərdir.  

---

### **3. L'Hospital Qaydası**  
**Təlimat:**  
Əgər $f(x)$ və $g(x)$ funksiyaları aşağıdakı şərtləri təmin edirsə:  
- $\lim_{x \to c} f(x) = 0$ və $\lim_{x \to c} g(x) = 0$ (və ya $\pm \infty / \pm \infty$ forması),  
- $g'(x) \neq 0$ intervallar daxilində,  

O zaman:  
$$
\lim_{x \to c} \frac{f(x)}{g(x)} = \lim_{x \to c} \frac{f'(x)}{g'(x)}
$$  
şərti ilə sağ tərəfdəki limit mövcuddur.  

**Təsviri:**  
Bu teorem müəyyən mürəkkəb limitlər hesablanarkən istifadə edilir.  

---

### **4. Fermat Teoremi**  
**Təlimat:**  
Əgər $f(x)$ funksiyasının $c$ nöqtəsində maksimum və ya minimum qiyməti varsa və $f'(c)$ mövcuddursa, o zaman:  
$$
f'(c) = 0
$$  

**Təsviri:**  
Maksimum və ya minimum nöqtələrdə tangent xətti üfüqi olur.  

---

### **5. Taylor Teoremi**  
**Təlimat:**  
Əgər $f(x)$ funksiyası $n$-ci dərəcəyə qədər törəməlidirsə, o zaman $f(x)$ funksiyasını $a$-nın yaxınlığında belə ifadə etmək olar:  
$$
f(x) = f(a) + f'(a)(x-a) + \frac{f''(a)}{2!}(x-a)^2 + \cdots + \frac{f^{(n)}(a)}{n!}(x-a)^n + R_n(x)
$$  
Burada $R_n(x)$ qalıq terminidir.  

**Təsviri:**  
Bu teorem bir funksiyanı çoxhədli ilə yaxınlaşdırmağa imkan verir.  

---

### **6. Fundamental Teorem (Differensial-İnteqral Bağlantısı)**  
Differensial hesab və inteqral hesab arasında əsas əlaqəni göstərir:  
$$
\frac{d}{dx} \int_a^x f(t) dt = f(x)
$$
